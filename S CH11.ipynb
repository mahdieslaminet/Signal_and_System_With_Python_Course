# انواع سیگنال (Types of Signals)

سیگنال‌ها نمایش‌دهنده‌ی اطلاعات در قالب داده‌ها یا تغییرات در طول زمان هستند. آنها نقش حیاتی در سیستم‌های مخابراتی، کنترلی، پردازش داده و بسیاری از زمینه‌های دیگر ایفا می‌کنند. سیگنال‌ها را می‌توان بر اساس ویژگی‌های مختلف دسته‌بندی کرد. در اینجا به برخی از دسته‌بندی‌های مهم می‌پردازیم:

## ۱. سیگنال‌های آنالوگ و دیجیتال (Analog and Digital Signals)

*   **سیگنال آنالوگ:** سیگنالی پیوسته در زمان و مقدار است. به این معنی که مقدار آن در هر لحظه از زمان تعریف شده و می‌تواند هر مقداری در یک بازه مشخص را بگیرد.
    *   **مثال:** امواج صوتی، سیگنال‌های ولتاژ پیوسته، سیگنال دما.
    *   **نمایش:** نمودار آنالوگ به صورت خطی پیوسته رسم می‌شود.

*   **سیگنال دیجیتال:** سیگنالی گسسته در زمان و مقدار است. به این معنی که مقدار آن فقط در لحظات گسسته از زمان تعریف شده و فقط می‌تواند مقادیر محدودی (معمولاً صفر و یک) را بگیرد.
    *   **مثال:** داده‌های باینری در کامپیوترها، سیگنال‌های نمونه‌برداری شده و کوانتیزه شده.
    *   **نمایش:** نمودار دیجیتال به صورت پله‌ای و گسسته رسم می‌شود.

## ۲. سیگنال‌های پیوسته و گسسته در زمان (Continuous-time and Discrete-time Signals)

*   **سیگنال پیوسته در زمان:** سیگنالی است که در هر لحظه از زمان تعریف شده است.
    *   **نمایش ریاضی:** \(x(t)\) که \(t\) متغیر پیوسته است.

*   **سیگنال گسسته در زمان:** سیگنالی است که فقط در لحظات گسسته از زمان تعریف شده است.
    *   **نمایش ریاضی:** \(x[n]\) که \(n\) یک عدد صحیح است.

## ۳. سیگنال‌های متناوب و نامتناوب (Periodic and Aperiodic Signals)

*   **سیگنال متناوب:** سیگنالی است که پس از گذشت یک بازه زمانی مشخص (دوره تناوب \(T\)) دوباره تکرار می‌شود.
    *   **شرط تناوب:** \(x(t) = x(t + T)\) برای همه \(t\).
    *   **مثال:** توابع سینوسی و کسینوسی.

*   **سیگنال نامتناوب:** سیگنالی است که هیچ الگوی تکراری ندارد.
    *   **مثال:** سیگنال‌های نمایی میراشونده، سیگنال‌های تصادفی.

## ۴. سیگنال‌های قطعی و تصادفی (Deterministic and Random Signals)

*   **سیگنال قطعی:** سیگنالی است که رفتار آن کاملاً قابل پیش‌بینی است و می‌توان مقدار آن را در هر لحظه از زمان با استفاده از یک فرمول ریاضی تعیین کرد.
    *   **مثال:** \(x(t) = 2t + 3\)

*   **سیگنال تصادفی:** سیگنالی است که رفتار آن قابل پیش‌بینی نیست و با استفاده از احتمالات و آمار توصیف می‌شود.
    *   **مثال:** نویز حرارتی در مدارهای الکترونیکی.



## 5. سیگنال‌های یک‌بعدی، دوبعدی و چندبعدی (One-dimensional, Two-dimensional, and Multi-dimensional Signals)

*   **سیگنال یک‌بعدی:** تابعی از یک متغیر مستقل (معمولاً زمان).
    *   **مثال:** سیگنال صوتی \(x(t)\).

*   **سیگنال دوبعدی:** تابعی از دو متغیر مستقل (مانند مختصات مکانی).
    *   **مثال:** تصویر \(I(x, y)\).

*   **سیگنال چندبعدی:** تابعی از بیش از دو متغیر مستقل.
    *   **مثال:** ویدیو \(V(x, y, t)\) که در آن \(x\) و \(y\) مختصات مکانی و \(t\) زمان است.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Analog Signal
# An analog signal is continuous in time and amplitude.
def generate_analog_signal(frequency=5, duration=1, sample_rate=1000):
    time = np.linspace(0, duration, int(sample_rate * duration))
    signal = np.sin(2 * np.pi * frequency * time)
    return time, signal

# 2. Digital Signal
# A digital signal is discrete in both time and amplitude.
def generate_digital_signal(bit_sequence=[1, 0, 1, 1, 0], bit_duration=0.2):
    time = np.arange(0, len(bit_sequence) * bit_duration, 0.01)
    signal = np.repeat(bit_sequence, int(bit_duration / 0.01))
    return time[:len(signal)], signal

# 3. Continuous-Time, Discrete-Amplitude Signal
# This type of signal is continuous in time but has discrete amplitude levels.
def generate_quantized_signal(frequency=5, levels=4, duration=1, sample_rate=1000):
    time, analog_signal = generate_analog_signal(frequency, duration, sample_rate)
    step = 2 / levels
    quantized_signal = np.round(analog_signal / step) * step
    return time, quantized_signal

# 4. Discrete-Time, Continuous-Amplitude Signal
# This type of signal is sampled in time but continuous in amplitude.
def generate_sampled_signal(frequency=5, duration=1, sample_rate=100):
    time = np.linspace(0, duration, sample_rate)
    signal = np.sin(2 * np.pi * frequency * time)
    return time, signal

# 5. Deterministic and Random Signals
# A deterministic signal is predictable, while a random signal is not.
def generate_random_signal(duration=1, sample_rate=1000):
    time = np.linspace(0, duration, int(sample_rate * duration))
    signal = np.random.normal(0, 1, len(time))
    return time, signal

# Visualization of signals
if __name__ == "__main__":
    plt.figure(figsize=(15, 10))

    # Analog Signal
    plt.subplot(3, 2, 1)
    t, s = generate_analog_signal()
    plt.plot(t, s)
    plt.title("Analog Signal")

    # Digital Signal
    plt.subplot(3, 2, 2)
    t, s = generate_digital_signal()
    plt.plot(t, s, drawstyle='steps-post')  # Use plot instead of stem
    plt.title("Digital Signal")

    # Quantized Signal
    plt.subplot(3, 2, 3)
    t, s = generate_quantized_signal()
    plt.plot(t, s, drawstyle='steps-post')  # Use plot instead of stem
    plt.title("Quantized Signal")

    # Sampled Signal
    plt.subplot(3, 2, 4)
    t, s = generate_sampled_signal()
    plt.plot(t, s)  # Use plot instead of stem
    plt.title("Sampled Signal")

    # Random Signal
    plt.subplot(3, 2, 5)
    t, s = generate_random_signal()
    plt.plot(t, s)
    plt.title("Random Signal")

    plt.tight_layout()
    plt.show()
```

# سیگنال‌های زوج و فرد (Even and Odd Signals)

## سیگنال زوج (Even Signal)

یک سیگنال \(x(t)\) زوج نامیده می‌شود اگر شرط زیر برای تمام مقادیر \(t\) برقرار باشد:

$$\[
x(t) = x(-t)
\]$$

این یعنی سیگنال نسبت به محور عمودی (محور \(y\)) متقارن است.

**مثال:** $$\(\cos(t)\)$$

## سیگنال فرد (Odd Signal)

یک سیگنال \(x(t)\) فرد نامیده می‌شود اگر شرط زیر برای تمام مقادیر \(t\) برقرار باشد:

\[
x(t) = -x(-t)
\]

این یعنی سیگنال نسبت به مبدأ مختصات متقارن است.

**مثال:** \(\sin(t)\)

## تجزیه یک سیگنال به اجزای زوج و فرد

هر سیگنال دلخواه \(x(t)\) را می‌توان به صورت مجموع یک سیگنال زوج و یک سیگنال فرد نوشت:

$$\[
x(t) = x_{\text{even}}(t) + x_{\text{odd}}(t)
\]$$

که در آن:

*   بخش زوج:

$$\[
x_{\text{even}}(t) = \frac{1}{2} [x(t) + x(-t)]
\$$]

*   بخش فرد:

$$\[
x_{\text{odd}}(t) = \frac{1}{2} [x(t) - x(-t)]
\$$]

**نکات:**

*   حاصل ضرب دو سیگنال زوج یا دو سیگنال فرد، زوج است.
*   حاصل ضرب یک سیگنال زوج و یک سیگنال فرد، فرد است. به کار می‌رود. یا پردازش سیگنال‌هایی که تقارن حول محور عمودی دارند، به کار می‌رود.


```python
import numpy as np
import matplotlib.pyplot as plt

# سیگنال زوج: کسینوس
def even_signal(frequency=5, duration=1, sample_rate=1000):
    time = np.linspace(-duration, duration, int(sample_rate * duration * 2))
    signal = np.cos(2 * np.pi * frequency * time)  # سیگنال زوج: کسینوس
    return time, signal

# سیگنال فرد: سینوس
def odd_signal(frequency=5, duration=1, sample_rate=1000):
    time = np.linspace(-duration, duration, int(sample_rate * duration * 2))
    signal = np.sin(2 * np.pi * frequency * time)  # سیگنال فرد: سینوس
    return time, signal

# رسم سیگنال‌ها
plt.figure(figsize=(15, 8))

# سیگنال زوج
t, even_sig = even_signal(frequency=5, duration=1, sample_rate=1000)
plt.subplot(2, 1, 1)
plt.plot(t, even_sig, label="سیگنال زوج: cos(2πft)")
plt.title("سیگنال زوج (Even Signal)")
plt.xlabel("زمان (t)")
plt.ylabel("آمپلتیود")
plt.grid(True)
plt.legend()

# سیگنال فرد
t, odd_sig = odd_signal(frequency=5, duration=1, sample_rate=1000)
plt.subplot(2, 1, 2)
plt.plot(t, odd_sig, label="سیگنال فرد: sin(2πft)", color="orange")
plt.title("سیگنال فرد (Odd Signal)")
plt.xlabel("زمان (t)")
plt.ylabel("آمپلتیود")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# سیگنال‌های متناوب و نامتناوب (Periodic and Aperiodic Signals)

سیگنال‌ها بر اساس تکرارشان در طول زمان به دو دسته اصلی متناوب و نامتناوب تقسیم می‌شوند.

## سیگنال متناوب (Periodic Signal)

یک سیگنال $$(x(t))$$ متناوب نامیده می‌شود اگر شرط زیر برای تمام مقادیر \(t\) و یک مقدار ثابت $$(T)$$ (دوره تناوب) برقرار باشد:

$$[
x(t) = x(t + T)
]$$

این یعنی سیگنال پس از گذشت زمان (T) دوباره تکرار می‌شود. کوچکترین مقدار مثبت \(T\) را دوره تناوب اصلی می‌نامند.

**مثال‌ها:**

*   تابع سینوس و کسینوس: $$(sin(t)), (cos(t)) (T \2\pi))$$
*   موج مربعی (Square wave)
*   موج دندان اره‌ای (Sawtooth wave)

**فرمول کلی برای سیگنال‌های متناوب سینوسی:**

$$[
x(t) = A\cos(\omega t + \phi)
]$$

که در آن:

*   \(A\): دامنه سیگنال
*   \(\omega\): فرکانس زاویه‌ای $$(omega = 2\pi f)$$
*   \(f\): فرکانس (تعداد تکرار در واحد زمان)
*   \(phi\): فاز

## سیگنال نامتناوب (Aperiodic Signal)

یک سیگنال \(x(t)\) نامتناوب نامیده می‌شود اگر هیچ مقدار \(T\) وجود نداشته باشد که شرط \(x(t) = x(t + T)\) را برای همه \(t\) برقرار کند. به عبارت دیگر، سیگنال هیچ الگوی تکراری ندارد.

**مثال‌ها:**

*  $$ (e^{-at}u(t))$$ (تابع نمایی میراشونده ضرب در تابع پله واحد)
*   سیگنال‌های تصادفی (Random signals)
*   پالس منفرد (Single pulse)

**توضیح بیشتر با مثال:**

فرض کنید سیگنال \(x(t)\) به صورت زیر تعریف شده باشد:

$$[
x(t) = \begin{cases}
1, & 0 \le t < 1 \\
0, & \text{otherwise}
\end{cases}
$$

این سیگنال یک پالس منفرد است و هیچگاه تکرار نمی‌شود، بنابراین نامتناوب است.

حالا فرض کنید سیگنال \(y(t)\) به صورت زیر باشد:

$$[
y(t) = \sin(2\pi t)
]$$

این سیگنال هر \(1\) ثانیه تکرار می‌شود (دوره تناوب \(T=1\))، بنابراین متناوب است.

**نکات مهم:**

*   جمع دو سیگنال متناوب لزوماً متناوب نیست، مگر اینکه نسبت دوره‌های تناوب آنها عددی گویا باشد.
*   سیگنال‌های نامتناوب را می‌توان به عنوان سیگنال‌هایی با دوره تناوب بی‌نهایت در نظر گرفت.

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. تعریف سیگنال سینوسی متناوب
T_sin = 2 * np.pi  # دوره سیگنال سینوسی
f_sin = 1 / T_sin  # فرکانس سیگنال سینوسی
t_sin = np.linspace(0, 4*T_sin, 1000)  # زمان برای رسم سیگنال
x_sin = np.cos(t_sin)  # سیگنال سینوسی

# 2. تعریف سیگنال مربعی متناوب
T_square = 2  # دوره سیگنال مربعی
t_square = np.linspace(0, 4*T_square, 1000)  # زمان برای رسم سیگنال
x_square = np.sign(np.sin(2 * np.pi * t_square / T_square))  # سیگنال مربعی

# رسم سیگنال‌ها
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# رسم سیگنال سینوسی
axs[0].plot(t_sin, x_sin, label=r'$x(t) = \cos(t)$', color='b')
axs[0].set_title('Periodic Sine Wave Signal', fontsize=14, loc='right')  # Title in English
axs[0].set_xlabel('Time (t)', fontsize=12)
axs[0].set_ylabel('Signal Value', fontsize=12)
axs[0].grid(True)
axs[0].legend()

# رسم سیگنال مربعی
axs[1].plot(t_square, x_square, label=r'$x(t) = \text{sign}(\sin(t))$', color='r')
axs[1].set_title('Periodic Square Wave Signal', fontsize=14, loc='right')  # Title in English
axs[1].set_xlabel('Time (t)', fontsize=12)
axs[1].set_ylabel('Signal Value', fontsize=12)
axs[1].grid(True)
axs[1].legend()

# تنظیم فاصله‌ها
plt.tight_layout()

# نمایش نمودار
plt.show()

```

# سیگنال‌های انرژی و توان

در تحلیل سیگنال‌ها، یکی از ویژگی‌های مهم که باید در نظر گرفته شود، میزان **انرژی** و **توان** سیگنال است. این ویژگی‌ها به ویژه در تحلیل سیستم‌های مخابراتی، فیزیک و مهندسی برق کاربرد دارند. در اینجا به توضیح این دو مفهوم پرداخته می‌شود.

## 1. سیگنال انرژی (Energy Signal)

سیگنال \( x(t) \) را سیگنال انرژی می‌نامیم اگر **انرژی** آن پایان‌پذیر باشد. انرژی سیگنال به صورت زیر تعریف می‌شود:

$$[
E_x = \int_{-\infty}^{\infty} |x(t)|^2 dt
]$$

اگر این مقدار انرژی، یک عدد معین باشد و بی‌نهایت نشود، سیگنال را سیگنال انرژی می‌نامیم. این سیگنال معمولاً در زمان‌های محدود غیر صفر است و برای تحلیل‌های سیستم‌های فیزیکی با سیگنال‌هایی با طول محدود کاربرد دارد.

### ویژگی‌های سیگنال انرژی:
- انرژی آن یک مقدار ثابت و محدود است.
- انرژی این سیگنال معمولاً برای سیگنال‌هایی با **طول محدود** استفاده می‌شود.

## 2. سیگنال توان (Power Signal)

سیگنال \( x(t) \) را سیگنال توان می‌نامیم اگر **توان متوسط** آن در طول زمان‌های بی‌نهایت ثابت باشد. توان متوسط سیگنال به صورت زیر تعریف می‌شود:

$$
P_x = \lim_{T \to \infty} \frac{1}{2T} \int_{-T}^{T} |x(t)|^2 dt
$$

در اینجا، اگر سیگنال دارای توان ثابت باشد، به آن سیگنال توان می‌گوییم. سیگنال‌های دوره‌ای مانند سیگنال‌های سینوسی و مربعی معمولاً سیگنال‌های توان هستند.

### ویژگی‌های سیگنال توان:
- توان متوسط آن یک مقدار ثابت و محدود است.
- توان سیگنال معمولاً برای سیگنال‌هایی با **طول نامحدود** یا **دوره‌ای** استفاده می‌شود.

## 3. رابطه انرژی و توان سیگنال

یک سیگنال ممکن است به یکی از دو دسته سیگنال انرژی یا سیگنال توان تعلق داشته باشد، اما هیچ سیگنال واحدی نمی‌تواند همزمان هم انرژی و هم توان باشد. به عبارت دیگر:
- اگر انرژی یک سیگنال محدود باشد، توان آن صفر است.
- اگر توان یک سیگنال محدود باشد، انرژی آن بی‌نهایت است.

## 4. مثال‌ها

### مثال 1: سیگنال انرژی

یک سیگنال $$( x(t) = e^{-t^2} )$$ که یک سیگنال گوسی است، انرژی آن محدود است. این سیگنال انرژی در محدوده زمان‌های محدود مقدار غیر صفر دارد و انرژی آن یک عدد معین است.

### مثال 2: سیگنال توان

سیگنال $$( x(t) = \cos(2\pi f t) )$$ که یک سیگنال سینوسی است، انرژی نامحدود دارد اما توان متوسط آن ثابت است. بنابراین، این سیگنال یک سیگنال توان است.

## 5. نتیجه‌گیری

- سیگنال انرژی: سیگنالی است که انرژی آن محدود است و معمولاً برای سیگنال‌هایی با طول محدود به کار می‌رود.
- سیگنال توان: سیگنالی است که توان متوسط آن ثابت است و معمولاً برای سیگنال‌های دوره‌ای یا طول نامحدود کاربرد دارد.

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. تعریف سیگنال انرژی
# سیگنال انرژی: x(t) = e^(-t^2)
T = 5  # محدوده زمانی برای رسم سیگنال انرژی
t_energy = np.linspace(-T, T, 1000)
x_energy = np.exp(-t_energy**2)

# محاسبه انرژی سیگنال
energy = np.trapz(np.abs(x_energy)**2, t_energy)  # انرژی سیگنال

# 2. تعریف سیگنال توان
# سیگنال توان: x(t) = cos(2*pi*f*t)
f = 1  # فرکانس
T_power = 10  # محدوده زمانی برای رسم سیگنال توان
t_power = np.linspace(0, T_power, 1000)
x_power = np.cos(2 * np.pi * f * t_power)

# محاسبه توان متوسط سیگنال
P_x = np.mean(np.abs(x_power)**2)  # توان متوسط سیگنال

# رسم سیگنال‌ها
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# رسم سیگنال انرژی
axs[0].plot(t_energy, x_energy, label=r'$x(t) = e^{-t^2}$', color='b')
axs[0].set_title('Energy Signal: $x(t) = e^{-t^2}$', fontsize=14, loc='right')  # Title in English
axs[0].set_xlabel('Time (t)', fontsize=12)
axs[0].set_ylabel('Signal Value', fontsize=12)
axs[0].grid(True)
axs[0].legend()
axs[0].text(0, 0.5, f'Energy = {energy:.2f}', fontsize=12, color='black')  # نمایش انرژی در نمودار

# رسم سیگنال توان
axs[1].plot(t_power, x_power, label=r'$x(t) = \cos(2\pi ft)$', color='r')
axs[1].set_title('Power Signal: $x(t) = \cos(2\pi ft)$', fontsize=14, loc='right')  # Title in English
axs[1].set_xlabel('Time (t)', fontsize=12)
axs[1].set_ylabel('Signal Value', fontsize=12)
axs[1].grid(True)
axs[1].legend()
axs[1].text(0, 0.5, f'Average Power = {P_x:.2f}', fontsize=12, color='black')  # نمایش توان متوسط در نمودار

# تنظیم فاصله‌ها
plt.tight_layout()

# نمایش نمودار
plt.show()

# نمایش نتایج در کنسول
print(f"Energy of the signal: {energy:.2f}")
print(f"Average power of the signal: {P_x:.2f}")
```

# شیفت زمانی (Time Shifting)

شیفت زمانی یکی از عملیات‌های اساسی در تحلیل سیگنال‌ها است که به تغییر موقعیت یک سیگنال در محور زمان اشاره دارد. این عملیات در سیستم‌های خطی و زمان‌ناهماهنگ کاربرد زیادی دارد.

## 1. تعریف شیفت زمانی

اگر $$( x(t) )$$ یک سیگنال باشد، آنگاه **شیفت زمانی** آن به صورت زیر تعریف می‌شود:

\[
y(t) = x(t - t_0)
\]

در اینجا:
- \( t_0 \) مقدار شیفت زمانی است که ممکن است مثبت یا منفی باشد.
- اگر \( t_0 > 0 \)، سیگنال \( x(t) \) به اندازه \( t_0 \) واحد به سمت راست (در زمان آینده) جابجا می‌شود.
- اگر \( t_0 < 0 \)، سیگنال \( x(t) \) به اندازه \( |t_0| \) واحد به سمت چپ (در زمان گذشته) جابجا می‌شود.

### مثال:
فرض کنید که سیگنال $$( x(t) = \cos(t) )$$ داریم. اگر این سیگنال به اندازه 2 واحد به سمت راست شیفت شود، سیگنال جدید به صورت زیر خواهد بود:

$$
y(t) = \cos(t - 2)
$$

این به این معنی است که سیگنال \( \cos(t) \) به اندازه 2 واحد به سمت راست جابجا شده است.

## 2. ویژگی‌های شیفت زمانی

- **خطی بودن**: شیفت زمانی یک عملیات خطی است. به این معنی که اگر سیگنال‌ها به صورت $$( x_1(t) ) و ( x_2(t) )$$ باشند، آنگاه شیفت خطی آن‌ها به صورت زیر خواهد بود:
  
  $$
  a \cdot x_1(t - t_0) + b \cdot x_2(t - t_0) = a \cdot x_1(t - t_0) + b \cdot x_2(t - t_0)
$$

- **تغییر در ویژگی‌های سیگنال**: شیفت زمانی معمولاً بر روی ویژگی‌های آمپلیتوئید (مقدار سیگنال) یا فرکانس سیگنال تأثیری ندارد، بلکه فقط آن را در محور زمان جابجا می‌کند.

## 3. نمودار شیفت زمانی

### سیگنال اصلی: $$( x(t) = \cos(t) )$$

اگر سیگنال اصلی $$( x(t) = \cos(t))$$ را داشته باشیم و بخواهیم این سیگنال را به اندازه 2 واحد به سمت راست و 3 واحد به سمت چپ شیفت دهیم، سیگنال‌های حاصل به ترتیب به صورت زیر خواهند بود:

- شیفت 2 واحد به سمت راست: $$( y(t) = \cos(t - 2) )$$
- شیفت 3 واحد به سمت چپ: $$( y(t) = \cos(t + 3) )$$

## 4. نتیجه‌گیری

شیفت زمانی یک عملیات ساده و کاربردی است که برای تحلیل و طراحی سیستم‌های خطی و زمان‌ناهماهنگ بسیار مفید است. این عملیات به ما کمک می‌کند که رفتار سیگنال‌ها را در زمان‌های مختلف تحلیل کنیم.

</div>
گنال‌ها را در زمان‌های مختلف تحلیل کنیم.

</div>


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the original signal: x(t) = cos(t)
T = 2 * np.pi  # One period of the cosine function
t = np.linspace(0, T, 1000)  # Time range for one period
x_t = np.cos(t)  # Original signal

# 2. Time shift: Shifting by +2 and -3
t_shifted_pos = np.linspace(-T, T, 1000)  # Time for shifted signal (right shift)
x_shifted_pos = np.cos(t_shifted_pos - 2)  # Shifted by 2 units to the right

t_shifted_neg = np.linspace(-T, T, 1000)  # Time for shifted signal (left shift)
x_shifted_neg = np.cos(t_shifted_neg + 3)  # Shifted by 3 units to the left

# 3. Plotting the signals
fig, axs = plt.subplots(3, 1, figsize=(10, 9))

# Plotting the original signal
axs[0].plot(t, x_t, label=r'$x(t) = \cos(t)$', color='b')
axs[0].set_title('Original Signal: $x(t) = \cos(t)$', fontsize=14, loc='right')
axs[0].set_xlabel('Time (t)', fontsize=12)
axs[0].set_ylabel('Signal Value', fontsize=12)
axs[0].grid(True)
axs[0].legend()

# Plotting the signal shifted by 2 units to the right
axs[1].plot(t_shifted_pos, x_shifted_pos, label=r'$x(t - 2)$', color='r')
axs[1].set_title('Time Shifted Signal by +2: $x(t - 2)$', fontsize=14, loc='right')
axs[1].set_xlabel('Time (t)', fontsize=12)
axs[1].set_ylabel('Signal Value', fontsize=12)
axs[1].grid(True)
axs[1].legend()

# Plotting the signal shifted by 3 units to the left
axs[2].plot(t_shifted_neg, x_shifted_neg, label=r'$x(t + 3)$', color='g')
axs[2].set_title('Time Shifted Signal by -3: $x(t + 3)$', fontsize=14, loc='right')
axs[2].set_xlabel('Time (t)', fontsize=12)
axs[2].set_ylabel('Signal Value', fontsize=12)
axs[2].grid(True)
axs[2].legend()

# Adjusting the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

# وارون‌سازی زمانی (Time Reversal)

وارون‌سازی زمانی یک عملیات است که به معنای معکوس کردن جهت زمان یک سیگنال است. در این عملیات، زمان \( t \) در فرمول سیگنال با \( -t \) جایگزین می‌شود. به عبارت دیگر، در وارون‌سازی زمانی، سیگنال اصلی در جهت زمان معکوس می‌شود.

## 1. تعریف وارون‌سازی زمانی

اگر \( x(t) \) یک سیگنال باشد، آنگاه سیگنال وارونه‌شده آن به صورت زیر تعریف می‌شود:

$$
y(t) = x(-t)
$$

این بدان معناست که سیگنال $$( x(t) ) $$در زمان \( t \) به سیگنال $$( y(t) = x(-t) )$$ تبدیل می‌شود که در آن تمام مقادیر زمانی معکوس شده‌اند.

### مثال:
فرض کنید سیگنال $$( x(t) = \cos(t) ) $$داریم. اگر این سیگنال وارونه شود، سیگنال جدید به صورت زیر خواهد بود:

$$
y(t) = \cos(-t) = \cos(t)
$$

در اینجا، چون $$( \cos(t) )$$ یک تابع زوج است، اثر وارون‌سازی زمانی بر روی آن تاثیری ندارد. اما برای توابع دیگر مانند سینوس، وارون‌سازی زمان تغییراتی را در سیگنال ایجاد خواهد کرد.

## 2. ویژگی‌های وارون‌سازی زمانی

- **تغییر جهت زمان**: در وارون‌سازی زمانی، سیگنال در جهت زمان معکوس می‌شود، ولی ویژگی‌های دیگر مانند آمپلیتوئید یا فرکانس تغییر نمی‌کنند.
- **عملیات خطی**: این عملیات خطی است و به همین دلیل می‌تواند برای ترکیب سیگنال‌ها و تحلیل‌های پیچیده‌تر استفاده شود.

## 3. نمودار وارون‌سازی زمانی

### سیگنال اصلی: $$( x(t) = \cos(t) )$$

اگر بخواهیم سیگنال $$( \cos(t) )$$ را وارونه کنیم، نتیجه به صورت زیر خواهد بود:

$$
y(t) = \cos(-t)
$$

### نتیجه‌گیری

وارون‌سازی زمانی یک عملیات ساده اما قدرتمند است که در تحلیل سیگنال‌ها و طراحی سیستم‌های زمان‌ناهماهنگ و غیرخطی کاربرد زیادی دارد.

</div>


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the original signal: x(t) = cos(t)
T = 2 * np.pi  # One period of the cosine function
t = np.linspace(0, T, 1000)  # Time range for one period
x_t = np.cos(t)  # Original signal

# 2. Time reversal: Reversing the time: x(-t)
t_reversed = -t  # Time axis reversed
x_reversed = np.cos(t_reversed)  # Reversed signal: x(-t)

# 3. Plotting the signals
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# Plotting the original signal
axs[0].plot(t, x_t, label=r'$x(t) = \cos(t)$', color='b')
axs[0].set_title('Original Signal: $x(t) = \cos(t)$', fontsize=14, loc='right')
axs[0].set_xlabel('Time (t)', fontsize=12)
axs[0].set_ylabel('Signal Value', fontsize=12)
axs[0].grid(True)
axs[0].legend()

# Plotting the time-reversed signal
axs[1].plot(t_reversed, x_reversed, label=r'$x(-t)$', color='r')
axs[1].set_title('Time Reversed Signal: $x(-t)$', fontsize=14, loc='right')
axs[1].set_xlabel('Time (t)', fontsize=12)
axs[1].set_ylabel('Signal Value', fontsize=12)
axs[1].grid(True)
axs[1].legend()

# Adjusting the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

# تغییر مقیاس سیگنال (Signal Scaling)

تغییر مقیاس سیگنال‌ها می‌تواند به دو صورت انجام شود: 

1. **تغییر مقیاس زمانی**: در این حالت، زمان سیگنال با یک عامل \( a \) ضرب می‌شود. به عبارت دیگر، سیگنال \( x(t) \) به صورت \( x(at) \) تغییر می‌کند.
   - اگر \( a > 1 \)، سیگنال فشرده‌تر می‌شود (زمان به جلو می‌رود).
   - اگر \( a < 1 \)، سیگنال کشیده‌تر می‌شود (زمان به عقب می‌رود).

2. **تغییر مقیاس دامنه‌ای**: در این حالت، سیگنال با یک ثابت \( A \) ضرب می‌شود. سیگنال به صورت \( A \cdot x(t) \) تغییر می‌کند.
   - اگر \( A > 1 \)، دامنه سیگنال بزرگتر می‌شود.
   - اگر \( A < 1 \)، دامنه سیگنال کوچکتر می‌شود.

## 1. تغییر مقیاس زمانی:

اگر سیگنال اصلی به صورت \( x(t) \) باشد، بعد از اعمال تغییر مقیاس زمانی به صورت زیر خواهد بود:

$$
y(t) = x(at)
$$

- اگر \( a > 1 \)، سیگنال فشرده‌تر می‌شود.
- اگر \( a < 1 \)، سیگنال کشیده‌تر می‌شود.

### مثال:
اگر سیگنال اصلی $$( x(t) = \cos(t) )$$ باشد و \( a = 2 \) را انتخاب کنیم، سیگنال تغییر یافته به صورت $$( x(2t) = \cos(2t) )$$ خواهد بود.

## 2. تغییر مقیاس دامنه‌ای:

اگر سیگنال اصلی به صورت \( x(t) \) باشد، بعد از اعمال تغییر مقیاس دامنه‌ای به صورت زیر خواهد بود:

$$
y(t) = A \cdot x(t)
$$

- اگر \( A > 1 \)، دامنه سیگنال افزایش می‌یابد.
- اگر \( A < 1 \)، دامنه سیگنال کاهش می‌یابد.

### مثال:
اگر سیگنال اصلی $$( x(t) = \cos(t) )$$ باشد و $$( A = 1.5 )$$ را انتخاب کنیم، سیگنال تغییر یافته به صورت $$( 1.5 \cdot \cos(t) )$$ خواهد بود.

</div>


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the original signal: x(t) = cos(t)
T = 2 * np.pi  # One period of the cosine function
t = np.linspace(0, T, 1000)  # Time range for one period
x_t = np.cos(t)  # Original signal

# 2. Applying time scaling: y(t) = x(at)
a = 2  # Time scaling factor (a > 1: compress, a < 1: stretch)
t_scaled = np.linspace(0, T/a, 1000)  # New time axis after scaling
x_scaled = np.cos(a * t_scaled)  # Scaled signal

# 3. Applying amplitude scaling: y(t) = A * x(t)
A = 1.5  # Amplitude scaling factor (A > 1: increase amplitude, A < 1: decrease amplitude)
x_scaled_amplitude = A * x_t  # Amplitude scaled signal

# 4. Plotting the signals
fig, axs = plt.subplots(3, 1, figsize=(10, 9))

# Plotting the original signal
axs[0].plot(t, x_t, label=r'$x(t) = \cos(t)$', color='b')
axs[0].set_title('Original Signal: $x(t) = \cos(t)$', fontsize=14, loc='right')
axs[0].set_xlabel('Time (t)', fontsize=12)
axs[0].set_ylabel('Signal Value', fontsize=12)
axs[0].grid(True)
axs[0].legend()

# Plotting the time-scaled signal
axs[1].plot(t_scaled, x_scaled, label=r'$x(at) = \cos(2t)$', color='r')
axs[1].set_title('Time Scaled Signal: $x(at) = \cos(2t)$', fontsize=14, loc='right')
axs[1].set_xlabel('Time (t)', fontsize=12)
axs[1].set_ylabel('Signal Value', fontsize=12)
axs[1].grid(True)
axs[1].legend()

# Plotting the amplitude-scaled signal
axs[2].plot(t, x_scaled_amplitude, label=r'$A \cdot x(t)$', color='g')
axs[2].set_title('Amplitude Scaled Signal: $A \cdot x(t)$', fontsize=14, loc='right')
axs[2].set_xlabel('Time (t)', fontsize=12)
axs[2].set_ylabel('Signal Value', fontsize=12)
axs[2].grid(True)
axs[2].legend()

# Adjusting the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

# سیگنال نمایی

سیگنال نمایی یک نوع سیگنال است که مقدار آن به صورت نمایی با زمان تغییر می‌کند. سیگنال نمایی به طور عمومی به صورت زیر تعریف می‌شود:

$$
x(t) = A e^{\alpha t}
$$

که در آن:
- \( A \) دامنه سیگنال است.
- $$( \alpha )$$ نرخ رشد یا کاهش سیگنال است.
- \( t \) زمان است.

### ویژگی‌ها:
1. اگر $$( \alpha > 0 )$$، سیگنال به طور نمایی افزایش می‌یابد.
2. اگر $$(alpha < 0 )$$، سیگنال به طور نمایی کاهش می‌یابد.

### مثال:
اگر ( A = 1 ) و ( alpha = 0.5 )، سیگنال به صورت $$( x(t) = e^{0.5t} )$$ به طور نمایی افزایش می‌یابد. اگر \( alpha = -0.5 \)، سیگنال به صورت $$( x(t) = e^{-0.5t} )$$ کاهش خواهد یافت.

در زیر، نمودارهای مربوط به سیگنال‌های رشد و کاهش نمایی آورده شده است.



```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the exponential signal: x(t) = A * exp(alpha * t)
A = 1  # Amplitude of the signal
alpha = 0.5  # Growth rate (can be positive for growth or negative for decay)
t = np.linspace(0, 5, 1000)  # Time range from 0 to 5 seconds
x_t = A * np.exp(alpha * t)  # Exponential growth signal

# 2. Defining the decaying exponential signal (alpha = -0.5)
alpha_decay = -0.5  # Negative alpha for decay
x_t_decay = A * np.exp(alpha_decay * t)  # Exponential decay signal

# 3. Plotting the signals
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plotting the growing exponential signal
axs[0].plot(t, x_t, label=r'$x(t) = e^{0.5t}$', color='b')
axs[0].set_title('Exponential Growth Signal: $x(t) = e^{0.5t}$', fontsize=14, loc='right')
axs[0].set_xlabel('Time (t)', fontsize=12)
axs[0].set_ylabel('Signal Value', fontsize=12)
axs[0].grid(True)
axs[0].legend()

# Plotting the decaying exponential signal
axs[1].plot(t, x_t_decay, label=r'$x(t) = e^{-0.5t}$', color='r')
axs[1].set_title('Exponential Decay Signal: $x(t) = e^{-0.5t}$', fontsize=14, loc='right')
axs[1].set_xlabel('Time (t)', fontsize=12)
axs[1].set_ylabel('Signal Value', fontsize=12)
axs[1].grid(True)
axs[1].legend()

# Adjusting the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

# سیگنال نمایی گسسته

سیگنال نمایی گسسته مشابه سیگنال نمایی پیوسته است، با این تفاوت که زمان به صورت گسسته و در نقاط مشخص \( n \) تعریف می‌شود. این سیگنال به صورت زیر است:

$$
x[n] = A e^{\alpha n}
$$

که در آن:
- \( A \) دامنه سیگنال است.
- \( alpha \) نرخ رشد یا کاهش سیگنال است.
- \( n \) شاخص زمانی گسسته است.

### ویژگی‌ها:
1. اگر \( alpha > 0 \)، سیگنال به طور نمایی افزایش می‌یابد.
2. اگر \( alpha < 0 \)، سیگنال به طور نمایی کاهش می‌یابد.

### مثال:
اگر \( A = 1 \) و \( alpha = 0.5 \)، سیگنال به صورت $$( x[n] = e^{0.5n} )$$ به طور نمایی افزایش می‌یابد. اگر \( \alpha = -0.5 \)، سیگنال به صورت $$( x[n] = e^{-0.5n} )$$ کاهش خواهد یافت.

در زیر، نمودارهای مربوط به سیگنال‌های نمایی گسسته رشد و کاهش نمایی آورده شده است.


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the discrete exponential signal: x[n] = A * exp(alpha * n)
A = 1  # Amplitude of the signal
alpha = 0.5  # Growth rate (can be positive for growth or negative for decay)
n = np.arange(0, 20, 1)  # Discrete time range (n = 0, 1, 2, ..., 19)
x_n = A * np.exp(alpha * n)  # Exponential growth signal

# 2. Defining the decaying exponential signal (alpha = -0.5)
alpha_decay = -0.5  # Negative alpha for decay
x_n_decay = A * np.exp(alpha_decay * n)  # Exponential decay signal

# 3. Plotting the signals
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plotting the growing exponential signal
axs[0].plot(n, x_n, label=r'$x[n] = e^{0.5n}$', color='b', marker='o', linestyle='-')
axs[0].set_title('Discrete Exponential Growth Signal: $x[n] = e^{0.5n}$', fontsize=14, loc='right')
axs[0].set_xlabel('Discrete Time (n)', fontsize=12)
axs[0].set_ylabel('Signal Value', fontsize=12)
axs[0].grid(True)
axs[0].legend()

# Plotting the decaying exponential signal
axs[1].plot(n, x_n_decay, label=r'$x[n] = e^{-0.5n}$', color='r', marker='o', linestyle='-')
axs[1].set_title('Discrete Exponential Decay Signal: $x[n] = e^{-0.5n}$', fontsize=14, loc='right')
axs[1].set_xlabel('Discrete Time (n)', fontsize=12)
axs[1].set_ylabel('Signal Value', fontsize=12)
axs[1].grid(True)
axs[1].legend()

# Adjusting the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

# سیگنال سینوسی میرا شونده

سیگنال سینوسی میرا شونده ترکیبی از یک سیگنال سینوسی و یک نمایی است که به تدریج دامنه آن کاهش می‌یابد. این نوع سیگنال در بسیاری از سیستم‌های فیزیکی دیده می‌شود، مانند سیستم‌های ارتعاشی یا مدارهای RLC.

فرمول کلی سیگنال سینوسی میرا شونده به صورت زیر است:

$$
x(t) = A e^{-\alpha t} \cos(\omega t + \phi)
$$

که در آن:
- \( A \) دامنه اولیه است.
- \( alpha \) نرخ میرایی است که تأثیر آن در کاهش دامنه سیگنال نمایان می‌شود.
- \( omega \) فرکانس زاویه‌ای سیگنال است.
- \( phi \) فاز اولیه است.
- \( t \) زمان است.

### ویژگی‌ها:
1. سیگنال سینوسی میرا شونده ترکیبی از دو اثر است: میرایی نمایی (کاهش دامنه) و نوسانات سینوسی.
2. اگر \( alpha > 0 \)، سیگنال به تدریج دامنه خود را کاهش می‌دهد.
3. اگر \( alpha = 0 \)، سیگنال سینوسی ساده بدون میرایی خواهد بود.

### مثال:
اگر \( A = 1 \)، \( alpha = 0.2 \)، $$( \omega = 2\pi )$$ و \( phi = 0 \)، سیگنال سینوسی میرا شونده به صورت $$( x(t) = e^{-0.2t} \cos(2\pi t) )$$ خواهد بود که دامنه آن به تدریج کاهش می‌یابد.


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the damped sinusoidal signal: x(t) = A * exp(-alpha * t) * cos(omega * t + phi)
A = 1          # Initial amplitude
alpha = 0.2    # Damping factor
omega = 2 * np.pi  # Angular frequency (for 1 Hz frequency)
phi = 0        # Initial phase
t = np.linspace(0, 10, 1000)  # Time from 0 to 10 seconds

# Defining the damped sinusoidal signal
x_t = A * np.exp(-alpha * t) * np.cos(omega * t + phi)

# 2. Plotting the damped sinusoidal signal
plt.figure(figsize=(10, 6))
plt.plot(t, x_t, label=r'$x(t) = e^{-0.2t} \cos(2\pi t)$', color='b')
plt.title('Damped Sinusoidal Signal', fontsize=14)
plt.xlabel('Time (t)', fontsize=12)
plt.ylabel('Signal Value', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

# تابع پله واحد

تابع پله واحد یک تابع گسسته است که معمولاً با \( u(t) \) یا \( H(t) \) نشان داده می‌شود. این تابع معمولاً برای مدل‌سازی سیگنال‌هایی که از زمان مشخصی به بعد فعال می‌شوند، استفاده می‌شود. تابع پله واحد در زمان \( t = 0 \) یک تغییر ناگهانی دارد.

فرمول تابع پله واحد به صورت زیر است:

$$
u(t) = 
\begin{cases} 
0, & t < 0 \\
1, & t \geq 0
\end{cases}
$$

که در آن:
- \( u(t) \) مقدار تابع در زمان \( t \) است.
- \( t \) زمان است.

### ویژگی‌ها:
1. اگر \( t < 0 \)، تابع پله واحد برابر با صفر است.
2. اگر \( t geq 0 \)، تابع پله واحد برابر با یک است.
3. تابع پله واحد معمولاً برای مدل‌سازی سیگنال‌هایی که از زمان خاصی به بعد شروع به تغییر می‌کنند، استفاده می‌شود.

### مثال:
تابع پله واحد \( u(t) \) نشان‌دهنده این است که یک سیگنال از زمان \( t = 0 \) به بعد به مقدار 1 می‌رسد و قبل از آن صفر است.


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the unit step function u(t)
def unit_step(t):
    return np.where(t >= 0, 1, 0)  # Returns 1 for t >= 0, otherwise 0

# 2. Generating time values
t = np.linspace(-5, 5, 1000)  # Time values from -5 to 5

# 3. Applying the unit step function
u_t = unit_step(t)

# 4. Plotting the unit step function
plt.figure(figsize=(10, 6))
plt.plot(t, u_t, label=r'$u(t)$', color='b')
plt.title('Unit Step Function: $u(t)$', fontsize=14)
plt.xlabel('Time (t)', fontsize=12)
plt.ylabel('u(t)', fontsize=12)
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)  # x-axis line
plt.axvline(0, color='black',linewidth=1)  # y-axis line
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

# تابع ضربه (Impulse Function)

تابع ضربه که معمولاً با \( \delta(t) \) نمایش داده می‌شود، یک تابع خاص است که در آن مقدار تابع در تمام نقاط به جز در \( t = 0 \) صفر است. در \( t = 0 \)، این تابع مقدار بی‌نهایت می‌گیرد. تابع ضربه معمولاً برای مدل‌سازی اختلالات آنی یا رویدادهای لحظه‌ای در سیستم‌ها استفاده می‌شود.


فرمول تابع ضربه به صورت زیر است:

$$
\delta(t) =
\begin{cases}
\infty, & t = 0 \\
0, & t \neq 0
\end{cases}
$$

### ویژگی‌ها:
1. در \( t = 0 \)، تابع ضربه مقدار بی‌نهایت دارد (که به طور نظری به عنوان یک "نقطه بی‌نهایت" در نظر گرفته می‌شود).
2. در هر نقطه دیگر (برای \( t \neq 0 \))، مقدار تابع برابر صفر است.
3. تابع ضربه در تحلیل‌های سیگنال‌ها و سیستم‌ها برای مدل‌سازی اختلالات آنی کاربرد دارد.

### خاصیت مهم تابع ضربه:
یکی از ویژگی‌های مهم تابع ضربه این است که:

$$
\int_{-\infty}^{\infty} \delta(t) \, dt = 1
$$

این خاصیت به این معناست که مساحت زیر منحنی تابع ضربه برابر با 1 است.

### مثال:
تابع ضربه معمولاً در سیستم‌های فیزیکی و تحلیل‌های ریاضی برای مدل‌سازی اختلالات لحظه‌ای به کار می‌رود. در این موارد، تابع ضربه به جای استفاده از یک سیگنال گسسته، اختلال را در یک نقطه زمانی خاص مدل می‌کند.


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the impulse function delta(t)
def impulse(t):
    return np.where(t == 0, 1, 0)  # Impulse function is 1 at t = 0, 0 otherwise

# 2. Generating time values
t = np.linspace(-5, 5, 1000)  # Time values from -5 to 5

# 3. Applying the impulse function
delta_t = impulse(t)

# 4. Plotting the impulse function
plt.figure(figsize=(10, 6))
plt.plot(t, delta_t, label=r'$\delta(t)$', color='r', linewidth=2)
plt.title('Impulse Function: $\delta(t)$', fontsize=14)
plt.xlabel('Time (t)', fontsize=12)
plt.ylabel(r'$\delta(t)$', fontsize=12)
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)  # x-axis line
plt.axvline(0, color='black',linewidth=1)  # y-axis line
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()

```

# تابع شیب (Ramp Function)

تابع شیب یکی از توابع پایه‌ای در تحلیل سیگنال‌ها است که معمولاً به شکل \( r(t) \) یا $$( \text{ramp}(t) )$$ نشان داده می‌شود. این تابع از یک نقطه شروع، به طور پیوسته و خطی افزایش می‌یابد و معمولاً برای مدل‌سازی سیگنال‌هایی که به صورت خطی از صفر افزایش می‌کنند، استفاده می‌شود.

فرمول تابع شیب به صورت زیر است:

$$
r(t) = 
\begin{cases} 
0, & t < 0 \\
t, & t \geq 0
\end{cases}
$$

که در آن:
- \( r(t) \) مقدار تابع در زمان \( t \) است.
- \( t \) زمان است.

### ویژگی‌ها:
1. اگر \( t < 0 \)، تابع شیب برابر با صفر است.
2. اگر \( t \geq 0 \)، تابع شیب برابر با \( t \) است، یعنی تابع به صورت خطی افزایش می‌یابد.
3. تابع شیب معمولاً برای مدل‌سازی سیگنال‌هایی که از زمان خاصی شروع به افزایش می‌کنند، به کار می‌رود.

### مثال:
اگر \( t = 0 \)، مقدار تابع شیب برابر با صفر است. برای \( t = 1 \)، مقدار تابع شیب برابر با 1 خواهد بود، و به همین ترتیب برای مقادیر بیشتر \( t \) تابع به صورت خطی افزایش خواهد یافت.

### نمودار تابع شیب:
نمودار تابع شیب به صورت یک خط مستقیم است که از \( t = 0 \) به بعد از صفر شروع به افزایش می‌کند.


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the ramp function r(t)
def ramp(t):
    return np.where(t >= 0, t, 0)  # Returns t for t >= 0, otherwise 0

# 2. Generating time values
t = np.linspace(-5, 5, 1000)  # Time values from -5 to 5

# 3. Applying the ramp function
r_t = ramp(t)

# 4. Plotting the ramp function
plt.figure(figsize=(10, 6))
plt.plot(t, r_t, label=r'$r(t)$', color='g', linewidth=2)
plt.title('Ramp Function: $r(t)$', fontsize=14)
plt.xlabel('Time (t)', fontsize=12)
plt.ylabel(r'$r(t)$', fontsize=12)
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)  # x-axis line
plt.axvline(0, color='black',linewidth=1)  # y-axis line
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

# تابع دوبلت واحد (Double Unit Step Function)

تابع دوبلت واحد یک تابع گسسته است که در آن سیگنال به صورت همزمان در دو نقطه خاص تغییر می‌کند. این تابع معمولاً برای مدل‌سازی سیگنال‌هایی که در دو زمان مختلف فعال می‌شوند یا دارای تغییرات پله‌ای در دو زمان مختلف هستند، استفاده می‌شود.

فرمول تابع دوبلت واحد به صورت زیر است:

$$
u_2(t) = u(t + a) - u(t - b)
$$

که در آن:
- \( u(t) \) تابع پله واحد است.
- \( a \) و \( b \) زمان‌هایی هستند که تابع دوبلت واحد در آن‌ها تغییر می‌کند.
- \( u(t + a) \) یک پله واحد است که در \( t = -a \) فعال می‌شود.
- \( u(t - b) \) یک پله واحد است که در \( t = b \) فعال می‌شود.

### ویژگی‌ها:
1. اگر \( t < -a \)، تابع دوبلت واحد برابر با صفر است.
2. اگر \( -a \leq t < b \)، تابع دوبلت واحد برابر با یک است.
3. اگر \( t \geq b \)، تابع دوبلت واحد دوباره صفر می‌شود.

### مثال:
برای تابع دوبلت واحد با \( a = 2 \) و \( b = 5 \)، تابع به صورت زیر خواهد بود:
$$
u_2(t) = u(t + 2) - u(t - 5)
$$
که در آن:
- تابع \( u(t + 2) \) در \( t = -2 \) فعال می‌شود.
- تابع \( u(t - 5) \) در \( t = 5 \) فعال می‌شود.

### نمودار تابع دوبلت واحد:
تابع دوبلت واحد معمولاً به صورت یک مستطیل بین دو نقطه \( t = -a \) و \( t = b \) رسم می‌شود.


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the double unit step function u_2(t)
def double_unit_step(t, a, b):
    return np.where(t >= -a, 1, 0) - np.where(t >= b, 1, 0)  # u(t+a) - u(t-b)

# 2. Generating time values
t = np.linspace(-10, 10, 1000)  # Time values from -10 to 10

# 3. Applying the double unit step function with specific a and b values
a = 2  # Shift for the first step
b = 5  # Shift for the second step
u_2_t = double_unit_step(t, a, b)

# 4. Plotting the double unit step function
plt.figure(figsize=(10, 6))
plt.plot(t, u_2_t, label=r'$u_2(t)$', color='orange', linewidth=2)
plt.title('Double Unit Step Function: $u_2(t)$', fontsize=14)
plt.xlabel('Time (t)', fontsize=12)
plt.ylabel(r'$u_2(t)$', fontsize=12)
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)  # x-axis line
plt.axvline(0, color='black',linewidth=1)  # y-axis line
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

# سیستم (System)

در مهندسی و ریاضیات، **سیستم** به مجموعه‌ای از اجزاء یا واحدهایی اطلاق می‌شود که به یکدیگر متصل شده‌اند و به طور مشترک یک ورودی را به خروجی تبدیل می‌کنند. سیستم‌ها می‌توانند در زمینه‌های مختلف مانند سیستم‌های فیزیکی، ریاضیاتی، کامپیوتری و غیره وجود داشته باشند.

### تعریف سیستم:
یک سیستم می‌تواند به صورت یک تابع یا رابطه ریاضیاتی بین ورودی و خروجی تعریف شود. به عبارت دیگر، سیستم‌ها ورودی‌های خاصی را دریافت کرده و خروجی‌های متناسبی تولید می‌کنند. 

فرمول کلی یک سیستم به صورت زیر است:

$$
y(t) = T\{x(t)\}
$$

که در آن:
- \( y(t) \) خروجی سیستم است.
- \( x(t) \) ورودی سیستم است.
- \( T \) عملگر سیستم است که ورودی را به خروجی تبدیل می‌کند.

### انواع سیستم‌ها:
سیستم‌ها را می‌توان به دسته‌های مختلفی تقسیم کرد که برخی از آنها عبارتند از:
1. **سیستم‌های خطی (Linear Systems):** سیستم‌هایی که اصول ابراد و هم‌خطی را رعایت می‌کنند.
2. **سیستم‌های غیرخطی (Nonlinear Systems):** سیستم‌هایی که اصول ابراد و هم‌خطی را رعایت نمی‌کنند.
3. **سیستم‌های زمان پیوسته (Continuous-Time Systems):** سیستم‌هایی که ورودی و خروجی آن‌ها در طول زمان به صورت پیوسته تغییر می‌کنند.
4. **سیستم‌های زمان گسسته (Discrete-Time Systems):** سیستم‌هایی که ورودی و خروجی آن‌ها در زمان‌های گسسته تغییر می‌کنند.

### ویژگی‌های سیستم‌ها:
1. **خطی بودن:** اگر سیستم در برابر ترکیب ورودی‌ها و تغییر مقیاس‌ها به طور متناسب پاسخ دهد، سیستم خطی است.
2. **پایداری:** یک سیستم پایدار است اگر پاسخ به ورودی‌های محدود، محدود باقی بماند.
3. **زمان-مستقل بودن:** اگر خواص سیستم در طول زمان تغییر نکند، سیستم زمان-مستقل است.

### مثال:
یک مثال ساده از یک سیستم، **مدار الکتریکی** است که در آن ورودی ولتاژ و خروجی جریان است. در اینجا عملگر سیستم می‌تواند معادله‌ای باشد که رابطه بین ولتاژ و جریان را بیان می‌کند.


# اجزای چهارگانه سیستم (Four Components of a System)

سیستم‌ها در علم مهندسی و ریاضیات به طور کلی از چهار جزء اصلی تشکیل می‌شوند که هرکدام نقش خاصی در عملکرد کلی سیستم دارند. این اجزا به طور معمول عبارتند از:

1. **ورودی (Input)**
2. **عملگر (Operator یا System)**
3. **خروجی (Output)**
4. **بازخورد (Feedback)**

### 1. ورودی (Input)
ورودی‌ها اطلاعات یا سیگنال‌هایی هستند که به سیستم وارد می‌شوند. ورودی می‌تواند از منابع مختلفی مانند سیگنال‌های الکتریکی، داده‌های کامپیوتری، یا حتی سیگنال‌های صوتی یا تصویری باشد. ورودی‌ها معمولاً به عنوان داده‌هایی که نیاز به پردازش دارند، عمل می‌کنند.

فرمول ورودی به صورت کلی:

$$
x(t) \quad \text{(ورودی سیستم)}
$$

### 2. عملگر (Operator یا System)
عملگر، همان سیستم یا تابعی است که ورودی‌ها را دریافت کرده و آنها را به خروجی تبدیل می‌کند. این عملگر ممکن است یک سیستم فیزیکی، ریاضیاتی، یا حتی دیجیتال باشد که ورودی را پردازش کرده و خروجی مناسب را تولید می‌کند.

فرمول عملگر به صورت کلی:

$$
y(t) = T\{x(t)\} \quad \text{(خروجی سیستم)}
$$

در اینجا \( T \) نمایانگر عملگر سیستم است که ورودی \( x(t) \) را به خروجی \( y(t) \) تبدیل می‌کند.

### 3. خروجی (Output)
خروجی نتیجه پردازش ورودی توسط عملگر سیستم است. خروجی ممکن است به صورت سیگنال، داده، یا هر نوع نتیجه‌ای باشد که از سیستم تولید می‌شود. هدف اصلی هر سیستم، تولید خروجی مفید و مورد نیاز از ورودی‌ها است.

فرمول خروجی به صورت کلی:

$$
y(t) \quad \text{(خروجی سیستم)}
$$

### 4. بازخورد (Feedback)
بازخورد فرآیندی است که در آن خروجی سیستم به ورودی آن بازگشت داده می‌شود. این ویژگی معمولاً برای تنظیم و بهبود عملکرد سیستم‌ها به کار می‌رود. در بسیاری از سیستم‌ها، بازخورد می‌تواند باعث پایداری یا تغییرات در رفتار سیستم شود.

فرمول بازخورد به صورت کلی:

$$
x(t) = x_{\text{in}}(t) + \beta y(t)
$$

که در آن:
- $$( x_{\text{in}}(t) )$$ ورودی اولیه است.
- $$( \beta y(t) )$$ بازخورد از خروجی به ورودی است.

### جمع‌بندی:
اجزای چهارگانه یک سیستم به طور کلی شامل ورودی، عملگر، خروجی و بازخورد هستند که با همکاری یکدیگر عملکرد کلی سیستم را تعیین می‌کنند. این اجزا در بسیاری از سیستم‌های مهندسی، کنترل، پردازش سیگنال و دیگر شاخه‌ها کاربرد دارند.


# طبقه‌بندی سیستم‌ها (Classification of Systems)

سیستم‌ها را می‌توان بر اساس ویژگی‌های مختلفی طبقه‌بندی کرد. برخی از مهم‌ترین طبقه‌بندی‌ها به شرح زیر است:

## 1. سیستم‌های خطی و غیرخطی (Linear and Nonlinear Systems)

- **سیستم‌های خطی (Linear Systems):** سیستم‌هایی که اصول ابراد و هم‌خطی را رعایت می‌کنند. در این سیستم‌ها، خروجی سیستم متناسب با ورودی‌ها است.
  
  \[
  y(t) = a_1 x_1(t) + a_2 x_2(t) + \cdots + a_n x_n(t)
  \]
  
  در این معادله، \( a_1, a_2, \dots, a_n \) ضرایب ثابت هستند و \( x_1(t), x_2(t), \dots, x_n(t) \) ورودی‌ها هستند.

- **سیستم‌های غیرخطی (Nonlinear Systems):** سیستم‌هایی که اصول ابراد و هم‌خطی را رعایت نمی‌کنند و خروجی به طور غیرخطی با ورودی‌ها مرتبط است.

 $$
  y(t) = f(x(t))
$$

  که در آن $$( f(x(t)) )$$ یک تابع غیرخطی است.

## 2. سیستم‌های زمان پیوسته و زمان گسسته (Continuous-Time and Discrete-Time Systems)

- **سیستم‌های زمان پیوسته (Continuous-Time Systems):** سیستم‌هایی که ورودی‌ها و خروجی‌های آن‌ها در زمان پیوسته تغییر می‌کنند. در این سیستم‌ها، ورودی و خروجی به صورت یک تابع پیوسته از زمان \( t \) تعریف می‌شوند.
  
 $$
  y(t) = T\{x(t)\}
  $$
  
  که در آن \( T \) یک عملگر سیستم است.

- **سیستم‌های زمان گسسته (Discrete-Time Systems):** سیستم‌هایی که ورودی‌ها و خروجی‌های آن‌ها در زمان‌های گسسته تغییر می‌کنند. این سیستم‌ها معمولاً در پردازش سیگنال‌های دیجیتال به کار می‌روند.
  
 $$
  y[n] = T\{x[n]\}
$$

## 3. سیستم‌های پایدار و ناپایدار (Stable and Unstable Systems)

- **سیستم‌های پایدار (Stable Systems):** سیستم‌هایی که در برابر اختلالات خارجی و تغییرات ورودی، پاسخ محدود و قابل پیش‌بینی تولید می‌کنند.

- **سیستم‌های ناپایدار (Unstable Systems):** سیستم‌هایی که به تغییرات ورودی‌ها یا اختلالات حساس هستند و ممکن است پاسخ‌های غیرمحدود یا غیرقابل پیش‌بینی تولید کنند.

## 4. سیستم‌های خطی زمانی-نامتغیر و زمانی-متغیر (Time-Invariant and Time-Varying Systems)

- **سیستم‌های خطی زمانی-نامتغیر (Time-Invariant Systems):** سیستم‌هایی که خواص آن‌ها در طول زمان تغییر نمی‌کند. به عبارت دیگر، اگر ورودی سیستم در زمان \( t_0 \) اعمال شود، همان خروجی در زمان $$( t_0 + \Delta t )$$ خواهد بود.

- **سیستم‌های خطی زمانی-متغیر (Time-Varying Systems):** سیستم‌هایی که خواص آن‌ها در طول زمان تغییر می‌کند.

## سیستم‌های باز و بسته (Open-Loop and Closed-Loop Systems)

### 1. سیستم‌های باز (Open-Loop Systems)

سیستم‌های باز به سیستم‌هایی اطلاق می‌شود که در آن‌ها هیچ‌گونه بازخوردی از خروجی به ورودی وجود ندارد. در این سیستم‌ها، ورودی به طور مستقیم به خروجی تبدیل می‌شود بدون آنکه خروجی تاثیری بر ورودی بگذارد.

معمولاً این سیستم‌ها ساده‌تر هستند و برای مواردی که نیاز به کنترل دقیق ندارند، استفاده می‌شوند. مثال معروف سیستم باز یک **ماکروویو** است که برای پخت غذا استفاده می‌شود. ورودی (دما و زمان) به طور مستقیم بر خروجی (پخت غذا) تاثیر می‌گذارد.

فرمول کلی سیستم باز:

$$
y(t) = T\{x(t)\}
$$

در اینجا \( T \) همان عملگر سیستم است که ورودی $$( x(t) )$$ را به خروجی $$( y(t) )$$ تبدیل می‌کند.

### 2. سیستم‌های بسته (Closed-Loop Systems)

در سیستم‌های بسته، خروجی به ورودی بازگشت داده می‌شود و از آن برای تنظیم یا تصحیح عملکرد سیستم استفاده می‌شود. این سیستم‌ها به سیستم‌های **کنترل بازخورد** یا **سیستم‌های خودتنظیم** معروف هستند.

در این سیستم‌ها، هرگونه تغییر در خروجی می‌تواند ورودی را تغییر دهد و عملکرد سیستم را اصلاح کند.

فرمول سیستم بسته با بازخورد به صورت زیر است:

$$
x(t) = x_{\text{in}}(t) + \beta y(t)
$$

که در آن:
- $$( x_{\text{in}}(t) )$$ ورودی اولیه است.
- $$( \beta y(t) )$$ بازخورد از خروجی به ورودی است.
  
سیستم‌های بسته معمولاً در کنترل و تنظیمات دقیق‌تر کاربرد دارند، مانند **سیستم‌های کنترل دما** یا **سیستم‌های راداری**.

### جمع‌بندی:
سیستم‌ها می‌توانند بر اساس ویژگی‌های مختلفی مانند خطی بودن، پیوستگی زمانی و نوع کنترل طبقه‌بندی شوند. سیستم‌های باز و بسته مهم‌ترین دسته‌بندی‌ها در سیستم‌های کنترل و مهندسی هستند.


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining a simple system (linear system for simplicity)
def open_loop_system(x):
    return x  # For simplicity, the output is the same as input (identity system)

def closed_loop_system(x, feedback_gain):
    y = open_loop_system(x)  # Get output from the open loop
    return y - feedback_gain * y  # Apply negative feedback to the output

# 2. Creating time vector (from 0 to 10 with small intervals)
t = np.linspace(0, 10, 1000)

# 3. Define input signal (for example, a step input)
x_in = np.ones_like(t)  # Constant input signal (step function)

# 4. Simulating Open-Loop System (no feedback)
y_open_loop = open_loop_system(x_in)

# 5. Simulating Closed-Loop System with a feedback gain of 0.5
feedback_gain = 0.5
y_closed_loop = closed_loop_system(x_in, feedback_gain)

# 6. Plotting the results
plt.figure(figsize=(12, 6))

# Plot Open-Loop Response
plt.subplot(1, 2, 1)
plt.plot(t, y_open_loop, label="Open-Loop System")
plt.title('Open-Loop System Response')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()

# Plot Closed-Loop Response
plt.subplot(1, 2, 2)
plt.plot(t, y_closed_loop, label="Closed-Loop System", color='orange')
plt.title('Closed-Loop System Response')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

# حافظه دار بودن و نبودن سیستم (Memory and Memoryless Systems)

سیستم‌ها می‌توانند بر اساس ویژگی‌های خود به دسته‌های مختلفی تقسیم شوند. یکی از مهم‌ترین ویژگی‌ها، وجود یا عدم وجود **حافظه** در سیستم است. در اینجا به بررسی سیستم‌های حافظه دار و سیستم‌های بدون حافظه می‌پردازیم.

## 1. سیستم‌های بدون حافظه (Memoryless Systems)

سیستم‌های بدون حافظه به سیستم‌هایی اطلاق می‌شود که خروجی آن‌ها تنها به ورودی آن در همان لحظه زمانی خاص وابسته است. به عبارت دیگر، در سیستم‌های بدون حافظه، وضعیت‌های گذشته یا آینده ورودی هیچ تأثیری بر خروجی ندارند.

فرمول کلی سیستم بدون حافظه به صورت زیر است:

$$
y(t) = T\{x(t)\}
$$

که در آن:
- \( y(t) \) خروجی سیستم است.
- \( x(t) \) ورودی سیستم است.
- \( T \) عملگر سیستم است که ورودی را به خروجی تبدیل می‌کند.

در این سیستم‌ها، مقدار خروجی در زمان \( t \) تنها به مقدار ورودی در همان زمان \( t \) وابسته است و هیچ تأثیری از ورودی‌های گذشته یا آینده ندارد.

### مثال:
یک مثال از سیستم بدون حافظه می‌تواند **مدار الکتریکی ساده** باشد که در آن خروجی (مثلاً ولتاژ) به طور مستقیم تابع ورودی (مانند جریان) در همان لحظه است.

## 2. سیستم‌های حافظه دار (Memory Systems)

سیستم‌های حافظه دار به سیستم‌هایی اطلاق می‌شود که خروجی آن‌ها نه تنها به ورودی فعلی بلکه به ورودی‌های گذشته و آینده نیز وابسته است. به عبارت دیگر، در سیستم‌های حافظه دار، وضعیت سیستم در زمان‌های گذشته بر خروجی آن تأثیر می‌گذارد.

فرمول کلی سیستم حافظه دار به صورت زیر است:

$$
y(t) = T\{x(t), x(t-\tau), x(t+\tau), \dots \}
$$

که در آن:
- \( y(t) \) خروجی سیستم است.
- $$( x(t), x(t-\tau), x(t+\tau), \dots )$$ ورودی‌های سیستم در زمان‌های مختلف هستند.
- \( T \) عملگر سیستم است که خروجی را از ورودی‌های مختلف در زمان‌های متفاوت محاسبه می‌کند.

### مثال:
یک مثال از سیستم حافظه دار می‌تواند **یک فیلتر دیجیتال** باشد که خروجی آن به ورودی‌های قبلی آن بستگی دارد. برای مثال، در یک فیلتر میانگین، خروجی فعلی ممکن است میانگینی از ورودی‌های گذشته باشد.

### جمع‌بندی:
- **سیستم‌های بدون حافظه**: خروجی فقط به ورودی فعلی وابسته است.
- **سیستم‌های حافظه دار**: خروجی به ورودی‌های گذشته یا آینده وابسته است.


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Defining the memoryless system
def memoryless_system(x):
    return x  # In a memoryless system, the output is the same as the input

# 2. Defining the memory system (using a simple moving average filter)
def memory_system(x, window_size=5):
    # A simple moving average filter as a memory system (with window size)
    return np.convolve(x, np.ones(window_size)/window_size, mode='same')

# 3. Creating a time vector (from 0 to 10 with small intervals)
t = np.linspace(0, 10, 1000)

# 4. Define input signal (a step input)
x_in = np.ones_like(t)  # A constant input signal (step function)

# 5. Simulating Memoryless System (output is the same as input)
y_memoryless = memoryless_system(x_in)

# 6. Simulating Memory System (output depends on previous inputs, e.g., moving average)
y_memory = memory_system(x_in, window_size=50)  # Using a larger window for averaging

# 7. Plotting the results
plt.figure(figsize=(12, 6))

# Plot Memoryless System Response
plt.subplot(1, 2, 1)
plt.plot(t, y_memoryless, label="Memoryless System", color='blue')
plt.title('Memoryless System Response')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()

# Plot Memory System Response (Moving Average)
plt.subplot(1, 2, 2)
plt.plot(t, y_memory, label="Memory System (Moving Average)", color='orange')
plt.title('Memory System Response')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

# سیستم‌های متغیر با زمان و ثابت با زمان (Time-Varying and Time-Invariant Systems)

سیستم‌ها را می‌توان از نظر وابستگی به زمان به دو دسته تقسیم کرد: **سیستم‌های متغیر با زمان** و **سیستم‌های ثابت با زمان**. این دسته‌بندی بر اساس این است که آیا ویژگی‌های سیستم در طول زمان تغییر می‌کند یا خیر.

## 1. سیستم‌های ثابت با زمان (Time-Invariant Systems)

سیستم‌های ثابت با زمان به سیستم‌هایی اطلاق می‌شود که ویژگی‌های آن‌ها در طول زمان تغییر نمی‌کند. به عبارت دیگر، اگر ورودی سیستم در زمان \( t_0 \) اعمال شود و سیستم رفتار خاصی از خود نشان دهد، آنگاه اگر ورودی مشابهی در زمان \( t_0 + Delta t \) اعمال شود، سیستم همان رفتار را از خود نشان خواهد داد.

در این سیستم‌ها، پاسخ سیستم به ورودی در زمان‌های مختلف مشابه است و هیچ‌گونه تغییر در خواص سیستم در طول زمان مشاهده نمی‌شود.

فرمول عمومی سیستم ثابت با زمان:

$$
y(t) = T\{x(t)\}
$$

که در آن:
- \( y(t) \) خروجی سیستم است.
- \( x(t) \) ورودی سیستم است.
- \( T \) عملگر سیستم است که ورودی را به خروجی تبدیل می‌کند.

### مثال:
یک مثال ساده از سیستم ثابت با زمان می‌تواند یک **فیلتر خطی** باشد که خواص آن در طول زمان تغییر نمی‌کند.

## 2. سیستم‌های متغیر با زمان (Time-Varying Systems)

سیستم‌های متغیر با زمان به سیستم‌هایی اطلاق می‌شود که ویژگی‌های آن‌ها در طول زمان تغییر می‌کند. در این سیستم‌ها، رفتار سیستم بسته به زمانی که ورودی در آن اعمال می‌شود تغییر می‌کند. به عبارت دیگر، خواص سیستم در زمان‌های مختلف متفاوت است.

فرمول عمومی سیستم متغیر با زمان:

$$
y(t) = T\{x(t), t\}
$$

که در آن:
- \( y(t) \) خروجی سیستم است.
- \( x(t) \) ورودی سیستم است.
- \( t \) زمان است که بر خواص سیستم تأثیر می‌گذارد.
- \( T \) عملگر سیستم است که ورودی را به خروجی تبدیل می‌کند و بسته به زمان \( t \) تغییر می‌کند.

### مثال:
یک مثال از سیستم متغیر با زمان می‌تواند یک **سیستم فیلتر زمانی متغیر** باشد که در آن پارامترهای فیلتر (مانند ضرایب فیلتر) با زمان تغییر می‌کنند.

### جمع‌بندی:
- **سیستم‌های ثابت با زمان**: خواص سیستم در طول زمان تغییر نمی‌کند.
- **سیستم‌های متغیر با زمان**: خواص سیستم در طول زمان تغییر می‌کند.


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define Time-Invariant System (Constant system)
def time_invariant_system(x):
    return x  # The output is the same as the input, no time dependence

# 2. Define Time-Varying System (Variable system)
def time_varying_system(x, t):
    return x * np.cos(t)  # The output depends on both input and time (cosine modulation)

# 3. Create a time vector (from 0 to 10 with small intervals)
t = np.linspace(0, 10, 1000)

# 4. Define input signal (for example, a step input)
x_in = np.ones_like(t)  # Constant input signal (step function)

# 5. Simulating Time-Invariant System (output is the same as input)
y_time_invariant = time_invariant_system(x_in)

# 6. Simulating Time-Varying System (output depends on time and input)
y_time_varying = time_varying_system(x_in, t)

# 7. Plotting the results
plt.figure(figsize=(12, 6))

# Plot Time-Invariant System Response
plt.subplot(1, 2, 1)
plt.plot(t, y_time_invariant, label="Time-Invariant System", color='blue')
plt.title('Time-Invariant System Response')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()

# Plot Time-Varying System Response
plt.subplot(1, 2, 2)
plt.plot(t, y_time_varying, label="Time-Varying System", color='orange')
plt.title('Time-Varying System Response')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

# علیت سیستم (Causality of a System)

سیستم‌ها را می‌توان بر اساس ویژگی‌های مختلف آن‌ها، از جمله **علیت**، دسته‌بندی کرد. یک سیستم **علّی** به سیستمی اطلاق می‌شود که خروجی آن تنها به ورودی‌های گذشته یا حال بستگی دارد و به ورودی‌های آینده وابسته نیست.

## 1. تعریف علیت (Causal System)

سیستم‌های **علّی** سیستمی هستند که در آن خروجی در هر زمان فقط به ورودی‌های گذشته و حال بستگی دارد، نه ورودی‌های آینده. به عبارت دیگر، اگر ورودی در زمان \( t_0 \) اعمال شود، خروجی در زمان \( t_0 \) یا زمان‌های قبلی می‌تواند تحت تأثیر قرار گیرد، اما ورودی‌های آینده تأثیری بر خروجی ندارند.

فرمول عمومی یک سیستم علّی به شکل زیر است:

$$
y(t) = T\{x(t), x(t-\tau), x(t-2\tau), \dots \}
$$

که در آن:
- \( y(t) \) خروجی سیستم در زمان \( t \) است.
- $$( x(t), x(t-\tau), x(t-2\tau), \dots )$$ ورودی‌ها در زمان‌های مختلف هستند.
- \( T \) عملگر سیستم است که ورودی‌ها را به خروجی تبدیل می‌کند.

در این سیستم، خروجی در زمان \( t \) به ورودی‌های گذشته و حال بستگی دارد، اما به ورودی‌های آینده وابسته نیست.

### مثال:
یک مثال از سیستم علّی می‌تواند **سیستم کنترل دما** باشد که دما را به طور مداوم اندازه‌گیری می‌کند و بر اساس مقادیر قبلی و کنونی دما، تصمیماتی برای تنظیم دما می‌گیرد.

## 2. سیستم‌های غیر علّی (Non-Causal Systems)

سیستم‌های **غیر علّی** به سیستمی اطلاق می‌شود که خروجی آن به ورودی‌های آینده وابسته است. در این سیستم‌ها، برای تعیین خروجی در زمان \( t \) به ورودی‌های آینده نیز نیاز داریم.

فرمول عمومی یک سیستم غیر علّی به شکل زیر است:

$$
y(t) = T\{x(t), x(t+\tau), x(t+2\tau), \dots \}
$$

که در آن:
- \( y(t) \) خروجی سیستم در زمان \( t \) است.
- $$( x(t), x(t+\tau), x(t+2\tau), \dots )$$ ورودی‌ها در زمان‌های مختلف هستند.

در این سیستم، خروجی به ورودی‌های آینده بستگی دارد.

### مثال:
یک مثال از سیستم غیر علّی می‌تواند **سیستم پیش‌بینی وضعیت بازار** باشد که برای پیش‌بینی وضعیت آینده بازار به داده‌های آینده (که هنوز نیامده‌اند) نیاز دارد.

### جمع‌بندی:
- **سیستم‌های علّی**: خروجی به ورودی‌های گذشته و حال بستگی دارد.
- **سیستم‌های غیر علّی**: خروجی به ورودی‌های آینده نیز بستگی دارد.


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define Causal System (only depends on current and past inputs)
def causal_system(x):
    # Output depends only on current and past inputs (x(t), x(t-1), x(t-2), ...)
    return x

# 2. Define Non-Causal System (depends on current and future inputs)
def non_causal_system(x):
    # Output depends on current and future inputs (x(t), x(t+1), x(t+2), ...)
    # In practice, this is non-causal and only for illustration
    n = len(x)
    y = np.zeros_like(x)
    for i in range(n-2):
        y[i] = x[i+1]  # Shifted by one time unit (future)
    return y

# 3. Create a time vector (from 0 to 10 with small intervals)
t = np.linspace(0, 10, 1000)

# 4. Define input signal (e.g., a step input signal)
x_in = np.ones_like(t)  # Constant input signal (step function)

# 5. Simulating Causal System (output depends on past and present inputs)
y_causal = causal_system(x_in)

# 6. Simulating Non-Causal System (output depends on future inputs)
y_non_causal = non_causal_system(x_in)

# 7. Plotting the results
plt.figure(figsize=(12, 6))

# Plot Causal System Response
plt.subplot(1, 2, 1)
plt.plot(t, y_causal, label="Causal System", color='blue')
plt.title('Causal System Response')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()

# Plot Non-Causal System Response
plt.subplot(1, 2, 2)
plt.plot(t, y_non_causal, label="Non-Causal System", color='orange')
plt.title('Non-Causal System Response')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

# ویژگی‌های مستقل از زمان: پایداری، خطی بودن و معکوس‌پذیری

سیستم‌ها به طور کلی بر اساس ویژگی‌های مختلفشان ارزیابی می‌شوند. در این بخش، سه ویژگی مهم **پایداری**، **خطی بودن** و **معکوس‌پذیری** را بررسی می‌کنیم که از ویژگی‌های **مستقل از زمان** به حساب می‌آیند.

## 1. پایداری (Stability)

سیستم پایدار به سیستمی اطلاق می‌شود که اگر ورودی آن محدود (محدود به یک بازه خاص) باشد، خروجی آن نیز محدود باقی می‌ماند. به عبارت دیگر، در سیستم‌های پایدار، هر ورودی محدود باعث تولید یک خروجی محدود می‌شود.

### انواع پایداری:
- **پایداری بُرگ (Bounded-Input, Bounded-Output Stability - BIBO)**: اگر ورودی به سیستم محدود باشد، خروجی نیز محدود خواهد بود.
  
  فرمول ریاضی پایداری به صورت زیر است:
 $$
  |x(t)| \leq B_x \quad \Rightarrow \quad |y(t)| \leq B_y
 $$
  که در آن:
  - \( |x(t)| \) ورودی محدود است.
  - \( |y(t)| \) خروجی محدود است.
  - \( B_x \) و \( B_y \) ثابت‌های محدود کننده هستند.

### مثال:
یک فیلتر خطی که ورودی‌های بزرگ را به خروجی‌های بزرگ تبدیل نمی‌کند، می‌تواند یک سیستم پایدار باشد.

## 2. خطی بودن (Linearity)

یک سیستم خطی به سیستمی اطلاق می‌شود که خواص **افزایشی** و **همسان‌سازی** را رعایت کند. به عبارت دیگر، اگر دو ورودی به سیستم اعمال شوند، خروجی آن‌ها باید معادل مجموع خروجی‌های هر یک از ورودی‌ها باشد. همچنین اگر ورودی به سیستم یک مقیاس تغییر کند، خروجی نیز به همان نسبت تغییر خواهد کرد.

### فرمول خطی بودن:
برای هر دو ورودی $$( x_1(t) ) و ( x_2(t) )$$ و ضرایب ثابت \( a \) و \( b \)، سیستم باید رابطه زیر را برقرار کند:
$$
T\{a x_1(t) + b x_2(t)\} = a T\{x_1(t)\} + b T\{x_2(t)\}
$$

### مثال:
یک سیستم فیلتر که سیگنال ورودی را تقویت یا تضعیف می‌کند، همچنان خطی خواهد بود، زیرا حاصل جمع ورودی‌ها به طور مساوی در خروجی‌ها تأثیر می‌گذارد.

## 3. معکوس‌پذیری (Invertibility)

سیستم معکوس‌پذیر به سیستمی اطلاق می‌شود که برای هر ورودی، بتوان خروجی معکوس آن را تعیین کرد. به عبارت دیگر، اگر سیستم به ورودی خاصی پاسخ دهد، می‌توان به راحتی ورودی اصلی را از خروجی سیستم بازیابی کرد.

### شرط معکوس‌پذیری:
یک سیستم معکوس‌پذیر است اگر برای هر ورودی \( x(t) \) و خروجی \( y(t) \)، معکوس آن وجود داشته باشد، به طوری که:
$$
T^{-1}\{y(t)\} = x(t)
$$
که در آن $$( T^{-1} )$$ عملگر معکوس سیستم است.

### مثال:
یک سیستم فیلتر که از ورودی به خروجی یک تابع خطی می‌سازد، ممکن است معکوس‌پذیر باشد اگر بتوان با استفاده از خروجی، ورودی اصلی را بازسازی کرد.

### جمع‌بندی:
- **پایداری**: سیستم‌های پایدار خروجی‌های محدود برای ورودی‌های محدود تولید می‌کنند.
- **خطی بودن**: سیستم‌های خطی خواص افزایشی و همسان‌سازی دارند.
- **معکوس‌پذیری**: سیستم‌های معکوس‌پذیر برای هر ورودی خاص، یک خروجی معکوس قابل تعیین دارند.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define a Bounded-Input, Bounded-Output (BIBO) Stable System
def stable_system(x):
    # A simple system where output is the same as the input, ensuring stability
    return x

# Create a time vector (from 0 to 10 with small intervals)
t = np.linspace(0, 10, 1000)

# Define a bounded input signal (step input)
x_in = np.ones_like(t)  # Constant input signal (bounded)

# Simulate the stable system
y_stable = stable_system(x_in)

# Plotting the results
plt.figure(figsize=(8, 4))
plt.plot(t, y_stable, label="Stable System", color='blue')
plt.title('Bounded-Input, Bounded-Output Stability')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define a linear system (output is a weighted sum of inputs)
def linear_system(x, a, b):
    # System response is a linear combination of inputs
    return a * x + b

# Create a time vector (from 0 to 10 with small intervals)
t = np.linspace(0, 10, 1000)

# Define two input signals
x1 = np.sin(t)  # First input signal (sine wave)
x2 = np.cos(t)  # Second input signal (cosine wave)

# Simulate the linear system with two different inputs
a, b = 2, 1  # Weights for the linear combination
y_linear1 = linear_system(x1, a, b)
y_linear2 = linear_system(x2, a, b)

# Simulate the response for a linear combination of both inputs
y_combined = linear_system(x1 + x2, a, b)  # Add the inputs first, then pass to system

# Plotting the results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(t, y_linear1, label="Linear System Response for x1(t)", color='green')
plt.plot(t, y_linear2, label="Linear System Response for x2(t)", color='orange')
plt.title('Linear System Responses')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(t, y_combined, label="Linear Combination Response", color='purple')
plt.title('Linear Combination of Inputs')
plt.xlabel('Time (t)')
plt.ylabel('Output y(t)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Define an invertible system (a simple linear system with a known inverse)
def invertible_system(x):
    # For simplicity, assume the system is y = 2 * x (scaling)
    return 2 * x

# Create a time vector (from 0 to 10 with small intervals)
t = np.linspace(0, 10, 1000)

# Define an input signal
x_in = np.sin(t)  # Example input signal (sine wave)

# Simulate the invertible system (output is twice the input)
y_invertible = invertible_system(x_in)

# Inverse operation (since y = 2 * x, x = y / 2)
x_recovered = y_invertible / 2

# Plotting the results
plt.figure(figsize=(8, 4))
plt.plot(t, x_in, label="Original Input x(t)", color='blue', linestyle='dashed')
plt.plot(t, y_invertible, label="System Output y(t)", color='green')
plt.plot(t, x_recovered, label="Recovered Input (Inverse)", color='red', linestyle='dotted')
plt.title('Invertible System and Inverse Recovery')
plt.xlabel('Time (t)')
plt.ylabel('Signal Value')
plt.legend()
plt.grid(True)
plt.show()
```
