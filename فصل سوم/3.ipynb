# مقدمه :

 همانگونه که گفتیم هدف ما در این کتاب تجزیه و تحلیل سیستم های LTI است. روش حوزه زمان در تجزیه و تحلیل این سیستم ها با همه وضوح و روشنی در بعضی موارد به محاسبات پیچیده و خسته کننده منتهی میشود. این امر باعث شده است تا روش تجزیه و تحلیل فوریه در این موارد به کار گرفته شود.
 
 البته تنها بدلیل سادگی این روش مورد توجه قرار نگرفته است بلکه دلیل عمده توجه متخصصان به تحلیل فوریه در تجزیه و تحلیل سیستمهای LTI این است که در واقع یک نوع برداشت جدید از سیستم ها به خواننده میدهد و خواننده را به دنیای جدیدی به نام حوزه فرکانس برده و او را در این دنیای جدید با خواص سیستم ها آشنا میسازد. البته لازم است توجه شود که تحلیل فوریه فقط و فقط در مورد سیستمهای خطی کاربرد دارد تبدیل فوریه در واقع ابتدا به عنوان یک ابزار ریاضی کارآمد جهت تفسیر برخی از پدیده های فیزیکی و همچنین حل برخی مسایل پیچیده ریاضی ارائه گردید. تبدیل فوریه عملگری است که برخی معادلات پیچیده را به معادلات ساده جبری تبدیل می نماید.
 
 اساس کار این عملگر بسط یک تابع بر اساس مؤلفه های فرکانسی به صورت e <sup>jWt</sup> است. اهمیت تحلیل فوریه در تجزیه و تحلیل سیستم ها و سیگنالها نیز از همین جا ناشی شده است. توابعی به صورت e <sup>jWt</sup> نوع خاص و بسیار مهمی از توابع هستند که اگر پاسخ سیستم خطی به این توابع را داشته باشیم اطلاعات مهمی در مورد سیستم و خواص آن میتوان بدست آورد. این خواص به خواص سیستم در حوزه فرکانس موسوم هستند و از روی آنها میتوان سایر خواص سیستم را نیز بدست آورد. جالب است توجه کنیم که پاسخ سیستم خطی به این گونه توابع دقیقا مشابه ورودی به استثناء یک ضریب ثابت مختلط است. بعبارت دیگر توابعی بصورت e <sup>jWt</sup> توابع ویژه سیستم های LTI هستند.
 
 بنابراین مشاهده میشود که تحلیل فوریه یک ابزار کارآمد و سریع در بدست آوردن خواص سیستم های خطی و تجزیه و تحلیل عملکرد آنهاست در این فصل توجه خود را به تجزیه و تحلیل سیگنالها و سیستم های پیوسته زمان معطوف میداریم و در فصل چهارم به بررسی سیستمها و دنباله های گسسته زمان می پردازیم در اینجا جهت یادآوری لازم است تعریف انرژی و توان را برای سیگنال های انرژی و توان که در فصل اول ارائه شد را مجددا تکرار کنیم.

برای رسم سیگنال قطار پله (Stair Signal) در پایتون، می‌توانید از کتابخانه‌های matplotlib و numpy استفاده کنید. در این مثال، ما یک سیگنال ساده پله‌ای را ایجاد خواهیم کرد:

```python
# 
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
x = np.arange(0, 10, 0.1)
y = np.floor(x)  # ایجاد سیگنال پله‌ای

# رسم سیگنال
plt.step(x, y, where='post', label='Stair Signal', color='blue')
plt.title('Stair Signal')
plt.xlabel('Time')
plt.ylabel('Signal Level')
plt.grid(True)
plt.legend()
plt.ylim(-1, 10)
plt.xlim(0, 10)
plt.show()
```

برای رسم سیگنال ضربه‌ای (Impulse Signal) در پایتون، می‌توانید از کتابخانه‌های matplotlib و numpy استفاده کنید. در اینجا یک نمونه کد برای ایجاد و رسم سیگنال ضربه‌ای آورده شده است

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
n = np.arange(0, 20)  # محور زمان
impulse_signal = np.zeros_like(n)  # ایجاد آرایه با صفر
impulse_signal[10] = 1  # ایجاد سیگنال ضربه‌ای در نقطه n=10

# رسم سیگنال ضربه‌ای
plt.stem(n, impulse_signal, basefmt=" ", use_line_collection=True)
plt.title('Impulse Signal')
plt.xlabel('n (Time index)')
plt.ylabel('Amplitude')
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.show()
```

بنابراین میبینیم که اگر ، با توجه به اینکه مرتبه میل کردن A بسمت صفر و T بسمت بی نهایت از یک مرتبه است حاصلضرب آنها عدد ثابت و محدودی است پس سیگنال ضربه یک سیگنال توان است.

برای رسم سیگنال پالس (Pulse Signal) در پایتون، ما می‌توانیم یک سیگنال که در یک بازه زمانی مشخص مقدار غیر صفر داشته باشد (مثل یک پالس واحد) و خارج از آن صفر باشد، ایجاد کنیم. این سیگنال مشابه یک تابع هِویساید (Heaviside) یا پالس واحد است.

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
n = np.arange(0, 20)  # محور زمان
pulse_signal = np.zeros_like(n)  # ایجاد آرایه صفر
pulse_signal[5:15] = 1  # تعریف پالس بین ایندکس‌های 5 تا 15

# رسم سیگنال پالس
plt.stem(n, pulse_signal, basefmt=" ", use_line_collection=True)
plt.title('Pulse Signal')
plt.xlabel('n (Time index)')
plt.ylabel('Amplitude')
plt.ylim(-0.5, 1.5)
plt.xlim(0, 20)
plt.grid(True)
plt.show()
```

برای رسم یک سیگنال مثلثی پیوسته (Continuous Triangular Signal) در پایتون، می‌توانیم از توابع numpy و matplotlib استفاده کنیم. سیگنال مثلثی معمولاً برای مدل‌سازی سیگنال‌هایی به کار می‌رود که دارای تغییرات تدریجی و منظم هستند. سیگنال مثلثی به صورت یک سری از افزایش و کاهش خطی از یک مقدار به مقدار دیگر در بازه‌ای مشخص، شکل می‌گیرد.

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
t = np.linspace(0, 10, 1000)  # محور زمان: از 0 تا 10 با 1000 نقطه
triangle_signal = 1 - 2 * np.abs(np.mod(t, 2) - 1)  # سیگنال مثلثی

# رسم سیگنال
plt.plot(t, triangle_signal, label='Continuous Triangular Signal', color='green')
plt.title('Continuous Triangular Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

##### 1 = ||v(t)||  


دقیقاً مشابه آنچه که در مورد تجزیه هر بردار به بردارهای یکه متعامد میدانیم هر سیگنال را میتوان به سیگنالهای متعامد یکه تجزیه کرد بدین ترتیب هر عملیات مطلوب را میتوان روی سیگنالهای متعامد یکه انجام داد البته چون عملیات روی سیگنالهای متعامد یکه ساده تر بوده و سریعتر به جواب می رسد تجزیه سیگنال به سیگنالهای متعامد یکه بسیار رایج است.

## کد پایتون برای رسم سیگنال پالس و هارمونیک‌های آن:
در اینجا ما از توابع سینوسی برای نشان دادن هارمونیک‌های یک پالس مستطیلی استفاده می‌کنیم:

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
A = 1  # دامنه
T = 1  # عرض پالس
f0 = 1 / T  # فرکانس بنیادین
t = np.linspace(-2, 3, 1000)  # محور زمان

# تعریف سیگنال پالس
x_t = A * ((t >= 0) & (t < T))

# محاسبه توابع سینوسی برای 5 هارمونیک اول
harmonics = 5
sinusoids = np.zeros((harmonics, len(t)))

for n in range(harmonics):
    k = n * 2 + 1  # تنها هارمونیک‌های فرد
    sinusoids[n] = (A * (4 / (np.pi * k))) * np.sin(2 * np.pi * k * f0 * t)

# جمع کردن هارمونیک‌ها
fourier_representation = np.sum(sinusoids, axis=0)

# رسم سیگنال
plt.figure(figsize=(12, 6))

# سیگنال پالس
plt.subplot(3, 1, 1)
plt.plot(t, x_t, label='Rectangular Pulse', color='red')
plt.title('Rectangular Pulse Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.ylim(-0.5, 1.5)
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.grid(True)
plt.legend()

# هارمونیک‌ها
for n in range(harmonics):
    k = n * 2 + 1  # تنها هارمونیک‌های فرد
    plt.subplot(3, 1, 2)
    plt.plot(t, sinusoids[n], label=f'Harmonic {k}', linestyle='--')

plt.title('Harmonics of the Rectangular Pulse')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.ylim(-0.5, 0.5)
plt.grid(True)
plt.legend()

# تقریب سری فوریه
plt.subplot(3, 1, 3)
plt.plot(t, fourier_representation, label='Fourier Series Approximation', color='blue')
plt.title('Fourier Series Approximation of the Rectangular Pulse')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

برای رسم یک سیگنال سینوسی میراشونده در پایتون، می‌توانیم از کتابخانه‌های numpy و matplotlib استفاده کنیم. سیگنال سینوسی میراشونده به معنای ترکیب یک سیگنال سینوسی با یک تابع نمایی است که موجب کاهش تدریجی دامنه سیگنال می‌شود.

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
t = np.linspace(0, 10, 1000)  # محور زمان: از 0 تا 10 با 1000 نقطه
frequency = 1  # فرکانس سیگنال (هرتز)
damping_factor = 0.1  # ضریب میرایی

# محاسبه سیگنال سینوسی میراشونده
damped_sine_wave = np.exp(-damping_factor * t) * np.sin(2 * np.pi * frequency * t)

# رسم سیگنال
plt.plot(t, damped_sine_wave, label='Damped Sine Wave', color='blue')
plt.title('Damped Sine Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

# تقریب سیگنال از روی سری فوریه : 
از لحاظ ریاضی رابطه بسط سری فوریه یک رابطه تحقیقی است اما اگر حدود مجموع آن را محدود کنیم در آن صورت یک تقریب از X(t) بدست می آید. 
اگر 2N + 1 جمله از سری فوریه را در نظر بگیریم در این صورت یک سیگنال جدید خواهیم داشت. که با سیگنال اصلی (t)x) متفاوت است. 


<img src='Capture9.PNG'/></a>

از یک دیدگاه مقدار عددی خطا زیاد برایمان مهم نیست بلکه انرژی موجود در آن برایمان مهم است. انرژی خطا بصورت زیر تعریف می گردد.  

<img src='Capture10.PNG'/></a>

هر چه انرژی خطا کوچکتر باشد (t)X شباهت بیشتری به .(t)X دارد. البته این نوع شباهت تضمینی در مورد شباهت مقادیر لحظه ای دو سیگنال ایجاد نمیکند در حقیقت ممکن است در لحظات منفصل و محدود مقدار تفاضل دو سیگنال بزرگ شود اما سطح زیر منحنی مربع تفاضل مقداری محدود باشد و انتظار داریم هر چه N بزرگتر شود این مقدار نیز نسبت صفر میل نماید. 

بیایید یک نمونه از سیگنال زمان پیوسته را با در نظر گرفتن سری فوریه آن بررسی کنیم. این کار به ما کمک می‌کند تا چگونگی تجزیه یک سیگنال پیچیده به مجموعه‌ای از سیگنال‌های ساده‌تر (سینوسی و کسینوسی) را درک کنیم.

برای این مثال، از یک سیگنال دندان اره‌ای (sawtooth wave) استفاده می‌کنیم. سیگنال دندان اره‌ای یک سیگنال متناوب است که به‌طور خطی افزایش می‌یابد و سپس به صورت ناگهانی به مقدار اولیه خود برمی‌گردد.


## کد پایتون برای رسم سیگنال دندان اره‌ای و تقریب آن با سری فوریه:

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
T = 2  # دوره تناوب (ثانیه)
f = 1/T # فرکانس
t = np.linspace(0, 10, 1000)  # محور زمان: از 0 تا 10 با 1000 نقطه
N = 10  # تعداد هارمونیک‌های سری فوریه

# محاسبه سیگنال دندان اره‌ای
sawtooth_signal = (t % T) / T 


# محاسبه تقریب سری فوریه سیگنال دندان اره‌ای
fourier_approx = np.zeros_like(t)
for n in range(1, N + 1):
    fourier_approx += (2 / (n * np.pi)) * (-1)**(n+1) * np.sin(2 * np.pi * n * f * t)


# رسم سیگنال اصلی و تقریب سری فوریه
plt.figure(figsize=(10, 6))
plt.plot(t, sawtooth_signal, label='Sawtooth Signal', color='red')
plt.plot(t, fourier_approx, label='Fourier Series Approximation', color='blue', linestyle='--')
plt.title('Sawtooth Signal and Fourier Series Approximation')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

بسیار خب، بیایید یک مثال دیگر از سیگنال زمان پیوسته با در نظر گرفتن سری فوریه آن بررسی کنیم. این بار، ما از یک سیگنال موج مربعی (square wave) استفاده خواهیم کرد. موج مربعی یک سیگنال متناوب است که بین دو سطح مقدار ثابت متناوباً تغییر می‌کند.

## کد پایتون برای رسم سیگنال موج مربعی و تقریب آن با سری فوریه:

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
T = 2  # دوره تناوب (ثانیه)
f = 1/T # فرکانس
t = np.linspace(0, 10, 1000)  # محور زمان: از 0 تا 10 با 1000 نقطه
N = 10  # تعداد هارمونیک‌های سری فوریه

# محاسبه سیگنال موج مربعی
square_signal = np.sign(np.sin(2 * np.pi * f * t))

# محاسبه تقریب سری فوریه سیگنال موج مربعی
fourier_approx = np.zeros_like(t)
for n in range(1, N + 1, 2):  # فقط هارمونیک‌های فرد را محاسبه می‌کنیم
    fourier_approx += (4 / (n * np.pi)) * np.sin(2 * np.pi * n * f * t)

# رسم سیگنال اصلی و تقریب سری فوریه
plt.figure(figsize=(10, 6))
plt.plot(t, square_signal, label='Square Wave', color='red')
plt.plot(t, fourier_approx, label='Fourier Series Approximation', color='blue', linestyle='--')
plt.title('Square Wave and Fourier Series Approximation')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

برای رسم یک سیگنال “خوش‌رفتار” (well-behaved signal)، ما به یک سیگنال پیوسته و با تغییرات ملایم نیاز داریم که در عین حال اطلاعات مفیدی را ارائه دهد. یکی از بهترین مثال‌ها برای چنین سیگنالی، یک سیگنال سینوسی میراشونده است که با یک فرکانس متغیر در طول زمان ترکیب شده است. این سیگنال هم تغییرات سینوسی دارد و هم به تدریج دامنه آن کاهش می‌یابد، و تغییر فرکانس آن را جالب‌تر می‌کند.

## کد پایتون برای رسم یک سیگنال خوش‌رفتار با فرکانس متغیر و میرایی:

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
t = np.linspace(0, 10, 1000)  # محور زمان: از 0 تا 10 با 1000 نقطه
initial_frequency = 0.5  # فرکانس اولیه (هرتز)
final_frequency = 2  # فرکانس نهایی (هرتز)
damping_factor = 0.1  # ضریب میرایی

# محاسبه فرکانس متغیر با زمان
frequency_t = initial_frequency + (final_frequency - initial_frequency) * t / 10

# محاسبه سیگنال خوش‌رفتار
well_behaved_signal = np.exp(-damping_factor * t) * np.sin(2 * np.pi * frequency_t * t)

# رسم سیگنال
plt.plot(t, well_behaved_signal, label='Well-Behaved Signal', color='purple')
plt.title('Well-Behaved Signal with Variable Frequency and Damping')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

## کد پایتون برای رسم موج مربعی با میرایی:

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات سیگنال
t = np.linspace(0, 10, 1000)  # محور زمان: از 0 تا 10 با 1000 نقطه
frequency = 1  # فرکانس موج مربعی (هرتز)
damping_factor = 0.1  # ضریب میرایی

# تابع موج مربعی با میرایی
square_wave = 0.5 * (1 + np.sign(np.sin(2 * np.pi * frequency * t)))  # موج مربعی
damped_square_wave = square_wave * np.exp(-damping_factor * t)  # موج مربعی با میرایی

# رسم سیگنال
plt.plot(t, damped_square_wave, label='Damped Square Wave', color='blue')
plt.title('Damped Square Wave Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
