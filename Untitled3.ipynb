که همان عبارتی است که در بخش ۱-۴ به دست آوردیم معادله ی (۱-۶۷) را ببینید]. معادله ی (۲-۲) را خاصیت غربالی ضربه ی واحد گسسته در زمان مینامند چون رشته ی [n - k] تنها به ازای k = n غیر صفر است جمع سمت راست معادلهی (۲-۲) رشته ی []x را غربال میکند و تنها مقدار متناظر " = x را عبور میدهد در زیر بخش بعد این نمایش سیگنالهای گسسته در زمان را برای نمایش جمع کانولوشن سیستمهای LTI گسسته در زمان به کار میبریم.

```python
import numpy as np
import matplotlib.pyplot as plt

def delta(n):
    """Approximation of the Dirac delta function."""
    if n == 0:
        return 1  # or a very large value
    else:
        return 0

# Example usage
n_values = np.arange(-5, 6)
y_values = [delta(n) for n in n_values]

# Plotting the approximation of the delta function
plt.stem(n_values, y_values)
plt.title('Approximation of the Dirac Delta Function')
plt.xlabel('n')
plt.ylabel('δ(n)')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def u(t):
    """Unit step function."""
    return np.where(t < 0, 0, 1)

# Generate time values
t = np.linspace(-5, 5, 1000)  # from -5 to 5

# Calculate the unit step function values
x_t = u(t)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(t, x_t, label='u(t)', color='blue')
plt.title('Unit Step Function u(t)')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.axhline(0, color='black', lw=0.5, ls='--', alpha=0.7)
plt.axvline(0, color='black', lw=0.5, ls='--', alpha=0.7)
plt.grid()
plt.legend()
plt.ylim(-0.1, 1.1)
plt.xlim(-5, 5)
plt.show()
```
