## قضیه نمونه برداری

###  در حالت کلی انتظار نداریم که بدون وجود شرایط یا اطلاعات اضافی سیگنالی به طور یکتا توسط رشته ای از نمونه های متساوی الفاصله اش مشخص شود. برای مثال شکل ۷-۱ سه سیگنال پیوسته در زمان متفاوت را نشان می دهد که مقادیرشان در مضارب صحیح T یکسان است ؛ یعنی $$ x_1(kT) = x_2(kT) = x_3(kT) $$ 
###   واضح است که تعداد سیگنالهایی که نمونه های مشابهی تولید میکنند بینهایت است. ولی خواهیم دید که اگر باند یک سیگنال محدود باشد، یعنی تبدیل فوریه آن در خارج یک باند فرکانسی محدود صفر باشد، و اگر نمونه ها به حد کافی به هم نزدیک باشند آنگاه نمونه ها به طور یکتا سیگنال را مشخص میکنند، و می توان سیگنال را به طور کامل بازسازی کرد. نزدیکی نمونه ها با بزرگترین فرکانس موجود در سیگنال ارتباط دارد. این گزاره را قضیه نمونه برداری مینامند و در کاربردهای عملی روشهای تحلیل سیگنالها و سیستمها اهمیتی شایان دارد.





```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف پارامترها
T = 1
t = np.arange(-3*T, 3*T, 0.01)

# تعریف سیگنال‌ها
x1 = np.sin(2 * np.pi * t / T)
x2 = np.cos(2 * np.pi * t / T)
x3 = np.sin(2 * np.pi * t / T) * np.cos(2 * np.pi * t / T)

# رسم سیگنال‌ها
plt.plot(t, x1, 'r', label='x_1(t)')
plt.plot(t, x2, '--g', label='x_2(t)')
plt.plot(t, x3, ':b', label='x_3(t)')

# تنظیمات نمودار
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.title('Three continuous-time signals with identical values at integer multiples of T')
plt.suptitle('7-1', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True)
plt.show()
```

## نمونه‌برداری با قطار ضربه
###     برای بیان قضیه نمونه‌برداری به راه مناسبی برای بیان نمونه‌های یک سیگنال پیوسته در زمان در فاصله‌های مساوی نیاز داریم.  یک راه مناسب برای انجام این کار استفاده از حاصل‌ضرب یک قطار ضربه متناوب و سیگنال پیوسته در زمان  $x(t) $ است. این فرآیند، که نمونه‌برداری با قطار ضربه نام دارد، در رابطه زیر تعریف می‌شود:
### $$ x_p(t) = x(t) \cdot p(t)$$
###  که در آن: $ p(t) = \sum_{{n=-\infty}}^{{\infty}} \delta(t - nT)$ و فرکانس نمونه‌برداری به صورت زیر تعریف می‌شود:

### $$ \Omega_s = \frac{{2\pi}}{{T}} $$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.image as mpimg

# تنظیمات کلی برای شکل
plt.rcParams.update({
    "text.usetex": False,  # غیرفعال کردن LaTeX
    "font.family": "serif",
    "axes.grid": False
})

# 1. تعریف تابع اصلی x(t)
t = np.linspace(-1, 1, 1000)
x_t = np.sin(2 * np.pi * t) + 0.5 * np.sin(6 * np.pi * t)

# 2. تعریف قطار ضربه p(t)
T = 0.2  # فاصله نمونه‌برداری
p_t = np.zeros_like(t)
p_t[::int(len(t) * T / (t[-1] - t[0]))] = 1  # ضربه‌ها در فواصل T

# 3. نتیجه نمونه‌برداری x_p(t)
x_p_t = x_t * p_t

# رسم نمودارها
fig, axs = plt.subplots(4, 1, figsize=(8, 12))  # تغییر اندازه شکل

# نمودار 1: x(t)
axs[0].plot(t, x_t, label=r"$x(t)$")
axs[0].set_title("function $x(t)$")
axs[0].set_xlim(-1, 1)
axs[0].set_ylim(-2, 2)
axs[0].legend()

# نمودار 2: p(t)
axs[1].stem(t, p_t, linefmt='C1-', markerfmt='C1o', basefmt='C1-', label=r"$p(t)$")
axs[1].set_title("Impulse-Train Sampling $p(t)$")
axs[1].set_xlim(-1, 1)
axs[1].legend()

# نمودار 3: x_p(t) = x(t) * p(t)
axs[2].stem(t, x_p_t, linefmt='C2-', markerfmt='C2o', basefmt='C2-', label=r"$x_p(t)$")
axs[2].plot(t, x_t, 'k--', alpha=0.5, label=r"$x(t)$ (source)")
axs[2].set_title(r"Sampling $x_p(t) = x(t) \cdot p(t)$")
axs[2].set_xlim(-1, 1)
axs[2].set_ylim(-2, 2)
axs[2].legend()

# نمودار 4: نمایش تصویر به جای بلوک دیاگرام
axs[3].axis('off')

# بارگذاری تصویر و نمایش آن
img = mpimg.imread('diagram0.png')  # اطمینان حاصل کنید که نام فایل درست است و در مسیر قرار دارد
axs[3].imshow(img)
axs[3].set_title('System Diagram')

plt.subplots_adjust(hspace=0.5, wspace=0.5)

# نمایش نمودار
plt.show()
```

چون کانولوشن با ضرب تنها باعث جابجایی سیگنال می‌شود، یعنی:

$$X(j\omega) * \delta(\omega - \omega_0) = X(j(\omega - \omega_0))$$

نتیجه می‌گیریم:

$$X_p(j\omega) = \frac{1}{T} \sum_{k=-\infty}^{+\infty} X(j(\omega - k\omega_s))$$

یعنی مطابق شکل ۳-۷، $X_p(j\omega)$ تابع متناوبی از $\omega$ است از مجموع $X(j\omega)$ های انتقال یافته است که با ضریب $\frac{1}{T}$ 
تغییر مقیاس یافته‌اند. در شکل ۷-۳ (ج) $\omega_s > 2\omega_m$ ( $\omega_s = \frac{2\pi}{T}$ )، و بنابراین $X(j\omega)$ های انتقال یافته 
همپوشانی ندارند. ولی در شکل ۷-۲ (با $\omega_s < 2\omega_m$)، $\omega_s$ همپوشانی دارد. 

شکل ۷-۳ (ج): $X_p(j\omega)$ در ضریب صحیح فرکانسی نمونه‌برداری بدون تغییر تکرار شده است. پس اگر $\omega_s > 2 \omega_m$ می‌توان $X(j\omega)$ را 
از $X_p(j\omega)$ به کمک یک فیلتر پایین‌گذر با بهره ۱ و فرکانس قطع بزرگتر از $\omega_m$ بازیابی کرد.

### قضیه نمونه‌برداری:
فرض کنید:
1. سیگنالی با باند محدود است، یعنی در $\lvert \omega \rvert > \omega_M$، $X(j\omega) = 0$.
2. به صورت یکتا به وسیله نمونه‌های $x(nT)$ تعیین می‌شود، اگر $n = 0, \pm 1, \pm 2, ...$ و $\omega_s > 2\omega_M$

که در آن:
$$\omega_s = \frac{2\pi}{T}$$

با داشتن این نمونه‌ها می‌توانیم با ایجاد یک قطار ضربه متناوب که اندازه‌هایشان با مقادیر متوالی نمونه‌ها برابر باشد، $x(t)$ را بازسازی کنیم. این قطار ضربه باید از یک فیلتر پایین‌گذر با بهره $T$ و فرکانس قطع بزرگ‌تر از $\omega_M$ و کمتر از $(\omega_s - \omega_M)$ گذرانده شود. سیگنال خروجی این فیلتر دقیقاً برابر $x(t)$ است.

فرکانس $\omega_M$ طبق قضیه نمونه‌برداری فرکانس نمونه‌برداری باید از آن بزرگ‌تر باشد، آهنگ نایکوئیست خوانده می‌شود.

همان‌طور که در فصل ۴ گفتیم، در عمل به دلایل مختلف از فیلتر ایده‌آل استفاده نمی‌شود. در بسیاری از کاربردها به جای فیلتر ایده‌آل شکل ۴-۷، فیلتر غیرایده‌آل $H(j\omega)$ به کار می‌رود که تقریبی از مشخصات فرکانسی دلخواه را دارد (یعنی در $| \omega | < \omega_M$, $H(j\omega) \approx 1$ و $| \omega | > (\omega_s - \omega_M)$, $H(j\omega) \approx 0$). واضح است که هر تغییری در طبقه فیلتر پایین‌گذر به تفاوت $x_r(t)$ و $x(t)$ می‌انجامد، که از تفاوت $X_r(j\omega)$ و $X(j\omega)$ ناشی می‌شود. پس فیلتر ایده‌آل باید با توجه به اعوجاج قابل قبول در کاربرد مورد نظر برگزیده شود.

در این فصل و فصل بعد برای سادگی و تأکید بر اصول پایه قضیه نمونه‌برداری، فرض می‌کنیم فیلتر ایده‌آل موجود است و آن را به کار می‌بریم. ولی همراه داریم که در عمل باید به جای چنین فیلتری یک فیلتر غیرایده‌آل به کار رود که تفاوت قابل تحمل مشخصاتش با فیلتر ایده‌آل توسط مسئله مورد نظر تعیین می‌شود.

### ۲-۱-۷: نمونه‌برداری با نگهدار مرتبه صفر

قضیه نمونه‌برداری که ساده‌ترین بیان آن بر حسب نمونه‌برداری با قطار ضربه است، می‌گوید که سیگنال دارای باند محدود به‌صورتی یکتا با نمونه‌های مشخص می‌شود. در عمل تولید و انتقال پالس‌های باریک با دامنه زیاد که تقریبی از ضربه باشند، نسبتاً مشکل است و غالباً ساده‌تر این است که سیگنال نمونه‌برداری شده را به شکلی که نگهدار مرتبه صفر نامیده می‌شود تولید کرد. چنین سیستمی به صورت شکل ۷-۵ از $x(t)$ در یک لحظه نمونه‌برداری می‌کند و مقدار آن را تا لحظه نمونه‌برداری بعدی نگه می‌دارد. باز هم با فیلتر پایین‌گذر می‌توان $x(t)$ را از خروجی نگهدار مرتبه صفر بازیابی کرد. اما در این مورد فیلتر دیگر نباید در باند عبور بهره ثابت داشته باشد. برای یافتن مشخصه فیلتر لازم، ابتدا توجه کنید که مطابق شکل ۷-۶ با گذاشتن یک سیستم LTI با پاسخ ضربه مستطیلی به دنبال سیستم نمونه‌برداری می‌توان $x_0(t)$ خروجی نگهدار مرتبه صفر را تولید کرد.

برای بازسازی $x(t)$ از $x_0(t)$، $x_0(t)$ را با سیستم LTI دارای پاسخ ضربه $h_1(t)$ و پاسخ فرکانسی $H_1(j\omega)$ پردازش می‌کنیم. شکل ۷-۷ اتصال سری این سیستم با سیستم شکل ۷-۶ را نشان می‌دهد. می‌خواهیم $H_1(j\omega)$ را به نحوی تعیین کنیم که داشته باشیم $r(t) = x(t)$. مقایسه سیستم شکل ۷-۷ و سیستم شکل ۴-۷ نشان می‌دهد که اگر ترکیب سری $H_0(j\omega)$ و $H_1(j\omega)$ فیلتر ایده‌آل پایین‌گذر شکل ۴-۷ باشد، آنگاه $r(t) = x(t)$.

با توجه به مثال ۴-۴ و خاصیت جابجایی زمانی بخش ۲-۴-۲ داریم:

$$H_0(j\omega) = e^{-j\omega T/2} \left[\frac{2\sin(\omega T / 2)}{\omega}\right] \tag{۷-۷}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف پارامترها
T = 1  # دوره زمانی (می‌توانید مقدار دیگری بدهید)
omega = np.linspace(-10, 10, 1000)  # محدوده فرکانس ω

# محاسبه H0(jω)
H0_magnitude = (2 * np.sin(omega * T / 2)) / omega  # دامنه تابع (مقدار مطلق)
H0_magnitude[np.isnan(H0_magnitude)] = T  # برای ω = 0 مقدار را T قرار می‌دهیم (به دلیل صفر شدن مخرج)

# رسم نمودار
plt.figure(figsize=(10, 6))
plt.plot(omega, H0_magnitude, label=r"$|H_0(j\omega)|$", color='blue')

plt.title("Magnitude of $H_0(j\omega)$")
plt.xlabel(r"$\omega$")
plt.ylabel(r"$|H_0(j\omega)|$")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

### ۲-۷: بازسازی سیگنال از نمونه‌ها با درون‌یابی

درون‌یابی، یعنی منطبق کردن یک سیگنال پیوسته به مجموعه‌ای از نمونه‌ها، روش متداولی برای بازسازی یک تابع تقریبی یا دقیق است. یک روش ساده درون‌یابی، استفاده از نگهدار مرتبه صفر مورد بحث در بخش ۷-۱ است. شکل ساده و مفید دیگری از درون‌یابی، درون‌یابی خطی است که در آن نقاط دو نمونه مجاور مطابق شکل ۷-۹ توسط یک خط راست به هم وصل می‌شوند. در فرمول‌های درون‌یابی پیچیده‌تر، نقاط متوالی نمونه‌ها با چند جمله‌ای‌های درجه بالاتر یا توابع دیگر به هم وصل می‌شوند.

همان‌طور که در بخش ۷-۱ دیدیم، اگر لحظات نمونه‌برداری یک سیگنال دارای باند محدود به اندازه کافی نزدیک باشند، سیگنال را می‌توان به صورت دقیق بازسازی کرد، یعنی با استفاده از فیلتر پایین‌گذر می‌توان بین نقاط نمونه درون‌یابی دقیقی صورت داد. با در نظر گرفتن اثر فیلتر پایین‌گذر شکل ۷-۷ در حوزه زمان، تعبیر بازسازی سیگنال $x(t)$ به عنوان یک فرایند درون‌یابی آشکارتر می‌شود. خروجی عبارت است از:

$$
x_r(t) = x_p(t) * h(t)
$$

یعنی با $x_p(t)$ معادله )۷-۳( داریم:

$$
x_r(t) = \sum_{n=-\infty}^{+\infty} x(nT)h(t - nT) \tag{۹-۷}
$$
معادله )۷-۹( نشان می‌دهد که چگونه یک منحنی پیوسته بین نقاط نمونه منطبق می‌شود. بنابراین یک فرمول درون‌یابی را بیان می‌کند. برای فیلتر پایین‌گذر ایده‌آل شکل ۷-۴:

$$
h(t) = \frac{\omega_c T \sin(\omega_c t)}{\pi \omega_c t} \tag{۷-۱۰}
$$

بنابراین:

$$
x_r(t) = \sum_{n=-\infty}^{+\infty} x(nT) \frac{\omega_c T \sin(\omega_c (t - nT))}{\pi \omega_c (t - nT)} \tag{۷-۱۱}
$$

شکل ۷-۱۰ بازسازی بر اساس معادله )۷-۱۱( را به ازای $\omega_c = \pi/T$ نشان می‌دهد. شکل ۷-۱۰ )الف( سیگنال اصلی محدود باند $x(t)$ را نشان می‌دهد، و شکل ۷-۱۰ )ب( $x_p(t)$ یعنی قطار ضربه‌های نمونه است. شکل ۷-۱۰ )ج( بر هم نهش جملات معادله )۷-۱۱( را نشان می‌دهد.

درون‌یابی به کمک پاسخ ضربه فیلتر پایین‌گذر ایده‌آل مطابق معادله )۷-۱۱( معمولاً درون‌یابی محدود-باند نامیده می‌شود. زیرا اگر $x(t)$ محدود به باند باشد و فرکانس نمونه‌برداری شرط قضیه نمونه‌برداری را ارضا کند، بازسازی دقیق $x(t)$ منتج می‌شود. چنانچه گفتیم در بسیاری از موارد استفاده از فیلتری با دقت کمتر ولی ساده‌تر، نتایج خوبی به همراه دارد.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 1  # Sampling period
wc = np.pi / T  # Cut-off frequency
n_samples = 10  # Number of samples

# Sampled signal x[nT], defined as a vectorized function
def x(nT):
    return np.sin(2 * np.pi * nT)  # Example: a sinusoidal sampled signal

# Reconstruction function using sinc interpolation
def sinc_interpolation(t, nT_samples, wc, T):
    xr = np.zeros_like(t)
    for nT in nT_samples:
        xr += x(nT) * (wc * T / np.pi) * np.sin(wc * (t - nT)) / (wc * (t - nT) + 1e-12)
    return xr

# Time range
t = np.linspace(-2 * T, 2 * T, 1000)  # Time range for reconstruction
nT_samples = np.arange(-n_samples, n_samples + 1) * T  # Discrete sample points nT

# Reconstructed signal
xr = sinc_interpolation(t, nT_samples, wc, T)

# Plot the reconstructed signal
plt.figure(figsize=(10, 6))
plt.plot(t, xr, label=r"$x_r(t)$ (Reconstructed)", color='blue')

# Plot the sampled points using stem
nT_sample_values = x(nT_samples)  # Values of the sampled signal
plt.stem(nT_samples, nT_sample_values, linefmt='r-', markerfmt='ro', basefmt='k', label=r"$x[nT]$ (Samples)")

# Plot settings
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Signal Reconstruction from Samples Using Sinc Interpolation")
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```
