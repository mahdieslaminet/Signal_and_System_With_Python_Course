**مقدمه 9.0**

در فصل‌های پیشین، دیدیم که ابزارهای تحلیل فوریه در مطالعه بسیاری از مسائل عملی که شامل سیگنال‌ها و سیستم‌های LTI (خطی و زمان‌Invariant) هستند، بسیار مفید می‌باشند. این امر به طور عمده به این دلیل است که کلاس‌های وسیعی از سیگنال‌ها می‌توانند به صورت ترکیب‌های خطی از نماهای پیچیده دوره‌ای نمایش داده شوند و نماهای پیچیده ویژگی‌های ویژه سیستم‌های LTI هستند. تبدیل فوریه زمان پیوسته به ما این امکان را می‌دهد که سیگنال‌ها را به صورت ترکیب‌های خطی از نماهای پیچیده‌ای به شکل \( e^{st} \) با \( s = jw \) نمایش دهیم. با این حال، ویژگی ویژه‌ای که در بخش 3.2 معرفی شد و بسیاری از پیامدهای آن، به طور کلی برای مقادیر دلخواه \( s \) نیز اعمال می‌شود و نه فقط مقادیر موهومی خالص. این مشاهده منجر به تعمیم تبدیل فوریه زمان پیوسته می‌شود که به نام تبدیل لاپلاس شناخته می‌شود و در این فصل توسعه می‌دهیم. در فصل بعد، تعمیم متناظر زمانی گسسته آن که به نام تبدیل z شناخته می‌شود را توسعه خواهیم داد.

همانطور که خواهیم دید، تبدیل‌های لاپلاس و z بسیاری از ویژگی‌هایی را دارند که تحلیل فوریه را به ابزاری بسیار مفید تبدیل کرده است. علاوه بر این، این تبدیل‌ها نه تنها ابزارهای اضافی و بینش‌هایی برای سیگنال‌ها و سیستم‌هایی که می‌توانند با استفاده از تبدیل فوریه تحلیل شوند، فراهم می‌آورند، بلکه می‌توانند در برخی زمینه‌های بسیار مهم که تبدیل فوریه نمی‌تواند به کار گرفته شود، نیز کاربرد داشته باشند. به عنوان مثال، تبدیل‌های لاپلاس و z می‌توانند برای تحلیل بسیاری از سیستم‌های ناپایدار به کار روند و بنابراین نقش مهمی در بررسی پایداری یا ناپایداری سیستم‌ها ایفا می‌کنند. این واقعیت، به همراه ویژگی‌های جبری که تبدیل‌های لاپلاس و z با تبدیل فوریه مشترک دارند، مجموعه‌ای از ابزارهای بسیار مهم برای تحلیل سیستم‌ها و به ویژه برای تحلیل سیستم‌های بازخورد که در فصل 11 توسعه خواهیم داد، به وجود می‌آورد.

**بخش 9.1: تبدیل لاپلاس**

در فصل 3، دیدیم که پاسخ یک سیستم LTI با پاسخ ضربه \( h(t) \) به ورودی نمایی پیچیده به شکل \( e^{st} \) به صورت زیر است:

$$
y(t) = H(s) e^{st}, \quad \text{معادله} \, (9.1)
$$

که در آن:

$$
H(s) = \int_0^\infty h(t) e^{-st} dt, \quad \text{معادله} \, (9.2)
$$

برای \( s \) موهومی (یعنی \( s = jw \))، انتگرال در معادله (9.2) به تبدیل فوریه \( h(t) \) مربوط می‌شود. برای مقادیر عمومی متغیرهای پیچیده، این تبدیل به نام تبدیل لاپلاس پاسخ ضربه \( h(t) \) شناخته می‌شود.

تبدیل لاپلاس یک سیگنال عمومی \( x(t) \) به صورت زیر تعریف می‌شود:

$$
X(s) = \int_{-\infty}^{\infty} x(t) e^{-st} dt, \quad \text{معادله} \, (9.3)
$$

و توجه داشته باشید که به طور خاص این تبدیل تابعی از متغیر مستقل \( s \) است که مربوط به متغیر پیچیده در نمای \( e^{-st} \) است. متغیرهای پیچیده را می‌توان به صورت \( s = \sigma + jw \) نوشت، به طوری که \( \sigma \) و \( w \) به ترتیب بخش‌های حقیقی و موهومی آن هستند. برای راحتی، گاهی اوقات تبدیل لاپلاس را به صورت عملگری \( \mathcal{C} \{x(t)\} \) نمایش می‌دهیم و رابطه تبدیل بین \( x(t) \) و \( X(s) \) را به صورت زیر می‌نویسیم:

$$
\mathcal{C} \{ x(t) \} = X(s), \quad \text{معادله} \, (9.4)
$$

وقتی \( s = jw \)، معادله (9.3) به صورت زیر در می‌آید:

$$
X(jw) = \int_{-\infty}^{\infty} x(t) e^{-jwt} dt, \quad \text{معادله} \, (9.5)
$$

که معادل تبدیل فوریه \( x(t) \) است؛ یعنی:

$$
X(s)|_{s=jw} = \mathcal{F} \{x(t)\}, \quad \text{معادله} \, (9.6)
$$

تبدیل لاپلاس همچنین رابطه ساده‌ای با تبدیل

یا

$$
X(\sigma + jw) = \int_{-\infty}^{\infty} x(t) e^{-(\sigma + jw)t} dt, \quad \text{معادله} \, (9.8)
$$

ما سمت راست معادله (9.8) را به عنوان تبدیل فوریه \( x(t) e^{-\sigma t} \) می‌شناسیم؛ به عبارت دیگر، تبدیل لاپلاس \( x(t) \) می‌تواند به عنوان تبدیل فوریه \( x(t) \) پس از ضرب در یک سیگنال نمایی حقیقی تفسیر شود. سیگنال نمایی حقیقی \( e^{-\sigma t} \) ممکن است در زمان کاهش یابد یا رشد کند، بسته به اینکه \( \sigma \) مثبت یا منفی باشد.

برای توضیح تبدیل لاپلاس و رابطه آن با تبدیل فوریه، بیایید مثال زیر را در نظر بگیریم:

**مثال 9.1**  
سیگنال \( x(t) = e^{-\alpha t} u(t) \) را در نظر بگیرید. از مثال 4.1، تبدیل فوریه \( X(jw) \) برای \( \alpha > 0 \) همگرا است و به صورت زیر داده می‌شود:

$$
X(j\omega) = \int_{-\infty}^{\infty} e^{-at} u(t) e^{-j\omega t} dt = \int_0^\infty e^{-at} e^{-j\omega t} dt = \frac{1}{j\omega + a}, \quad a > 0.
$$

از معادله (9.3)، تبدیل لاپلاس به صورت زیر است:

$$
X(s) = \int_{-\infty}^{\infty} e^{-at} u(t) e^{-st} dt = \int_0^\infty e^{-(s + a)t} dt,
$$
یا با قرار دادن \( s = \sigma + j\omega \):

$$
X(\sigma + j\omega) = \int_0^\infty e^{-(\sigma + a)t} e^{-j\omega t} dt. \quad \text{معادله} \, (9.11)
$$

با مقایسه با معادله (9.9)، معادله (9.11) را به عنوان تبدیل فوریه \( e^{-(\sigma + a)t} u(t) \) می‌شناسیم و بنابراین:

$$
X(\sigma + j\omega) = \frac{1}{(\sigma + a) + j\omega}, \quad \sigma + a > 0.
$$

یا به طور معادل، چون \( s = \sigma + j\omega \) و \( \sigma = \text{Re}(s) \):

$$
X(s) = \frac{1}{s + a}, \quad \text{Re}(s) > -a.
$$

برای مثال، برای \( a = 0 \)، \( x(t) \) همان گام واحد است که تبدیل لاپلاس آن به صورت \( X(s) = \frac{1}{s} \) است.

توجه داشته باشید که همانطور که تبدیل فوریه برای همه سیگنال‌ها همگرا نمی‌شود، تبدیل لاپلاس نیز ممکن است برای مقادیری از \( \text{Re}(s) \) همگرا باشد و برای مقادیر دیگر همگرا نباشد. در معادله (9.13)، تبدیل لاپلاس تنها برای \( \text{Re}(s) > -a \) همگرا است. اگر \( a \) مثبت باشد،

سپس می‌توان تبدیل لاپلاس \( X(s) \) را در \( \sigma = 0 \) محاسبه کرد تا به نتیجه زیر برسیم:

$$
X(0+j\omega) = \frac{1}{\omega + a}, \quad \text{معادله} \, (9.15)
$$

همانطور که در معادله (9.6) نشان داده شده است، برای \( \sigma = 0 \)، تبدیل لاپلاس برابر با تبدیل فوریه است، همانطور که در مثال قبلی با مقایسه معادلات (9.9) و (9.15) مشهود است. اگر \( a \) منفی یا صفر باشد، تبدیل لاپلاس همچنان وجود دارد، اما تبدیل فوریه وجود ندارد.

### مثال 9.2

برای مقایسه با مثال 9.1، به عنوان مثال دوم سیگنال زیر را در نظر می‌گیریم:

$$
x(t) = -e^{-at} u(-t)
$$

بنابراین:

$$
X(s) = - \int_{-\infty}^{\infty} e^{-at} e^{-st} u(-t) dt = - \int_{-\infty}^{\infty} e^{-(s + a)t} dt,
$$

که معادله (9.16) است. با توجه به معادله (9.18):

$$
X(s) = \frac{-1}{s + a}, \quad \text{معادله} \, (9.17)
$$

برای همگرایی در این مثال، نیاز داریم که \( \text{Re}(s + a) < 0 \) یا \( \text{Re}(s) < -a \)، یعنی:

$$
X(s) = \frac{-1}{s + a}, \quad \text{Re}(s) < -a, \quad \text{معادله} \, (9.19)
$$

با مقایسه معادلات (9.14) و (9.19)، مشاهده می‌کنیم که عبارت جبری برای تبدیل لاپلاس در هر دو سیگنال در مثال‌های 9.1 و 9.2 یکسان است. با این حال، از همان معادلات می‌بینیم که مجموعه مقادیر \( s \) که این عبارت برای آن‌ها معتبر است، در دو مثال بسیار متفاوت است. این موضوع نشان می‌دهد که هنگام مشخص کردن تبدیل لاپلاس یک سیگنال، هم عبارت جبری و هم محدوده مقادیر \( s \) برای که این عبارت معتبر است، نیاز است. به طور کلی، محدوده مقادیر \( s \) برای که انتگرال در معادله (9.3) همگرا است، به عنوان منطقه همگرایی (ROC) تبدیل لاپلاس شناخته می‌شود. به عبارت دیگر، ROC شامل مقادیر \( s = \sigma + j\omega \) است که تبدیل فوریه سیگنال \( x(t) e^{-\sigma t} \) همگرا می‌شود.

### نمایش منطقه همگرایی

یک روش راحت برای نمایش منطقه همگرایی در شکل 9.1 نشان داده شده است. متغیر \( s \) یک عدد مختلط است و در این شکل، صفحه مختلط یا همان صفحه \( s \)-plane نشان داده شده است. محور افقی \( \text{Re}(s) \) و محور عمودی \( \text{Im}(s) \) است. این محورها گاهی به عنوان محور \( \sigma \) و محور \( j\omega \) شناخته می‌شوند. ناحیه سایه‌دار در شکل 9.1 (الف) مجموعه نقاط در صفحه \( s \)-plane را نشان می‌دهد که به منطقه همگرایی برای مثال 9.1 مربوط می‌شود. ناحیه سایه‌دار در شکل 9.1 (ب) منطقه همگرایی برای مثال 9.2 را نشان می‌دهد.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure and axes for the two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Define the parameters for the plots
a = -1
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)

# Plot (a)
axes[0].axvline(x=a, color='black', linestyle='--')  # Dashed vertical line at -a
axes[0].axhline(y=0, color='black', linestyle='-')   # Real axis
axes[0].axvline(x=0, color='black', linestyle='-')   # Imaginary axis
axes[0].set_xlim(-2, 2)
axes[0].set_ylim(-2, 2)
axes[0].set_title("(a)")
axes[0].set_xlabel("Re")
axes[0].set_ylabel("Im")
axes[0].grid(False)

# Plot (b)
axes[1].axvline(x=a, color='black', linestyle='--')  # Dashed vertical line at -a
axes[1].axhline(y=0, color='black', linestyle='-')   # Real axis
axes[1].axvline(x=0, color='black', linestyle='-')   # Imaginary axis
axes[1].set_xlim(-2, 2)
axes[1].set_ylim(-2, 2)
axes[1].set_title("(b)")
axes[1].set_xlabel("Re")
axes[1].set_ylabel("Im")
axes[1].grid(False)

# Set the overall layout and show the plots
plt.tight_layout()
plt.show()
```

## مثال 9.4

در این مثال، سیگنالی را در نظر می‌گیریم که مجموع یک نمایی واقعی و پیچیده است:

$$
x(t) = e^{-2t} u(t) + e^{-t} (\cos 3t) u(t) \tag{9.24}
$$

با استفاده از رابطه اویلر می‌توانیم بنویسیم:

$$
x(t) = \left[ e^{-2t} + \frac{1}{2} e^{-(1-3j)t} + \frac{1}{2} e^{-(1+3j)t} \right] u(t)
$$

تبدیل لاپلاس \( x(t) \) به صورت زیر بیان می‌شود:

$$
X(s) = \int_{0}^{\infty} e^{-2t} u(t) e^{-st} \, dt + \frac{1}{2} \int_{0}^{\infty} e^{-(1-3j)t} u(t) e^{-st} \, dt + \frac{1}{2} \int_{0}^{\infty} e^{-(1+3j)t} u(t) e^{-st} \, dt
$$

هر یک از انتگرال‌ها در معادله (9.25) تبدیل لاپلاس از نوع موجود در مثال 9.1 هستند. بنابراین داریم:

$$
e^{-2t} u(t) \leftrightarrow \frac{1}{s + 2}, \quad \Re(s) > -2
$$

$$
e^{-(1-3j)t} u(t) \leftrightarrow \frac{1}{s + (1 - 3j)}, \quad \Re(s) > -1
$$

$$
e^{-(1+3j)t} u(t) \leftrightarrow \frac{1}{s + (1 + 3j)}, \quad \Re(s) > -1
$$

برای اینکه هر سه تبدیل لاپلاس همزمان همگرا شوند، باید \( \Re(s) > -1 \) باشد. بنابراین، تبدیل لاپلاس \( x(t) \) به صورت زیر است:

$$
\frac{1}{s + 2} + \frac{1}{2} \left( \frac{1}{s + (1 - 3j)} \right) + \frac{1}{2} \left( \frac{1}{s + (1 + 3j)} \right), \quad \Re(s) > -1
$$

یا با ترکیب اصطلاحات روی یک مخرج مشترک:

$$
e^{-2t} u(t) + e^{-t} (\cos 3t) u(t) \leftrightarrow \frac{2s^2 + 5s + 12}{(s^2 + 2s + 10)(s + 2)}, \quad \Re(s) > -1
$$

در هر یک از این مثال‌ها، تبدیل لاپلاس منطقی است، به طوری که:

$$
X(s) = \frac{N(s)}{D(s)}
$$

که در آن \( N(s) \) و \( D(s) \) چند جمله‌ای‌های صورت و مخرج هستند.

## مثال 9.5

در این مثال، سیگنالی را در نظر می‌گیریم که ترکیبی از توابع گام واحد است:

$$
x(t) = \delta(t) - u(t) + u(t) \tag{9.32}
$$

تبدیل لاپلاس جمله دوم و سوم در سمت راست معادله (9.32) را می‌توان از مثال 9.1 ارزیابی کرد. تبدیل لاپلاس تکانه واحد را می‌توان مستقیماً به صورت زیر ارزیابی کرد:

$$
\mathcal{L}\left\{ \delta(t) \right\} = \int_{-\infty}^{\infty} \delta(t)e^{-st} dt = 1 \tag{9.33}
$$

که برای هر مقدار \( s \) معتبر است. به عبارت دیگر، منطقه همگرایی (ROC) تبدیل لاپلاس \( \mathcal{L}\left\{ \delta(t) \right\} \) کل صفحه \( s \) است. با استفاده از این نتیجه، به همراه تبدیل لاپلاس دو عبارت دیگر در معادله (9.32)، به دست می‌آوریم:

$$
X(s) = \frac{1}{3s + 1} + \frac{1}{3s - 2}, \quad \Re{s} > 2 \tag{9.34}
$$

یا معادله معادل:

$$
X(s) = \frac{(s - 1)^2}{(s + 1)(s^2)}, \quad \Re{s} > 2 \tag{9.35}
$$

که در آن ROC مجموعه‌ای از مقادیر \( s \) است که لاپلاس هر سه عبارت را در \( x(t) \) تبدیل می‌کند. نمودار قطب‌ها و صفرها برای این مثال همراه با ROC در شکل 9.3 نشان داده شده است. همچنین، از آنجایی که درجات صورت و مخرج \( X(s) \) برابر است، \( X(s) \) در بی‌نهایت نه قطب دارد و نه صفر.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
fig, ax = plt.subplots()

# Plot the poles and zeros
poles = [-1, 2]
zeros = [1]

# Plot poles
for pole in poles:
    ax.plot(np.real(pole), np.imag(pole), 'rx', markersize=10, label='Pole')

# Plot zeros
for zero in zeros:
    ax.plot(np.real(zero), np.imag(zero), 'bo', markersize=10, label='Zero')

# Set axis labels
ax.set_xlabel('Re')
ax.set_ylabel('Im')

# Set axis limits
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])

# Add grid
ax.grid(True)

# Add legend
ax.legend()

# Highlight the right half-plane
ax.axvspan(0, 3, color='yellow', alpha=0.3)

# Show the plot
plt.show()
```

    همچنین در این مثال، توجه کنید که دو صفر در معادله (9.35) در همان مقدار <i>s</i> رخ می‌دهد. به طور کلی، ما به ترتیب یک قطب یا صفر را به عنوان تعداد دفعات تکرار آن در یک مکان معین اشاره خواهیم کرد. در مثال 9.5 یک صفر مرتبه دوم در <i>s = 1</i> و دو قطب مرتبه اول، یکی در <i>s = 1</i> و دیگری در <i>s = 2</i> وجود دارد. در این مثال، منطقه همگرایی (ROC) در سمت راست سمت راست‌ترین قطب قرار دارد.
  </p>
  <p>
    به طور کلی، برای تبدیل‌های لاپلاس منطقی، یک رابطه نزدیک بین مکان‌های قطب‌ها و ROC‌های احتمالی وجود دارد که می‌توانند با نمودار قطب-صفر معین مرتبط شوند. محدودیت‌های خاص در ROC ارتباط نزدیکی با ویژگی‌های حوزه زمانی <i>x(t)</i> دارند. در بخش بعدی به بررسی برخی از این محدودیت‌ها و ویژگی‌ها می‌پردازیم.

\section*{ترجمه}

برای تبدیل لاپلاس \( x(t) \)، منطقه همگرایی (ROC) شامل مقادیر \( s \) است که در آن‌ها \( x(t) e^{-t} \) به‌طور کامل قابل انتگرال‌گیری است:

\[
\int_{T_1}^{\infty} x(t) e^{-t} \, dt < \infty
\]

\textbf{ویژگی 1}: این ویژگی به‌طور طبیعی پیروی می‌کند زیرا این شرط تنها به بخش حقیقی \( s \) بستگی دارد.

\textbf{ویژگی 2}: برای تبدیل‌های لاپلاس عقلانی، منطقه همگرایی هیچ قطبی ندارد.

ویژگی 2 به‌راحتی در تمام مثال‌های بررسی‌شده تاکنون مشاهده می‌شود. از آنجایی که \( X(s) \) در یک قطب بی‌نهایت است، انتگرال در معادله (9.3) به‌وضوح در یک قطب همگرا نمی‌شود، بنابراین منطقه همگرایی نمی‌تواند مقادیر \( s \) که قطب هستند را شامل شود.

\textbf{ویژگی 3}: اگر \( x(t) \) دارای مدت زمان محدود باشد و به‌طور کامل قابل انتگرال‌گیری باشد، منطقه همگرایی کل صفحه \( s \) است.

شهود پشت این نتیجه در شکل‌های 9.4 و 9.5 نشان داده شده است. به‌طور خاص، یک سیگنال با مدت زمان محدود این ویژگی را دارد که در خارج از یک بازه محدود صفر است، همانطور که در شکل 9.4 نشان داده شده است. در شکل 9.5(a)، سیگنال \( x(t) \) از شکل 9.4 ضرب‌شده با یک نمایی در حال زوال نشان داده شده است و در شکل 9.5(b) همان سیگنال ضرب‌شده با یک نمایی رو به رشد نشان داده شده است.



```python
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Top graph: Finite-duration signal
T1, T2 = 1, 5
t = np.linspace(0, 6, 1000)
signal = np.piecewise(t, [t < T1, (t >= T1) & (t <= T2), t > T2], [0, lambda t: np.sin(2 * np.pi * t), 0])
axs[0].plot(t, signal)
axs[0].set_title('Figure 9.4 Finite-duration signal')
axs[0].set_xlabel('t')
axs[0].set_ylabel('Amplitude')
axs[0].grid(True)

# Bottom left graph: Decaying exponential
T1, T2 = 1, 5
t = np.linspace(0, 6, 1000)
decaying_exponential = np.piecewise(t, [t < T1, t >= T1], [0, lambda t: np.exp(-t)])
axs[1].plot(t, decaying_exponential)
axs[1].set_title('Decaying exponential')
axs[1].set_xlabel('t')
axs[1].set_ylabel('Amplitude')
axs[1].grid(True)

# Bottom right graph: Growing exponential
T1, T2 = 1, 5
t = np.linspace(0, 6, 1000)
growing_exponential = np.piecewise(t, [t < T1, t >= T1], [0, lambda t: np.exp(t)])
axs[2].plot(t, growing_exponential)
axs[2].set_title('Growing exponential')
axs[2].set_xlabel('t')
axs[2].set_ylabel('Amplitude')
axs[2].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

            x(t) =
            \begin{cases}
            e^{\sigma t}, & 0 < t < T \\
            0, & \text{otherwise}

        <p>
            $$
            X(s) = \int_{0}^{T} e^{\sigma t} e^{-st} \, dt = \int_{0}^{T} e^{-(s+\sigma)t} \, dt
            $$

        <p>از آنجایی که در این مثال \( x(t) \) دارای مدت زمان محدود است، از ویژگی 3 نتیجه می‌گیریم که منطقه همگرایی کل صفحه \( s \) است. در فرم معادله (9.42)، \( X(s) \) به نظر می‌رسد که در \( s = -\sigma \) یک قطب دارد، که از ویژگی 2، این امر با منطقه همگرایی که شامل تمام صفحه \( s \) است ناسازگار است. اما در واقع، در عبارت جبری معادله (9.42)، هم صورت و هم مخرج در \( s = -\sigma \) صفر هستند و بنابراین برای تعیین \( X(s) \) در \( s = -\sigma \)، می‌توانیم از قانون لوپیتال استفاده کنیم تا به نتیجه زیر برسیم:</p>

        <p>
            $$
            \lim_{s \to -\sigma} X(s) = \lim_{s \to -\sigma} \frac{d}{ds} \left( 1 - e^{-(s+\sigma)T} \right) \Big/ (s + \sigma)
            $$
        </p>

        <p>که در نهایت خواهیم داشت:</p>

        <p>
            $$
            X(-\sigma) = T
            $$
        </p>

\textbf{ویژگی 4}: اگر \( x(t) \) راست‌جهت باشد و اگر خط \( \text{Re}(s) = \sigma_0 \) در منطقه همگرایی (ROC) باشد، آنگاه تمام مقادیر \( s \) که \( \text{Re}(s) > \sigma_0 \) هستند، نیز در منطقه همگرایی خواهند بود.

یک سیگنال راست‌جهت سیگنالی است که برای آن \( x(t) = 0 \) قبل از زمان محدود \( T_1 \) است، همانطور که در شکل 9.7 نشان داده شده است. همچنین در این حالت، اگر یک نقطه \( s \) در منطقه همگرایی باشد، تمام نقاط به سمت راست آن، یعنی تمام نقاط با بخش حقیقی بزرگتر، نیز در منطقه همگرایی خواهند بود. به همین دلیل، منطقه همگرایی در این مورد معمولاً به‌عنوان نیم‌صفحه راست شناخته می‌شود.

\textbf{ویژگی 5}: اگر \( x(t) \) چپ‌جهت باشد و اگر خط \( \text{Re}(s) = \sigma_0 \) در منطقه همگرایی باشد، آنگاه تمام مقادیر \( s \) که \( \text{Re}(s) < \sigma_0 \) هستند، نیز در منطقه همگرایی خواهند بود.

یک سیگنال چپ‌جهت سیگنالی است که برای آن \( x(t) = 0 \) پس از زمان محدود \( T_2 \) است، همانطور که در شکل 9.8 نشان داده شده است. استدلال و شهود پشت این ویژگی دقیقاً مشابه استدلال و شهود ویژگی 4 است. همچنین برای یک سیگنال چپ‌جهت، منطقه همگرایی معمولاً به‌عنوان نیم‌صفحه چپ شناخته می‌شود، به‌طوری که اگر یک نقطه \( s \) در منطقه همگرایی باشد، تمام نقاط به سمت چپ آن نیز در منطقه همگرایی خواهند بود.

\textbf{ویژگی 6}: اگر \( x(t) \) دوطرفه باشد و اگر خط \( \text{Re}(s) = \sigma_0 \) در منطقه همگرایی باشد، آنگاه منطقه همگرایی به‌صورت یک نوار در صفحه \( s \) خواهد بود که شامل خط \( \text{Re}(s) = \sigma_0 \) می‌شود.

یک سیگنال دوطرفه سیگنالی است که برای آن، دامنه سیگنال در هر دو جهت \( t > 0 \) و \( t < 0 \) به‌طور نامتناهی ادامه دارد، همانطور که در شکل 9.9(a) نشان داده شده است. برای چنین سیگنالی، می‌توانیم با انتخاب یک زمان دلخواه \( T_0 \) و تقسیم \( x(t) \) به مجموع یک سیگنال راست‌جهت \( X_R(t) \) و یک سیگنال چپ‌جهت \( X_L(t) \)، منطقه همگرایی را بررسی کنیم، همانطور که در شکل‌های 9.9(b) و 9.9(c) نشان داده شده است. تبدیل لاپلاس \( x(t) \) زمانی همگرا می‌شود که تبدیل‌های لاپلاس هر دو سیگنال \( X_R(t) \) و \( X_L(t) \) همگرا باشند. از ویژگی 4، منطقه همگرایی \( L\{X_R(t)\} \) شامل نیم‌صفحه \( \text{Re}(s) > \sigma_R \) برای مقداری از \( \sigma_R \) است و از ویژگی 5، منطقه همگرایی \( L\{X_L(t)\} \) شامل نیم‌صفحه \( \text{Re}(s) < \sigma_L \) برای مقداری از \( \sigma_L \) است. منطقه همگرایی \( L\{x(t)\} \) سپس تلاقی این دو نیم‌صفحه خواهد بود، همانطور که در شکل 9.10 نشان داده شده است. این فرض با این پیش‌فرض است که \( \sigma_R < \sigma_L \)، به‌طوری که تلاقی‌ای وجود دارد. اگر اینطور نباشد، حتی اگر تبدیل‌های لاپلاس هر دو سیگنال \( X_R(t) \) و \( X_L(t) \) به‌طور جداگانه وجود داشته باشند، تبدیل لاپلاس \( x(t) \) وجود نخواهد داشت.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the time variable
t = np.linspace(0, 10, 1000)

# Define the function x(t)
x_t = np.sin(t) * np.exp(-0.1 * t)

# Plot the function
plt.plot(t, x_t, label='x(t)')

# Mark T1
plt.axvline(x=1, color='gray', linestyle='--')
plt.text(1, 1.1, 'T1', horizontalalignment='center')

# Set axis labels
plt.xlabel('t')
plt.ylabel('x(t)')

# Set title
plt.title('Graph of x(t) over time')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

<h3>9.3 تبدیل معکوس لاپلاس</h3>

<p>
در بخش 9.1، تفسیر تبدیل لاپلاس یک سیگنال به‌عنوان تبدیل فوریه نسخه‌ای نمایی‌وزن‌شده از سیگنال مورد بحث قرار گرفت.
به این صورت که با نوشتن \( s \) به شکل \( s = \sigma + j\omega \)، تبدیل لاپلاس یک سیگنال \( x(t) \) به صورت زیر است:
</p>

<p align="center">
$$
X(s) = X(\sigma + j\omega) = \int_{-\infty}^{\infty} x(t)e^{-\sigma t}e^{-j\omega t} \, dt,
$$
</p>

<p align="center">(معادله 9.53)</p>

<p>
برای مقادیر \( s = \sigma + j\omega \) در منطقه همگرایی (ROC).
می‌توانیم این رابطه را با استفاده از تبدیل معکوس فوریه معکوس کنیم. همان‌طور که در معادله (4.9) آمده است، داریم:
</p>

<p align="center">
$$
x(t)e^{-\sigma t} = \mathcal{F}^{-1} \{X(\sigma + j\omega)\},
$$
</p>

<p>
و با ضرب طرفین در \( e^{\sigma t} \)، نتیجه می‌گیریم:
</p>

<p align="center">
$$
x(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} X(\sigma + j\omega)e^{j\omega t} \, d\omega.
$$
</p>

<p align="center">(معادله 9.54)</p>

<p>
یا معادل آن:
</p>

<p align="center">
$$
x(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} X(j\omega)e^{(\sigma + j\omega)t} \, d\omega.
$$
</p>

<p align="center">(معادله 9.55)</p>

</div>

<div dir="rtl" align="justify">

<p>
به‌عبارت‌دیگر، می‌توانیم \( x(t) \) را از تبدیل لاپلاس آن برای مجموعه‌ای از مقادیر \( s = \sigma + j\omega \) در منطقه همگرایی (ROC) بازیابی کنیم، به‌طوری‌که \( \sigma \) ثابت باشد و \( \omega \) از \(-\infty\) تا \(+\infty\) تغییر کند.
برای درک بهتر این موضوع و دستیابی به دیدگاه‌های بیشتر در مورد بازیابی \( x(t) \) از \( X(s) \)،
می‌توان متغیر انتگرال‌گیری در معادله (9.55) را از \( \omega \) به \( s \) تغییر داد و از این واقعیت استفاده کرد که \( \sigma \) ثابت است، بنابراین \( ds = j\,d\omega \) خواهد بود.
نتیجه معادله پایه تبدیل لاپلاس معکوس است:
</p>

<p align="center">
$$
x(t) = \frac{1}{2\pi j} \int_{-\infty}^{\infty} X(s)e^{st} \, ds.
$$
</p>

<p align="center">(معادله 9.56)</p>

<p>
این معادله بیان می‌کند که \( x(t) \) می‌تواند به‌عنوان انتگرالی وزن‌دار از نمایی‌های مختلط نمایش داده شود.
مسیر انتگرال‌گیری در معادله (9.56)، خط مستقیمی در صفحه \( s \) است که تمام نقاط \( s \) را شامل می‌شود و \( \text{Re}\{s\} = \sigma \) را ارضا می‌کند.
این خط موازی با محور \( j\omega \) است.
علاوه‌براین، می‌توانیم هر خطی در منطقه همگرایی (ROC) را انتخاب کنیم، یعنی هر مقداری از \( \sigma \) که در آن \( X(\sigma + j\omega) \) همگرا باشد.
محاسبه دقیق انتگرال برای یک \( X(s) \) کلی، نیازمند استفاده از انتگرال‌گیری در صفحه مختلط است، موضوعی که در اینجا مورد بحث قرار نمی‌گیرد.
بااین‌حال، برای دسته تبدیل‌های لاپلاس عقلانی، تبدیل لاپلاس معکوس را می‌توان بدون محاسبه مستقیم معادله (9.56) و با استفاده از تکنیک بسط کسر جزئی، مشابه روشی که در فصل 4 برای محاسبه تبدیل فوریه معکوس استفاده شد، تعیین کرد.
</p>

<p>
این روش اساساً شامل بسط عبارت جبری عقلانی به ترکیبی خطی از عبارات مرتبه پایین‌تر است.
</p>

<p>
به‌عنوان مثال، فرض کنید هیچ قطب چندتایی وجود نداشته باشد و فرض کنیم که مرتبه چندجمله‌ای مخرج از مرتبه چندجمله‌ای صورت بیشتر است، بنابراین می‌توانیم \( X(s) \) را به‌صورت زیر بسط دهیم:
</p>

<p align="center">
$$
X(s) = \frac{A}{s + 1} + \frac{B}{s + 2}.
$$
</p>

<p align="center">(معادله 9.59)</p>

<p>
از منطقه همگرایی \( X(s) \)، می‌توان منطقه همگرایی هر یک از عبارات فردی در معادله (9.59) را استنباط کرد و سپس از مثال‌های 9.1 و 9.2، تبدیل لاپلاس معکوس هر یک از این عبارات را می‌توان تعیین کرد.
برای معکوس تبدیل هر عبارت از نوع \( \frac{A}{s + a_1} \) در معادله، دو انتخاب ممکن وجود دارد.
اگر منطقه همگرایی به‌سمت راست قطب در \( s = -a_1 \) باشد، تبدیل معکوس این عبارت به صورت \( A e^{a_1 t} u(t) \) است که یک سیگنال راست‌سایده است.
اگر منطقه همگرایی به‌سمت چپ قطب در \( s = a_1 \) باشد، تبدیل معکوس عبارت به صورت \( A e^{a_1 t} u(-t) \) است که یک سیگنال چپ‌سایده است.
جمع کردن تبدیل‌های معکوس عبارت‌های فردی در معادله (9.59) تبدیل معکوس \( X(s) \) را به‌دست می‌دهد.
جزئیات این فرایند بهتر است از طریق مثال‌های متعدد ارائه شود.
</p>

<p>
مثال 9.9:
</p>

<p>
فرض کنید:
</p>

<p align="center">
$$
X(s) = \frac{1}{(s + 1)(s + 2)} \quad \text{برای} \quad \text{Re}\{s\} > -1.
$$
</p>

<p align="center">(معادله 9.58)</p>

<p>
برای به‌دست آوردن تبدیل لاپلاس معکوس، ابتدا بسط کسر جزئی انجام می‌دهیم تا \( X(s) \) به شکل زیر درآید:
</p>

<p align="center">
$$
X(s) = \frac{A}{s + 1} + \frac{B}{s + 2}.
$$
</p>

</div>

<div dir="rtl" align="justify">

<p>
همان‌طور که در پیوست بحث شد، می‌توانیم ضرایب \( A \) و \( B \) را با ضرب کردن دو طرف معادله \( (9.59) \) در \( (s + 1)(s + 2) \) و سپس برابر قرار دادن ضرایب توان‌های مساوی \( s \) در هر دو طرف به‌دست آوریم. به‌طور جایگزین، می‌توانیم از رابطه زیر استفاده کنیم:
</p>

<p align="center">
$$
A = \left[ (s + 1)X(s) \right]_{s=-1} = 1،
$$
</p>

<p align="center">
$$
B = \left[ (s + 2)X(s) \right]_{s=-2} = -1۔
$$
</p>

<p>
بنابراین، گسترش کسر جزئی برای \( X(s) \) به شکل زیر است:
</p>

<p align="center">
$$
X(s) = \frac{1}{s + 1} + \frac{1}{s + 2}۔
$$
</p>

<p>
از مثال‌های 9.1 و 9.2، می‌دانیم که برای یک تبدیل از فرم \( \frac{1}{s + a} \)، دو انتخاب مختلف برای تبدیل معکوس وجود دارد که بستگی به این دارد که منطقه همگرایی (ROC) در سمت چپ یا راست قطب قرار دارد. در نتیجه، باید مشخص کنیم که کدام منطقه همگرایی را برای هر یک از جمله‌های مرتبه اول در معادله \( (9.62) \) در نظر بگیریم. این کار با مراجعه به ویژگی‌های منطقه همگرایی که در بخش 9.2 توضیح داده شده، انجام می‌شود.
</p>

<p>
از آنجا که منطقه همگرایی برای \( X(s) \) به‌صورت \( \text{Re}\{s\} > -1 \) است، منطقه همگرایی برای جمله‌های فردی در گسترش کسر جزئی معادله \( (9.62) \) شامل \( \text{Re}\{s\} > -1 \) است. منطقه همگرایی برای هر جمله سپس می‌تواند به سمت چپ یا راست (یا هر دو) گسترش یابد تا محدود به یک قطب یا بی‌نهایت شود. این موضوع در شکل 9.14 نشان داده شده است.
</p>

<h4>شکل 9.14</h4>
<ul>
    <li>(a) نمودار قطب-صفر و منطقه همگرایی برای \( X(s) \) همان‌طور که در معادله \( (9.58) \) مشخص شده است.</li>
    <li>(b) و (c) نمایانگر جمله‌های فردی در گسترش کسر جزئی در معادله \( (9.62) \) هستند. منطقه همگرایی برای جمع با سایه روشن‌تر نشان داده شده است.</li>
</ul>

<p>
برای جمله‌ای که در شکل 9.14(c) نشان داده شده، منطقه همگرایی برای جمع می‌تواند به سمت چپ گسترش یابد، همان‌طور که نشان داده شده است، تا محدود به یک قطب شود.
</p>

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Top graph: Right-sided signal
t = np.linspace(0, 10, 1000)
right_sided_signal = np.exp(-t) * (t >= 0)
axs[0].plot(t, right_sided_signal)
axs[0].set_title('Right-sided signal')
axs[0].set_xlabel('t')
axs[0].set_ylabel('Amplitude')
axs[0].grid(True)

# Middle graph: Left-sided signal
t = np.linspace(-10, 0, 1000)
left_sided_signal = np.exp(t) * (t <= 0)
axs[1].plot(t, left_sided_signal)
axs[1].set_title('Left-sided signal')
axs[1].set_xlabel('t')
axs[1].set_ylabel('Amplitude')
axs[1].grid(True)

# Bottom graph: Two-sided signal
t = np.linspace(-10, 10, 1000)
two_sided_signal = np.exp(-np.abs(t))
axs[2].plot(t, two_sided_signal)
axs[2].set_title('Two-sided signal')
axs[2].set_xlabel('t')
axs[2].set_ylabel('Amplitude')
axs[2].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

<div dir="rtl" align="justify">

<p>
همان‌طور که در بخش 9.1 دیدیم، تبدیل فوریه یک سیگنال، تبدیل لاپلاس است که بر روی محور \( j\omega \) ارزیابی می‌شود. در این بخش، یک روش برای ارزیابی هندسی تبدیل فوریه و به طور کلی، تبدیل لاپلاس در هر مجموعه‌ای از مقادیر از الگوی قطب-صفر مرتبط با یک تبدیل لاپلاس نسبی را مورد بحث قرار می‌دهیم. برای توسعه این روش، ابتدا یک تبدیل لاپلاس با یک صفر واحد را در نظر می‌گیریم (یعنی \( X(s) = s \)) و آن را در یک مقدار خاص از \( s \)، مثلاً \( s = s_1 \) ارزیابی می‌کنیم.
</p>

<p>
عبارت جبری \( s_1 - a \) که ما آن را به عنوان جمع دو عدد مختلط \( s_1 \) و \( a \) می‌بینیم، هرکدام از این دو عدد مختلط را می‌توان به صورت یک بردار در صفحه مختلط نمایش داد، همان‌طور که در شکل 9.15 نشان داده شده است. برداری که عدد مختلط \( s_1 - a \) را نمایش می‌دهد، سپس جمع بردارهای \( s_1 \) و \( -a \) است، که در شکل می‌بینیم این یک بردار از صفر در \( s = a \) به نقطه \( s_1 \) است. مقدار \( X(s_1) \) سپس دارای مقادیری است که برابر با طول این بردار است و زاویه‌ای که برابر با زاویه بردار نسبت به محور حقیقی است.
</p>

<p>
اگر به جای آن، \( X(s) \) یک قطب واحد در \( s = a \) داشته باشد (یعنی \( X(s) = \frac{1}{s - a} \))، در این صورت مخرج به وسیله همان جمع برداری \( s_1 - a \) نمایان می‌شود، و مقدار \( X(s_1) \) دارای مقادیری خواهد بود که معکوس طول بردار از قطب به \( s = s_1 \) است و زاویه آن معکوس زاویه بردار با محور حقیقی خواهد بود.
</p>

<h4>شکل 9.15</h4>
<p>
نمایش هندسی در صفحه مختلط از بردارهای \( s_1 \)، \( a \)، و \( s_1 - a \) که اعداد مختلط \( s_1 \)، \( a \)، و \( s_1 - a \) را نمایش می‌دهند.
</p>

<p>
یک تبدیل لاپلاس نسبی عمومی‌تر شامل یک حاصل‌ضرب از اصطلاحات قطب و صفر از فرم بحث شده در پاراگراف قبلی است؛ به عبارت دیگر، این تبدیل می‌تواند به فرم زیر فاکتوری شود:
</p>

<p align="center">
$$
X(s) = M \prod (s - b) \prod (s - \alpha)
$$
</p>

<p>
برای ارزیابی \( X(s) \) در \( s = s_1 \)، هر یک از عبارت‌ها در حاصل‌ضرب با یک بردار از صفر یا قطب به نقطه \( s_1 \) نمایش داده می‌شود. بزرگی \( X(s_1) \) سپس برابر با بزرگی عامل مقیاس \( M \)، ضرب در طول‌های بردارهای صفر (یعنی بردارهایی که از صفرها به \( s_1 \) می‌روند) تقسیم بر ضرب طول‌های بردارهای قطب (یعنی بردارهایی که از قطب‌ها به \( s_1 \) می‌روند) خواهد بود. زاویه عدد مختلط \( X(s_1) \) مجموع زاویه‌های بردارهای صفر منهای مجموع زاویه‌های بردارهای قطب است. اگر عامل مقیاس \( M \) در معادله (9.70) منفی باشد، زاویه‌ای اضافی باید در نظر گرفته شود. اگر \( X(s) \) دارای قطب یا صفرهای چندگانه باشد، هر بردار باید به تعداد مرتبه قطب یا صفر در نظر گرفته شود.
</p>

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the complex plane
fig, ax = plt.subplots()

# Define the points
s1 = 1 + 2j
a = 2 + 0j
s1_minus_a = s1 - a

# Plot the points
ax.plot([0, s1.real], [0, s1.imag], 'bo-', label='$s_1$')
ax.plot([0, a.real], [0, a.imag], 'ro-', label='$a$')
ax.plot([a.real, s1.real], [a.imag, s1.imag], 'go-', label='$s_1 - a$')

# Mark the points
ax.text(s1.real, s1.imag, '$s_1$', fontsize=12, ha='right')
ax.text(a.real, a.imag, '$a$', fontsize=12, ha='right')
ax.text(s1_minus_a.real, s1_minus_a.imag, '$s_1 - a$', fontsize=12, ha='right')

# Set the axes
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Set the labels
ax.set_xlabel('Re')
ax.set_ylabel('Im')

# Set the title
ax.set_title('Complex Plane Representation')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

<h3>بخش 9.4: ارزیابی هندسی تبدیل فوریه از روی نمودار قطب-صفر</h3>

<p>
اگر برخی از \( a_j \)ها یا برخی از \( b_i \)ها (یا هر دو) برابر باشند، طول‌ها و زوایای بردارهای مربوط به هر یک از این قطب‌ها یا صفرها باید به تعداد برابر با مرتبه قطب یا صفر محاسبه شوند.
</p>

<h4>مثال 9.12</h4>

<p>
فرض کنید:
</p>

<p align="center">
$$
\text{Re}\{s\} > -1. \tag{9.71}
$$
</p>

<p>
تبدیل لاپلاس به صورت زیر است:
</p>

<p align="center">
$$
X(s) = \frac{1}{s+1}, \quad X(j\omega) = \frac{1}{j\omega + \frac{1}{2}}. \tag{9.72}
$$
</p>

<p>
تبدیل فوریه به صورت \( X(s) \) با جایگذاری \( s = j\omega \) به دست می‌آید. برای این مثال، تبدیل فوریه به صورت زیر است:
</p>

<p align="center">
$$
X(j\omega) = \frac{1}{\sqrt{\omega^2 + \left(\frac{1}{2}\right)^2}}. \tag{9.73}
$$
</p>

<p>
زاویه تبدیل فوریه به صورت زیر است:
</p>

<p align="center">
$$
\angle X(j\omega) = -\tan^{-1}(2\omega). \tag{9.74}
$$
</p>

<p>
نمودار قطب-صفر مربوط به \( X(s) \) در شکل 9.16 نشان داده شده است. برای تعیین تبدیل فوریه به صورت گرافیکی، بردار قطب همان‌طور که نشان داده شده است، رسم می‌شود. اندازه تبدیل فوریه در فرکانس \( \omega \) برابر با معکوس طول بردار از قطب تا نقطه \( j\omega \) روی محور موهومی است. فاز تبدیل فوریه نیز برابر با منفی زاویه این بردار است.
</p>

نمودار قطب-صفر برای مثال 9.12. اندازه \( |X(j\omega)| \) برابر با معکوس طول بردار نشان داده شده است و \( \angle X(j\omega) \) برابر با منفی زاویه این بردار است.

</p>

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the complex plane
fig, ax = plt.subplots()

# Define the points
s1 = 1 + 2j
a = -1/2
s1_minus_a = s1 - a

# Plot the points
ax.plot([0, s1.real], [0, s1.imag], 'bo-', label='$s_1$')
ax.plot([0, a.real], [0, a.imag], 'ro-', label='$a$')
ax.plot([a.real, s1.real], [a.imag, s1.imag], 'go-', label='$s_1 - a$')

# Mark the points
ax.text(s1.real, s1.imag, '$s_1$', fontsize=12, ha='right')
ax.text(a.real, a.imag, '$a$', fontsize=12, ha='right')
ax.text(s1_minus_a.real, s1_minus_a.imag, '$s_1 - a$', fontsize=12, ha='right')

# Set the axes
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5)

# Set the labels
ax.set_xlabel('Re')
ax.set_ylabel('Im')

# Set the title
ax.set_title('Complex Plane Representation')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

<h3>9.4.1 سیستم‌های مرتبه اول</h3>

<p>
به‌عنوان یک تعمیم از مثال 9.12، کلاس سیستم‌های مرتبه اول را که در بخش 6.5.1 به تفصیل مورد بحث قرار گرفت، در نظر می‌گیریم. پاسخ ضربه‌ای برای چنین سیستمی به‌صورت زیر است:
</p>

<p align="center">
$$
h(t) = \frac{1}{\tau} e^{-\frac{t}{\tau}} u(t),
$$
</p>

<p align="center">(معادله 9.75)</p>

<h4>شکل 9.17</h4>
<p>

برای \( \zeta > 1 \)، \( C_1 \) و \( C_2 \) حقیقی هستند و در نتیجه هر دو قطب بر روی محور حقیقی قرار دارند، همانطور که در شکل 9.19(a) نشان داده شده است. حالت \( \zeta > 1 \) اساساً یک حاصل‌ضرب از دو جمله مرتبه اول است، همانطور که در بخش 9.4.1 توضیح داده شد. بنابراین، در این حالت، \( H(j\omega) \) به‌طور یکنواخت کاهش می‌یابد و با افزایش \( \omega \)، \( H(j\omega) \) از 0 در \( \omega = 0 \) به \( -\pi/2 \) در \( \omega \to \infty \) تغییر می‌کند. این موضوع را می‌توان از شکل 9.19(a) تأیید کرد، زیرا مشاهده می‌شود که طول بردار از هر یک از دو قطب به نقطه \( s = j\omega \) به‌طور یکنواخت با افزایش \( \omega \) از 0 افزایش می‌یابد و زاویه هر یک از این بردارها از 0 تا \( \pi/2 \) افزایش می‌یابد.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the poles for different damping ratios
poles = {
    'Overdamped': [-1, -2],
    'Critically Damped': [-1, -1],
    'Underdamped': [-1 + 1j, -1 - 1j]
}

# Create a new figure
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Plot each pole-zero plot
for ax, (label, pole) in zip(axs, poles.items()):
    ax.plot(np.real(pole), np.imag(pole), 'rx', markersize=10, label='Poles')
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(True)
    ax.set_title(f'{label} System')
    ax.set_xlabel('Re')
    ax.set_ylabel('Im')
    ax.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

<div dir="rtl" align="justify">

<p>
بنابراین، برای \( 0 < \zeta < 1 \)، سیستم مرتبه دوم یک فیلتر باندگذر غیرایده‌آل است که پارامتر \( \zeta \) کنترل‌کننده تیزی و عرض قله در پاسخ فرکانسی است. به‌طور خاص، از هندسه در شکل 9.19(d)، مشاهده می‌شود که طول بردار قطب از قطب ربع دوم با ضریبی برابر با \( \sqrt{2} \) از حداقل خود در \( \omega = \omega_n \sqrt{1 - \zeta^2} \) زمانی که \( \omega \) از این مقدار به‌طور افزایشی یا کاهشی به میزان \( \zeta \omega_n \) تغییر می‌کند، افزایش می‌یابد. بنابراین، برای مقادیر کوچک \( \zeta \)، و نادیده گرفتن اثر قطب دورتر از ربع سوم، \( H(j\omega) \) در یک ضریب \( \sqrt{2} \) از مقدار قله خود در بازه فرکانسی زیر باقی می‌ماند:
</p>

<p align="center">
$$
\omega_n \sqrt{1 - \zeta^2} - \zeta \omega_n < \omega < \omega_n \sqrt{1 - \zeta^2} + \zeta \omega_n.
$$
</p>

<p>
اگر عرض باند نسبی \( B \) را به‌عنوان طول این بازه فرکانسی تقسیم بر فرکانس طبیعی بدون میرا شدن \( \omega_n \) تعریف کنیم، می‌بینیم که:
</p>

<p align="center">
$$
B = 2\zeta.
$$
</p>

<p>
بنابراین، هرچقدر که \( \zeta \) به صفر نزدیک‌تر باشد، قله در پاسخ فرکانسی تیزتر و باریک‌تر خواهد بود. همچنین توجه داشته باشید که \( B \) معکوس معیار کیفیت \( Q \) برای سیستم‌های مرتبه دوم است که در بخش 6.5.2 تعریف شده است. بنابراین، با افزایش کیفیت، عرض باند نسبی کاهش می‌یابد و فیلتر به طور فزاینده‌ای انتخابی فرکانسی می‌شود.
</p>

<p>
تصویر مشابهی می‌توان برای \( \omega_n = 1 \) و مقادیر مختلف \( \zeta \) ترسیم کرد. همان‌طور که در شکل 9.20(b) مشاهده می‌شود، بردار قطب ربع دوم از \( H(j\omega) \) تغییر می‌کند، که از شکل 9.19(d) دیده می‌شود. زاویه از \( \pi/4 \) به 0 و سپس به \( -\pi/4 \) تغییر می‌کند زیرا \( \omega \) از \( \omega_n \sqrt{1 - \zeta^2} - \zeta \omega_n \) به \( \omega_n \sqrt{1 - \zeta^2} + \zeta \omega_n \) تغییر می‌کند. برای مقادیر کوچک \( \zeta \)، زاویه برای قطب ربع سوم در این بازه فرکانسی تغییر بسیار کمی دارد، که منجر به تغییر سریع \( H(j\omega) \) به‌طور تقریبی \( 7/2 \) در این بازه می‌شود، همانطور که در شکل نشان داده شده است.
</p>

<p>
تغییر \( \omega_n \) با ثابت نگه داشتن \( \zeta \) فقط مقیاس فرکانسی را در بحث قبلی تغییر می‌دهد. به عبارت دیگر، \( H(j\omega) \) و \( H(w) \) تنها به \( \omega/\omega_n \) بستگی دارند. از شکل 9.19(c)، همچنین می‌توان به راحتی تعیین کرد که چگونه قطب‌ها و ویژگی‌های سیستم با تغییر \( \zeta \) و ثابت نگه داشتن \( \omega_n \) تغییر می‌کنند. از آنجا که \( \cos 0 = 4 \)، قطب‌ها در طول یک نیم‌دایره با شعاع ثابت \( \omega_n \) حرکت می‌کنند. برای \( \zeta = 0 \)، دو قطب روی محور موهومی قرار دارند. به‌طور متناسب، در دامنه زمانی، پاسخ ضربه‌ای سینوسی با میرا نشدن خواهد بود. همان‌طور که \( \zeta \) از 0 به 1 افزایش می‌یابد، دو قطب پیچیده باقی می‌مانند و به سمت نیمه‌صفحه چپ صفحه \( s \) حرکت می‌کنند و بردارهای از مبدا به سمت قطب‌ها، اندازه کلی ثابت \( \omega_n \) را حفظ می‌کنند. همانطور که بخش حقیقی قطب‌ها منفی‌تر می‌شود، پاسخ زمانی با سرعت بیشتری در \( t \to \infty \) کاهش می‌یابد. همچنین، همانطور که مشاهده کردیم، همانطور که \( \zeta \) از 0 به سمت 1 افزایش می‌یابد، عرض باند نسبی پاسخ فرکانسی افزایش می‌یابد و پاسخ فرکانسی کمتر تیز و کمتر انتخابی فرکانسی می‌شود.
</p>

<h4>شکل 9.20:</h4>
<p> (الف) مقدار و (ب) فاز پاسخ فرکانسی برای یک سیستم مرتبه دوم با \( 0 < \zeta < 1 \).</p>

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the frequency range
omega = np.logspace(-1, 2, 500)

# Define the magnitude response for different damping ratios
zeta_01 = 0.1
zeta_02 = 0.2
zeta_04 = 0.4
omega_n = 1

# Magnitude response
magnitude_01 = 20 * np.log10(1 / np.sqrt((1 - (omega / omega_n)**2)**2 + (2 * zeta_01 * omega / omega_n)**2))
magnitude_02 = 20 * np.log10(1 / np.sqrt((1 - (omega / omega_n)**2)**2 + (2 * zeta_02 * omega / omega_n)**2))
magnitude_04 = 20 * np.log10(1 / np.sqrt((1 - (omega / omega_n)**2)**2 + (2 * zeta_04 * omega / omega_n)**2))

# Phase response
phase_01 = -np.arctan2(2 * zeta_01 * omega / omega_n, 1 - (omega / omega_n)**2)
phase_02 = -np.arctan2(2 * zeta_02 * omega / omega_n, 1 - (omega / omega_n)**2)
phase_04 = -np.arctan2(2 * zeta_04 * omega / omega_n, 1 - (omega / omega_n)**2)

# Create a new figure
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot the magnitude response
axs[0].semilogx(omega, magnitude_01, label='ζ = 0.1')
axs[0].semilogx(omega, magnitude_02, label='ζ = 0.2')
axs[0].semilogx(omega, magnitude_04, label='ζ = 0.4')
axs[0].set_title('Magnitude of the Frequency Response')
axs[0].set_xlabel('Normalized Frequency (ω)')
axs[0].set_ylabel('Magnitude (dB)')
axs[0].legend()
axs[0].grid(True)

# Plot the phase response
axs[1].semilogx(omega, phase_01, label='ζ = 0.1')
axs[1].semilogx(omega, phase_02, label='ζ = 0.2')
axs[1].semilogx(omega, phase_04, label='ζ = 0.4')
axs[1].set_title('Phase of the Frequency Response')
axs[1].set_xlabel('Normalized Frequency (ω)')
axs[1].set_ylabel('Phase (radians)')
axs[1].legend()
axs[1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

<h3>9.5.1 خطی بودن تبدیل لاپلاس</h3>

<p>
اگر
\[
x_1(t) \xrightarrow{\mathcal{L}} X_1(s)
\]
با ناحیه همگرایی که به‌عنوان \( R_1 \) نشان داده شده است و
\[
x_2(t) \xrightarrow{\mathcal{L}} X_2(s)
\]
با ناحیه همگرایی که به‌عنوان \( R_2 \) نشان داده شده است،
</p>

<p>
آنگاه تبدیل لاپلاس ترکیب خطی به‌صورت زیر است:
</p>

<p align="center">
$$
\mathcal{L} \left( ax_1(t) + bx_2(t) \right) = aX_1(s) + bX_2(s),
$$
</p>

<p>
که در آن ناحیه همگرایی \( X(s) \) حداقل تقاطع ناحیه‌های همگرایی \( R_1 \) و \( R_2 \) است. این تقاطع ممکن است خالی باشد، در این صورت \( X(s) \) هیچ ناحیه همگرایی ندارد، یعنی \( x(t) \) تبدیل لاپلاس ندارد. برای مثال، برای \( x(t) \) همانطور که در معادله (9.47) در مثال 9.7 آمده است، اگر \( b > 0 \)، ناحیه همگرایی \( X(s) \) تقاطع ناحیه‌های همگرایی دو جمله در مجموع است. اگر \( b < 0 \)، هیچ نقطه مشترکی در \( R_1 \) و \( R_2 \) وجود ندارد؛ یعنی تقاطع خالی است و بنابراین، \( x(t) \) تبدیل لاپلاس ندارد. ناحیه همگرایی همچنین می‌تواند بزرگتر از تقاطع باشد. به عنوان یک مثال ساده، برای \( x_1(t) = x_2(t) \) و \( a = b \) در معادله (9.82)، \( x(t) = 0 \) و بنابراین، \( X(s) = 0 \) است. ناحیه همگرایی \( X(s) \) سپس کل صفحه \( s \) خواهد بود.
</p>

<p>
ناحیه همگرایی مربوط به یک ترکیب خطی از جمله‌ها همیشه می‌تواند با استفاده از ویژگی‌های ناحیه همگرایی که در بخش 9.2 توسعه یافته است، ساخته شود. به طور خاص، از تقاطع ناحیه‌های همگرایی برای جمله‌های فردی (فرض بر این است که خالی نباشد)، می‌توانیم یک خط یا نوار را پیدا کنیم که در ناحیه همگرایی ترکیب خطی است. سپس این ناحیه را به سمت راست (افزایش قسمت حقیقی \( s \)) و به سمت چپ (کاهش قسمت حقیقی \( s \)) به نزدیک‌ترین قطب‌ها (که ممکن است در بی‌نهایت باشند) گسترش می‌دهیم.
</p>

<h4>مثال 9.13</h4>

<p>
در این مثال، نشان داده می‌شود که ناحیه همگرایی تبدیل لاپلاس یک ترکیب خطی از سیگنال‌ها گاهی اوقات می‌تواند فراتر از تقاطع ناحیه‌های همگرایی برای جمله‌های فردی گسترش یابد. فرض کنید:
</p>

<p align="center">
$$
x(t) = x_1(t) - x_2(t),
$$
</p>

</div>

<div dir="rtl" align="justify">

<p>
تبدیل لاپلاس سیگنال‌های \( x_1(t) \) و \( x_2(t) \) به‌ترتیب به‌صورت زیر است:
</p>

<p align="center">
$$
X_1(s) = \frac{1}{s + 1} + \frac{1}{s},
\quad \text{Re}\{s\} > -1,
\quad (9.84)
$$
</p>

<p align="center">
و
</p>

<p align="center">
$$
X_2(s) = \frac{1}{(s + 1)(s + 2)}, \quad \text{Re}\{s\} > -1.
\quad (9.85)
$$
</p>

<p>
نمودار قطب-صفر، شامل نواحی همگرایی \( X_1(s) \) و \( X_2(s) \)، در شکل‌های 9.22(a) و (b) نشان داده شده است. از معادله (9.82)، داریم:
</p>

<p align="center">
$$
X(s) = \frac{1}{s + 1} - \frac{1}{(s + 1)(s + 2)} = \frac{1}{(s + 1)(s + 2)} \cdot \frac{(s + 2)}{(s + 1)} = \frac{1}{(s + 1)(s + 2)}.
\quad (9.86)
$$
</p>

<p>
بنابراین، در ترکیب خطی از \( x_1(t) \) و \( x_2(t) \)، قطب در \( s = -1 \) با یک صفر در \( s = -1 \) لغو می‌شود. نمودار قطب-صفر برای \( X(s) = X_1(s) - X_2(s) \) در شکل 9.22(c) نشان داده شده است. تقاطع نواحی همگرایی برای \( X_1(s) \) و \( X_2(s) \) به‌صورت \( \text{Re}\{s\} > 1 \) است. اما از آنجا که ناحیه همگرایی همیشه توسط یک قطب یا بی‌نهایت محدود می‌شود، برای این مثال، ناحیه همگرایی برای \( X(s) \) می‌تواند به سمت چپ گسترش یابد تا توسط قطب در \( s = -2 \) محدود شود، که نتیجه لغو قطب-صفر در \( s = -1 \) است.
</p>

<h4>شکل 9.22:</h4>
<p>نمودارهای قطب-صفر و نواحی همگرایی برای مثال 9.13: (الف) \( X_1(s) \)، (ب) \( X_2(s) \)، (ج) \( X_1(s) - X_2(s) \). ناحیه همگرایی برای \( X_1(s) - X_2(s) \) شامل تقاطع \( R_1 \) و \( R_2 \) است که سپس می‌تواند به سمت چپ گسترش یابد تا توسط قطب در \( s = -2 \) محدود شود.</p>

<h3>9.5.2 انتقال زمانی</h3>

<p>
اگر
\[
x(t) \xrightarrow{\mathcal{L}} X(s),
\]
با ناحیه همگرایی \( R \)،
</p>

<p>
آنگاه
\[
x(t - t_0) \xrightarrow{\mathcal{L}} e^{-s t_0} X(s),
\]
با ناحیه همگرایی \( R \).
\quad (9.87)
</p>

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the poles and zeros
poles = [-2, -2, -2]
zeros = []

# Create a new figure
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Plot each pole-zero plot
for ax, pole in zip(axs, poles):
    ax.plot(np.real(pole), np.imag(pole), 'rx', markersize=10, label='Pole')
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(True)
    ax.set_title(f'Pole at s = {pole}')
    ax.set_xlabel('Re')
    ax.set_ylabel('Im')
    ax.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

<h3>9.5.3 انتقال در دامنه \( s \)</h3>

<p>
اگر
\[
\mathcal{L} \{ x(t) \} = X(s),
\]
با ناحیه همگرایی \( R \)،
</p>

<p>
آنگاه
\[
e^{s t_0} x(t) \xrightarrow{\mathcal{L}} X(s - s_0), \quad \text{با ناحیه همگرایی} \, R + \text{Re}\{s_0\}.
\quad (9.88)
\]
</p>

<p>
به عبارت دیگر، ناحیه همگرایی مربوط به \( X(s - s_0) \) همان ناحیه همگرایی \( X(s) \) است که به اندازه \( \text{Re}\{s_0\} \) جابجا شده است. بنابراین، برای هر مقداری از \( s \) که در \( R \) باشد، مقدار \( s + \text{Re}\{s_0\} \) در \( R_1 \) خواهد بود. این موضوع در شکل 9.23 نشان داده شده است. توجه داشته باشید که اگر \( X(s) \) یک قطب یا صفر در \( s = a \) داشته باشد، آنگاه \( X(s - s_0) \) یک قطب یا صفر در \( s - s_0 = a \) خواهد داشت، یعنی \( s = a + s_0 \).
</p>

<p>
یک مورد خاص مهم از معادله (9.88) زمانی است که \( s_0 = j \omega_0 \)، یعنی وقتی که یک سیگنال \( x(t) \) برای مدوله کردن یک نمایی مختلط دوره‌ای \( e^{j \omega_0 t} \) استفاده می‌شود. در این صورت، معادله (9.88) به شکل زیر درمی‌آید:
</p>

<p align="center">
$$
e^{j \omega_0 t} x(t) \xrightarrow{\mathcal{L}} X(s - j \omega_0), \quad \text{با ناحیه همگرایی} \, R.
\quad (9.89)
$$
</p>

<p>
سمت راست معادله (9.89) را می‌توان به‌عنوان یک جابجایی در دامنه \( s \) به‌طور موازی با محور \( j \omega \) تعبیر کرد. به عبارت دیگر، اگر تبدیل لاپلاس از \( x(t) \) یک قطب یا صفر در \( s = a \) داشته باشد، آنگاه تبدیل لاپلاس از \( e^{j \omega_0 t} x(t) \) یک قطب یا صفر در \( s = a + j \omega_0 \) خواهد داشت.
</p>

</div>

<div dir="rtl" align="justify">

<p>
تبدیل لاپلاس سیگنال‌های \( x_1(t) \) و \( x_2(t) \) به‌ترتیب به‌صورت زیر است:
</p>

<p align="center">
$$
X_1(s) = \frac{1}{s + 1} + \frac{1}{s},
\quad \text{Re}\{s\} > -1,
\quad (9.84)
$$
</p>

<p align="center">
و
</p>

<p align="center">
$$
X_2(s) = \frac{1}{(s + 1)(s + 2)}, \quad \text{Re}\{s\} > -1.
\quad (9.85)
$$
</p>

<p>
نمودار قطب-صفر، شامل نواحی همگرایی \( X_1(s) \) و \( X_2(s) \)، در شکل‌های 9.22(a) و (b) نشان داده شده است. از معادله (9.82)، داریم:
</p>

<p align="center">
$$
X(s) = \frac{1}{s + 1} - \frac{1}{(s + 1)(s + 2)} = \frac{1}{(s + 1)(s + 2)} \cdot \frac{(s + 2)}{(s + 1)} = \frac{1}{(s + 1)(s + 2)}.
\quad (9.86)
$$
</p>

<p>
بنابراین، در ترکیب خطی از \( x_1(t) \) و \( x_2(t) \)، قطب در \( s = -1 \) با یک صفر در \( s = -1 \) لغو می‌شود. نمودار قطب-صفر برای \( X(s) = X_1(s) - X_2(s) \) در شکل 9.22(c) نشان داده شده است. تقاطع نواحی همگرایی برای \( X_1(s) \) و \( X_2(s) \) به‌صورت \( \text{Re}\{s\} > 1 \) است. اما از آنجا که ناحیه همگرایی همیشه توسط یک قطب یا بی‌نهایت محدود می‌شود، برای این مثال، ناحیه همگرایی برای \( X(s) \) می‌تواند به سمت چپ گسترش یابد تا توسط قطب در \( s = -2 \) محدود شود، که نتیجه لغو قطب-صفر در \( s = -1 \) است.
</p>

<h3>9.5.2 انتقال زمانی</h3>

<p>
اگر
\[
x(t) \xrightarrow{\mathcal{L}} X(s),
\]
با ناحیه همگرایی \( R \)،
</p>

<p>
آنگاه
\[
x(t - t_0) \xrightarrow{\mathcal{L}} e^{-s t_0} X(s),
\]
با ناحیه همگرایی \( R \).
\quad (9.87)
</p>

</div>

<div dir="rtl" align="justify">

<p>
اگر
$$
X_2(s) = \frac{1}{(s+2)(s+1)}, \quad \text{Re}\{s\} > -1,
$$
آنگاه:
$$
X_1(s) X_2(s) = 1,
$$
و ناحیه همگرایی آن تمام صفحه \( s \) خواهد بود. همان‌طور که در فصل 4 دیدیم، خاصیت کانولوشن در زمینه تبدیل فوریه نقش مهمی در تحلیل سیستم‌های خطی با ویژگی زمانی ثابت (LTI) ایفا می‌کند. در بخش‌های 9.7 و 9.8، ما این خاصیت کانولوشن را برای تبدیل لاپلاس در تحلیل سیستم‌های LTI به‌طور خاص‌تر برای دسته‌ای از سیستم‌ها که توسط معادلات دیفرانسیل خطی با ضرایب ثابت نشان داده می‌شوند، به‌طور مفصل بررسی خواهیم کرد.
</p>

<p>
9.5.7 مشتق‌گیری در دامنه زمان
</p>

<p>
اگر
$$
L \{ x(t) \} = X(s), \quad \text{ROC} = R,
$$
آنگاه:
$$
\frac{d x(t)}{d t} \xrightarrow{\mathcal{L}} s X(s), \quad \text{ROC} \supseteq R.
$$
</p>

<p>
این خاصیت از مشتق‌گیری هر دو طرف معکوس تبدیل لاپلاس به‌دست می‌آید، همان‌طور که در معادله (9.56) بیان شده است. به‌طور خاص، اگر \( x(t) = 1 + jx \)، آنگاه:
$$
X(s) = \int e^{st} \, dt.
$$
بنابراین، مشتق‌گیری \( x(t) \) برابر با معکوس تبدیل لاپلاس از \( s X(s) \) است. ناحیه همگرایی \( s X(s) \) شامل ناحیه همگرایی \( X(s) \) است و ممکن است بزرگ‌تر شود، اگر \( X(s) \) قطب مرتبه اول در \( s = 0 \) داشته باشد که توسط ضرب در \( s \) لغو می‌شود. به‌عنوان مثال، اگر \( x(t) = u(t) \) باشد، آنگاه \( X(s) = \frac{1}{s} \) است و ناحیه همگرایی آن \( \text{Re}\{s\} > 0 \) خواهد بود. مشتق‌گیری از \( x(t) \) یک پالس است که تبدیل لاپلاس آن برابر با واحد است و ناحیه همگرایی آن تمام صفحه \( s \) خواهد بود.
</p>

<p>
9.5.8 مشتق‌گیری در دامنه \( s \)
</p>

<p>
با مشتق‌گیری از هر دو طرف معادله تبدیل لاپلاس (9.3)، یعنی:
$$
x(t) = X(s) = \int e^{st} \, dt,
$$
ما به نتیجه زیر خواهیم رسید:
$$
\frac{d X(s)}{d s} = \int (-1) x(t) e^{-st} \, dt.
$$
</p>

</div>

<div dir="rtl" align="justify">

<p>
در نتیجه، اگر
$$
\mathcal{L}\{ x(t) \} = X(s), \quad \text{با ناحیه همگرایی} \, R,
$$
آنگاه:
$$
\mathcal{L}\{-t x(t) \} = \frac{d X(s)}{d s}, \quad \text{با ناحیه همگرایی} \, R.
\quad (9.100)
$$
</p>

<p>
مثال 9.14
</p>

<p>
بیایید تبدیل لاپلاس \( x(t) = t e^{at} u(t) \) را پیدا کنیم.
$$
x(t) = t e^{at} u(t).
\quad (9.101)
$$
</p>

<p>
از آنجا که:
$$
\mathcal{L}\{ e^{-at} u(t) \} = \frac{1}{s + a},
$$
با توجه به معادله (9.100)، داریم:
$$
\frac{d}{ds} \left( \frac{1}{s + a} \right) = \frac{-1}{(s + a)^2}, \quad \text{با ناحیه همگرایی} \, \text{Re}\{s\} > -a.
\quad (9.102)
$$
</p>

<p>
در واقع، با استفاده از کاربرد مکرر معادله (9.100)، به نتیجه زیر می‌رسیم:
$$
\mathcal{L}\{ t e^{-at} u(t) \} = \frac{1}{(s + a)^3}, \quad \text{با ناحیه همگرایی} \, \text{Re}\{s\} > -a.
\quad (9.103)
$$
</p>

<p>
و به‌طور کلی:
$$
\mathcal{L}\{ t^{n-1} e^{-at} u(t) \} = \frac{1}{(s + a)^n}, \quad \text{با ناحیه همگرایی} \, \text{Re}\{s\} > -a.
\quad (9.104)
$$
</p>

<p>
همان‌طور که در مثال بعدی نشان داده می‌شود، این جفت تبدیل لاپلاس خاص زمانی که از روش گسترش کسر جزئی برای تعیین معکوس تبدیل لاپلاس یک تابع کسری با قطب‌های مرتبه بالا استفاده می‌کنیم، بسیار مفید است.
</p>

<p>
مثال 9.15
</p>

<p>
تبدیل لاپلاس زیر را در نظر بگیرید:
$$
X(s) = \frac{2s^2 + 5s + 5}{(s+1)^2 (s + 2)}, \quad \text{با ناحیه همگرایی} \, \text{Re}\{s\} > -1.
$$
با اعمال روش گسترش کسر جزئی که در ضمیمه توضیح داده شده است، می‌توانیم بنویسیم:
$$
X(s) = \frac{2}{(s + 1)^2} + \frac{3}{(s + 2)}, \quad \text{با ناحیه همگرایی} \, \text{Re}\{s\} > -1.
\quad (9.105)
$$
</p>

<h3>9.7 تحلیل و شبیه‌سازی سیستم‌های LTI با استفاده از تبدیل لاپلاس</h3>

<p>
یکی از کاربردهای مهم تبدیل لاپلاس در تحلیل و شبیه‌سازی سیستم‌های LTI است. نقش آن در این دسته از سیستم‌ها به‌طور مستقیم از خاصیت هم convolutions (بخش 9.5.6) نشأت می‌گیرد. به‌طور خاص، تبدیل لاپلاس ورودی و خروجی یک سیستم LTI از طریق ضرب در تبدیل لاپلاس پاسخ ضربه‌ای سیستم به یکدیگر مرتبط می‌شوند. بنابراین،
</p>

<p align="center">
$$
Y(s) = H(s)X(s).
\quad (9.112)
$$
</p>

<p>
که در آن \( X(s) \)، \( Y(s) \) و \( H(s) \) به‌ترتیب تبدیل‌های لاپلاس ورودی، خروجی و پاسخ ضربه‌ای سیستم هستند. معادله (9.112) معادل معادله (4.56) برای تبدیل فوریه است. همچنین، از بحث ما در بخش 3.2 در مورد پاسخ سیستم‌های LTI به نمایی‌های پیچیده، اگر ورودی یک سیستم LTI به‌صورت \( x(t) = e^{st} \) باشد و \( s \) در ناحیه همگرایی \( H(s) \) باشد، سپس خروجی \( H(s)e^{st} \) خواهد بود؛ یعنی \( e^{st} \) یک تابع ویژه سیستم با مقادیر ویژه برابر با تبدیل لاپلاس پاسخ ضربه‌ای است.
</p>

<p>
اگر ناحیه همگرایی \( H(s) \) شامل محور موهومی باشد، سپس برای \( s = j\omega \)، \( H(s) \) پاسخ فرکانسی سیستم LTI است. در زمینه وسیع‌تر تبدیل لاپلاس، \( H(s) \) معمولاً به‌عنوان تابع سیستم یا به‌طور جایگزین تابع انتقال شناخته می‌شود. بسیاری از ویژگی‌های سیستم‌های LTI به‌طور نزدیکی با ویژگی‌های تابع سیستم در صفحه \( s \) مرتبط هستند. در ادامه، این را با بررسی چند ویژگی و کلاس سیستم‌های مهم نشان می‌دهیم.
</p>

<h4>9.7.1 علیت</h4>

<p>
برای یک سیستم LTI علی، پاسخ ضربه‌ای برای \( t < 0 \) صفر است و بنابراین سمت راست است. بنابراین، از بحث در بخش 9.2، می‌بینیم که
</p>

<p align="center">
$$
ناحیه همگرایی مرتبط با تابع سیستم برای یک سیستم علی، یک صفحه نیمه راست است.
$$
</p>

<p>
اما باید تأکید کرد که معکوس این جمله لزوماً درست نیست. یعنی، همان‌طور که در مثال 9.19 به‌دنبال آن خواهیم دید، ناحیه همگرایی به سمت راست راست‌ترین قطب ممکن است شامل ناحیه‌های دیگری نیز باشد.
</p>

</div>

<div dir="rtl" align="justify">

<p>
قطب‌ها تضمین‌کننده این نیستند که یک سیستم علی است؛ بلکه تنها تضمین می‌کنند که پاسخ ضربه‌ای سمت راست است. اما اگر \( H(s) \) یک تابع سیستم گ rational باشد، آنگاه همانطور که در مثال‌های 9.17 و 9.18 به‌دنبال آن خواهیم دید، می‌توانیم با بررسی اینکه ناحیه همگرایی (ROC) آن یک صفحه نیمه راست است، تعیین کنیم که آیا سیستم علی است یا نه. به‌طور خاص،
</p>

<p>
برای یک سیستم با تابع سیستم گ rational، علیت سیستم معادل است با اینکه ناحیه همگرایی به سمت راست راست‌ترین قطب باشد.
</p>

<h4>مثال 9.17</h4>

<p>
یک سیستم با پاسخ ضربه‌ای را در نظر بگیرید:
</p>

<p align="center">
$$
h(t) = e^{-t}u(t).
\quad (9.113)
$$
</p>

<p>
از آنجا که \( h(t) = 0 \) برای \( t < 0 \)، این سیستم علی است. همچنین، تابع سیستم را می‌توان از مثال 9.1 به‌دست آورد:
</p>

<p align="center">
$$
H(s) = \frac{1}{s + 1}, \quad \text{Re}\{s\} > -1.
\quad (9.114)
$$
</p>

<p>
در این حالت، تابع سیستم گ rational است و ناحیه همگرایی در معادله (9.114) به سمت راست راست‌ترین قطب است که با بیان ما مبنی بر اینکه علیت برای سیستم‌هایی با توابع سیستم گ rational معادل است با ناحیه همگرایی که به سمت راست راست‌ترین قطب است، همخوانی دارد.
</p>

<h4>مثال 9.18</h4>

<p>
یک سیستم با پاسخ ضربه‌ای را در نظر بگیرید:
</p>

<p align="center">
$$
h(t) = e^{t}u(t).
$$
</p>

<p>
از آنجا که \( h(t) \neq 0 \) برای \( t < 0 \)، این سیستم علی نیست. همچنین، از مثال 9.7، تابع سیستم به‌صورت زیر است:
</p>

<p align="center">
$$
H(s) = \frac{2}{s + 1}, \quad -1 < \text{Re}\{s\} < +1.
$$
</p>

<p>
بنابراین، \( H(s) \) گ rational است و ناحیه همگرایی آن به‌صورت ناحیه‌ای نیست که به سمت راست راست‌ترین قطب باشد، که با این واقعیت که سیستم علی نیست همخوانی دارد.
</p>

<h4>مثال 9.19</h4>

<p>
یک تابع سیستم را در نظر بگیرید:
</p>

<p align="center">
$$
H(s) = \frac{1}{s + 1}, \quad \text{Re}\{s\} > -1.
\quad (9.115)
$$
</p>

<p>
برای این سیستم، ناحیه همگرایی به سمت راست راست‌ترین قطب است. بنابراین، پاسخ ضربه‌ای باید سمت راست باشد. برای تعیین پاسخ ضربه‌ای، ابتدا از نتیجه استفاده می‌کنیم:
</p>

</div>

<div dir="rtl" align="justify">

<p>
از مثال 9.1:
</p>

<p align="center">
$$
H(s) = \frac{1}{s + 1}, \quad \text{Re}\{s\} > -1.
\quad (9.116)
$$
</p>

<p>
سپس، از خاصیت انتقال زمانی در بخش 9.5.2 [معادله (9.87)]، فاکتور \( e^s \) در معادله (9.115) می‌تواند با انتقال در تابع زمان در معادله (9.116) حساب شود. بنابراین،
</p>

<p align="center">
$$
e^{-(t+1)}u(t + 1) \xrightarrow{\mathcal{L}} \frac{1}{s + 1}, \quad \text{Re}\{s\} > -1,
$$
</p>

<p>
پس پاسخ ضربه‌ای مرتبط با سیستم به‌صورت زیر است:
</p>

<p align="center">
$$
h(t) = e^{-(t+1)}u(t + 1),
\quad (9.118)
$$
</p>

<p>
که برای \( -1 < t < 0 \) غیرصفر است. بنابراین، سیستم علی نیست. این مثال به‌عنوان یادآوری است که علیت به این معناست که ناحیه همگرایی باید به سمت راست راست‌ترین قطب باشد، اما برعکس این قضیه در حالت عمومی درست نیست، مگر اینکه تابع سیستم گ rational باشد.
</p>

<p>
به‌طور مشابه، می‌توانیم با مفهوم ضد-علیت نیز برخورد کنیم. یک سیستم ضد-علی است اگر پاسخ ضربه‌ای آن \( h(t) = 0 \) برای \( t > 0 \) باشد. از آنجا که در این حالت \( h(t) \) سمت چپ است، می‌دانیم از بخش 9.2 که ناحیه همگرایی تابع سیستم \( H(s) \) باید یک صفحه نیمه چپ باشد. دوباره، در حالت عمومی، برعکس این قضیه درست نیست. یعنی، اگر ناحیه همگرایی \( H(s) \) یک صفحه نیمه چپ باشد، تنها چیزی که می‌دانیم این است که \( h(t) \) سمت چپ است. با این حال، اگر \( H(s) \) گ rational باشد، آنگاه داشتن ناحیه همگرایی به سمت چپ چپ‌ترین قطب معادل با این است که سیستم ضد-علی باشد.
</p>

<h4>9.7.2 پایداری</h4>

<p>
ناحیه همگرایی \( H(s) \) همچنین می‌تواند به پایداری یک سیستم مرتبط باشد. همانطور که در بخش 2.3.7 ذکر شده است، پایداری یک سیستم LTI معادل با این است که پاسخ ضربه‌ای آن قابل انتگرال‌گیری مطلق باشد، که در این صورت (بخش 4.4) تبدیل فوریه پاسخ ضربه‌ای همگرا می‌شود. از آنجا که تبدیل فوریه یک سیگنال برابر با تبدیل لاپلاس آن است که در امتداد محور \( j\omega \) ارزیابی می‌شود، ما به نتیجه زیر می‌رسیم:
</p>

<p>
یک سیستم LTI پایدار است اگر و تنها اگر ناحیه همگرایی تابع سیستم \( H(s) \) شامل تمام محور \( j\omega \) باشد [یعنی \( \text{Re}\{s\} = 0 \)].
</p>

<h4>مثال 9.20</h4>

<p>
یک سیستم LTI را با تابع سیستم زیر در نظر بگیرید:
</p>

<p align="center">
$$
H(s) = \frac{1}{(s - 1)(s + 1)(s - 2)},
\quad (9.119)
$$
</p>

<p>
از آنجا که ناحیه همگرایی مشخص نشده است، از بحث ما در بخش 9.2 می‌دانیم که چندین ناحیه همگرایی مختلف وجود دارد و در نتیجه، چندین پاسخ ضربه‌ای مختلف برای تابع سیستم \( H(s) \) که در معادله (9.119) داده شده است، می‌تواند وجود داشته باشد.
</p>

</div>

<div dir="rtl" align="justify">

<p>
با این حال، اگر اطلاعاتی در مورد علیت یا پایداری سیستم داشته باشیم، می‌توان ناحیه همگرایی مناسب را شناسایی کرد. به‌عنوان مثال، اگر سیستم شناخته شده باشد که علی است، ناحیه همگرایی به‌صورت ناحیه‌ای که در شکل 9.25(a) نشان داده شده است، خواهد بود و پاسخ ضربه‌ای آن به‌صورت زیر است:
</p>

<p align="center">
$$
h(t) = (e^t + e^{2t})u(t).
\quad (9.120)
$$
</p>

<p>
توجه داشته باشید که این انتخاب خاص از ناحیه همگرایی شامل محور \( j\omega \) نیست و بنابراین سیستم مربوطه ناپایدار است (چرا که می‌توان با مشاهده اینکه \( h(t) \) قابل انتگرال‌گیری مطلق نیست، این موضوع را بررسی کرد). از طرف دیگر، اگر سیستم شناخته شود که پایدار است، ناحیه همگرایی به‌صورت ناحیه‌ای که در شکل 9.25(b) نشان داده شده است، خواهد بود و پاسخ ضربه‌ای مربوطه به‌صورت زیر است:
</p>

<p align="center">
$$
h(t) = u(t) - u(t + 1),
$$
</p>

<p>
که قابل انتگرال‌گیری مطلق است. در نهایت، برای ناحیه همگرایی در شکل 9.25(c)، سیستم ضد-علی و ناپایدار است و پاسخ ضربه‌ای آن به‌صورت زیر خواهد بود:
</p>

<p align="center">
$$
h(t) = - (e^{3t} + e^{2t})u(-t).
$$
</p>

<p>
در شکل 9.25، ناحیه‌های همگرایی مختلف برای تابع سیستم در مثال 9.20 با قطب‌هایی در \( s = -1 \) و \( s = 2 \) و یک صفر در \( s = 1 \) نشان داده شده است: (الف) سیستم علی و ناپایدار، (ب) سیستم غیرعلی و پایدار، (ج) سیستم ضد-علی و ناپایدار.
</p>

<h4>شکل 9.25:</h4>
<p>
ناحیه‌های همگرایی ممکن برای تابع سیستم در مثال 9.20 با قطب‌هایی در \( s = -1 \) و \( s = 2 \) و یک صفر در \( s = 1 \): (الف) سیستم علی و ناپایدار؛ (ب) سیستم غیرعلی و پایدار؛ (ج) سیستم ضد-علی و ناپایدار.
</p>

</div>

<div dir="rtl" align="justify">

<p>
البته کاملاً ممکن است که یک سیستم پایدار (یا ناپایدار) باشد و تابع سیستم آن غیرعقلانی باشد. به‌عنوان مثال، تابع سیستم در معادله (9.115) غیرعقلانی است و پاسخ ضربه‌ای آن در معادله (9.118) قابل انتگرال‌گیری مطلق است، که نشان‌دهنده پایداری سیستم است. با این حال، برای سیستم‌هایی که تابع سیستم آن‌ها عقلانی است، پایداری به راحتی با توجه به موقعیت‌های قطب‌های سیستم تفسیر می‌شود. به‌عنوان مثال، برای نمودار قطب-صفر در شکل 9.25، پایداری به انتخاب ناحیه همگرایی‌ای که بین دو قطب باشد مربوط می‌شود، به‌طوری که محور \( j\omega \) در ناحیه همگرایی قرار گیرد.
</p>

<p>
برای یک کلاس خاص و بسیار مهم از سیستم‌ها، پایداری می‌تواند به‌طور ساده‌ای از طریق موقعیت قطب‌ها توصیف شود. به‌طور خاص، یک سیستم علی LTI با تابع سیستم عقلانی \( H(s) \) را در نظر بگیرید. از آنجا که سیستم علی است، ناحیه همگرایی به‌صورت ناحیه‌ای به‌راستی از راست‌ترین قطب خواهد بود. بنابراین، برای اینکه این سیستم پایدار باشد (یعنی برای اینکه ناحیه همگرایی شامل محور \( j\omega \) باشد)، باید راست‌ترین قطب \( H(s) \) به سمت چپ محور \( j\omega \) قرار گیرد. به عبارت دیگر،
</p>

<p align="center">
$$
\text{یک سیستم علی با تابع سیستم عقلانی } H(s) \text{ پایدار است اگر و تنها اگر تمامی قطب‌های } H(s) \text{ در نیم‌صفحه چپ صفحه } s \text{ قرار گیرند؛ یعنی تمامی قطب‌ها دارای بخش حقیقی منفی هستند.}
$$
</p>

<p>
مثال 9.21:
</p>

<p>
سیستم علی را که در مثال 9.17 آورده شده دوباره در نظر بگیرید. پاسخ ضربه‌ای در معادله (9.113) قابل انتگرال‌گیری مطلق است و بنابراین سیستم پایدار است. بر اساس این موضوع، مشاهده می‌کنیم که قطب تابع سیستم \( H(s) \) در معادله (9.114) در \( s = -1 \) قرار دارد، که در نیم‌صفحه چپ صفحه \( s \) است. در مقابل، سیستم علی با پاسخ ضربه‌ای \( h(t) = e^{2t}u(t) \) ناپایدار است، چرا که \( h(t) \) قابل انتگرال‌گیری مطلق نیست. همچنین در این حالت، تابع سیستم \( H(s) = \frac{1}{s - 2} \) است و ناحیه همگرایی \( \text{Re}\{s\} > 2 \) دارد، بنابراین سیستم یک قطب در \( s = 2 \) در نیم‌صفحه راست صفحه \( s \) دارد.
</p>

<p>
مثال 9.22:
</p>

<p>
کلاس سیستم‌های علی مرتبه دوم را که پیش‌تر در بخش‌های 9.4.2 و 6.5.2 مورد بحث قرار داده‌ایم، در نظر بگیرید. پاسخ ضربه‌ای و تابع سیستم به ترتیب به‌صورت زیر است:
</p>

<p align="center">
$$
h(t) = M[e^{-\omega_n t} \sin(\omega_n t)]u(t),
\quad (9.121)
$$
</p>

<p align="center">
و
</p>

<p align="center">
$$
H(s) = \frac{\omega_n^2}{(s - C_1)(s - C_2)},
\quad \text{که در آن}
$$
</p>

<p align="center">
$$
C_1 = -\zeta\omega_n + \omega_n\sqrt{\zeta^2 - 1},
\quad C_2 = -\zeta\omega_n - \omega_n\sqrt{\zeta^2 - 1},
\quad M = \frac{1}{\zeta^2 - 1}.
\quad (9.122)
$$
</p>

</div>

<div dir="rtl" align="justify">

<p>
شکل 9.26 موقعیت‌های قطب‌ها و ناحیه همگرایی (ROC) برای یک سیستم علی مرتبه دوم را نشان می‌دهد که در آن \( \zeta < 0 \). در شکل 9.19، موقعیت‌های قطب‌ها برای \( \zeta > 0 \) نشان داده شده است. در شکل 9.26، موقعیت‌های قطب‌ها برای \( \zeta < 0 \) نشان داده شده است. همان‌طور که از این شکل و از معادلات (9.124) و (9.125) پیداست، برای \( \zeta < 0 \) هر دو قطب دارای بخش‌های حقیقی مثبت هستند. بنابراین، برای \( \zeta < 0 \)، سیستم مرتبه دوم علی نمی‌تواند پایدار باشد. این موضوع همچنین از معادله (9.121) مشهود است، زیرا با \( \text{Re}\{C_1\} > 0 \) و \( \text{Re}\{C_2\} > 0 \)، هر عبارت به‌طور نمایی با افزایش \( t \) رشد می‌کند و در نتیجه \( h(t) \) نمی‌تواند قابل انتگرال‌گیری مطلق باشد.
</p>

<p>
9.7.3 سیستم‌های LTI مشخص‌شده با معادلات تفاضلی با ضرایب ثابت خطی
</p>

<p>
در بخش 4.7، استفاده از تبدیل فوریه برای به‌دست آوردن پاسخ فرکانسی یک سیستم LTI که با یک معادله تفاضلی با ضرایب ثابت خطی مشخص شده است، مورد بحث قرار گرفت. به‌طور مشابه، می‌توان از ویژگی‌های تبدیل لاپلاس برای به‌دست آوردن مستقیم تابع سیستم برای یک سیستم LTI که با یک معادله تفاضلی با ضرایب ثابت خطی مشخص شده است، استفاده کرد. در مثال بعدی این روش توضیح داده می‌شود.
</p>

<p>
مثال 9.23
</p>

<p>
سیستمی LTI را در نظر بگیرید که ورودی آن \( x(t) \) و خروجی آن \( y(t) \) معادله تفاضلی با ضرایب ثابت خطی زیر را راضی می‌کنند:
</p>

<p align="center">
$$
\frac{d y(t)}{dt} + 3y(t) = x(t).
\quad (9.126)
$$
</p>

</div>

<div dir="rtl" align="justify">

<p>
با اعمال تبدیل لاپلاس به هر دو طرف معادله (9.126) و استفاده از ویژگی‌های خطی بودن و مشتق‌گیری که در بخش‌های 9.5.1 و 9.5.7 توضیح داده شده‌اند (معادلات (9.82) و (9.98) به ترتیب)، معادله جبری زیر به‌دست می‌آید:
</p>

<p align="center">
$$
sY(s) + 3Y(s) = X(s).
$$
</p>

<p>
از آنجا که طبق معادله (9.112) تابع سیستم به‌صورت زیر است:
</p>

<p align="center">
$$
H(s) = \frac{Y(s)}{X(s)},
$$
</p>

<p>
بنابراین، برای این سیستم داریم:
</p>

<p align="center">
$$
H(s) = \frac{1}{s + 3}.
\quad (9.127)
$$
</p>

<p>
این معادله به‌طور کامل بیان‌گر تابع سیستم است، اما ناحیه همگرایی (ROC) را مشخص نمی‌کند. در حقیقت، همان‌طور که در بخش 2.4 توضیح داده شده است، معادله تفاضلی به‌تنهایی مشخص‌کننده کامل سیستم LTI نیست و معمولاً پاسخ‌های مختلفی برای آن وجود دارد که همگی با معادله تفاضلی سازگارند. اگر علاوه بر معادله تفاضلی، بدانیم که سیستم علی است، آنگاه می‌توان نتیجه گرفت که ناحیه همگرایی باید به‌سمت راست‌ترین قطب باشد که در اینجا معادل \( \text{Re}\{s\} > -3 \) است. اگر سیستم به‌عنوان سیستم غیرعلی (آنتی‌کازال) شناخته شود، آنگاه ناحیه همگرایی تابع سیستم \( H(s) \) برابر با \( \text{Re}\{s\} < -3 \) خواهد بود. پاسخ ضربه‌ای معادل در حالت علی به‌صورت زیر است:
</p>

<p align="center">
$$
h(t) = e^{-3t}u(t),
\quad (9.129)
$$
</p>

<p>
در حالی که در حالت غیرعلی (آنتی‌کازال) پاسخ ضربه‌ای به‌صورت زیر خواهد بود:
</p>

<p align="center">
$$
h(t) = -e^{3t}u(-t).
\quad (9.130)
$$
</p>

<p>
همان‌طور که در مثال 9.23 نشان داده شد، از روش مشابهی برای به‌دست آوردن \( H(s) \) از معادله تفاضلی می‌توان به‌طور عمومی‌تر استفاده کرد. در نظر بگیرید که یک معادله تفاضلی خطی با ضرایب ثابت عمومی به‌صورت زیر است:
</p>

<p align="center">
$$
\sum_{k=0}^{N} a_k \frac{d^k y(t)}{dt^k} = \sum_{k=0}^{M} b_k \frac{d^k x(t)}{dt^k}.
\quad (9.131)
$$
</p>

<p>
با اعمال تبدیل لاپلاس به هر دو طرف معادله و استفاده مکرر از ویژگی‌های خطی بودن و مشتق‌گیری، معادله‌های زیر به‌دست می‌آید:
</p>

<p align="center">
$$
\sum_{k=0}^{N} a_k s^k Y(s) = \sum_{k=0}^{M} b_k s^k X(s).
\quad (9.132)
$$
</p>

<p>
یا معادله جبری به‌صورت زیر خواهد بود:
</p>

<p align="center">
$$
H(s) = \frac{\sum_{k=0}^{M} b_k s^k}{\sum_{k=0}^{N} a_k s^k}.
\quad (9.133)
$$
</p>

</div>

<div dir="rtl" align="justify">

<p>
بنابراین، تابع سیستم برای سیستمی که توسط یک معادله تفاضلی مشخص شده است، همیشه یک کسر متناسب است، با صفرها در مکان‌های جواب معادله زیر:
</p>

<p align="center">
$$
\sum_{k=0}^{M} b_k s^k = 0.
\quad (9.134)
$$
</p>

<p>
و قطب‌ها در مکان‌های جواب معادله زیر:
</p>

<p align="center">
$$
\sum_{k=0}^{M} a_k s^k = 0.
\quad (9.135)
$$
</p>

<p>
مطابق با بحث قبلی، معادله (9.133) شامل مشخصات ناحیه همگرایی تابع سیستم \( H(s) \) نیست، زیرا معادله تفاضلی با ضرایب ثابت به‌تنهایی ناحیه همگرایی را محدود نمی‌کند. با این حال، با اطلاعات اضافی، مانند آگاهی از پایداری یا علی بودن سیستم، می‌توان ناحیه همگرایی را استنباط کرد. به‌عنوان مثال، اگر شرط شروع صفر برای سیستم اعمال شود، به‌طوری که سیستم علی باشد، آنگاه ناحیه همگرایی به‌سمت راست‌ترین قطب خواهد بود.
</p>

<p>
مثال 9.24:
</p>

<p>
یک مدار RLC که ولتاژ خازن و جریان القاگر آن به‌طور اولیه صفر است، یک سیستم LTI است که می‌توان آن را با یک معادله تفاضلی با ضرایب ثابت خطی توصیف کرد. در نظر بگیرید مدار سری RLC در شکل 9.27. فرض کنید ولتاژ across منبع ولتاژ ورودی سیستم \( x(t) \) است و ولتاژ اندازه‌گیری شده across خازن خروجی سیستم \( y(t) \) است. با برابری مجموع ولتاژها across مقاوم، القاگر و خازن با ولتاژ منبع، معادله زیر به‌دست می‌آید:
</p>

<p align="center">
$$
RC \frac{d y(t)}{dt} + LC \frac{d^2 y(t)}{dt^2} + y(t) = x(t).
\quad (9.136)
$$
</p>

<p>
با اعمال معادله (9.133) به‌دست می‌آید:
</p>

<p align="center">
$$
H(s) = \frac{1}{LC} \frac{1}{s^2 + \left(\frac{R}{L}\right) s + \frac{1}{LC}}.
\quad (9.137)
$$
</p>

<p>
همان‌طور که در مسئله 9.64 نشان داده شده است، اگر مقادیر \( R \)، \( L \)، و \( C \) همگی مثبت باشند، قطب‌های این تابع سیستم دارای بخش حقیقی منفی خواهند بود و در نتیجه سیستم پایدار خواهد بود.
</p>

</div>

<p>
9.7.4 مثال‌های مربوط به رفتار سیستم نسبت به تابع سیستم
</p>

<p>
همان‌طور که دیده‌ایم، ویژگی‌های سیستم مانند علی بودن و پایداری می‌توانند مستقیماً به تابع سیستم و ویژگی‌های آن مربوط شوند. در واقع، هر یک از ویژگی‌های تبدیل لاپلاس که توضیح داده‌ایم، می‌توانند به این روش برای ارتباط رفتار سیستم با تابع سیستم استفاده شوند. در این بخش، چندین مثال برای توضیح این موضوع آورده شده است.
</p>

<p>
مثال 9.25:
</p>

<p>
فرض کنید که می‌دانیم اگر ورودی یک سیستم LTI به صورت زیر باشد:
</p>

<p align="center">
$$
x(t) = e^{3}u(t),
$$
</p>

<p>
آنگاه خروجی سیستم به شکل زیر است:
</p>

<p align="center">
$$
y(t) = \left[ e^{e^{-2t}} \right] u(t).
$$
</p>

<p>
همان‌طور که در ادامه نشان خواهیم داد، از این اطلاعات می‌توان تابع سیستم این سیستم را تعیین کرده و از آن به سرعت ویژگی‌های دیگر سیستم را استنباط کرد. با اعمال تبدیل لاپلاس به \( x(t) \) و \( y(t) \) داریم:
</p>

<p align="center">
$$
X(s) = \frac{1}{s+3}, \quad \text{Re}\{s\} > -3,
$$
</p>

<p align="center">
$$
Y(s) = \frac{1}{(s + 1)(s + 2)}, \quad \text{Re}\{s\} > -1.
$$
</p>

<p>
با توجه به معادله (9.112)، می‌توان نتیجه گرفت که تابع سیستم به‌صورت زیر است:
</p>

<p align="center">
$$
H(s) = \frac{Y(s)}{X(s)} = \frac{s+3}{(s+1)(s+2)} = \frac{s+3}{s^2 + 3s + 2}, \quad \text{Re}\{s\} > -1.
$$
</p>

<p>
علاوه بر این، می‌توانیم ناحیه همگرایی (ROC) این سیستم را نیز تعیین کنیم. به‌ویژه، از خاصیت هم‌گام‌سازی که در بخش 9.5.6 ذکر شده است، می‌دانیم که ناحیه همگرایی \( Y(s) \) باید حداقل تقاطع نواحی همگرایی \( X(s) \) و \( H(s) \) را دربر بگیرد. با بررسی سه انتخاب مختلف برای ناحیه همگرایی \( H(s) \) (یعنی به‌سمت چپ قطب در \( s = 2 \)، بین قطب‌ها در \( -2 \) و \( 1 \)، و به‌سمت راست قطب در \( s = -1 \))، می‌بینیم که تنها انتخابی که با نواحی همگرایی \( X(s) \) و \( Y(s) \) سازگار است، \( \text{Re}\{s\} > 1 \) است. از آنجا که این ناحیه به‌سمت راست‌ترین قطب تابع \( H(s) \) است، نتیجه می‌گیریم که تابع سیستم علی است و از آنجا که هر دو قطب \( H(s) \) دارای بخش حقیقی منفی هستند، سیستم پایدار است. علاوه بر این، از رابطه بین معادلات (9.131) و (9.133) می‌توان معادله تفاضلی که سیستم را با شرایط شروع صفر مشخص می‌کند، به‌دست آورد:
</p>

<p align="center">
$$
\frac{d^2 y(t)}{dt^2} + 3 \frac{dy(t)}{dt} + 2y(t) = \frac{dx(t)}{dt} + 3x(t).
$$
</p>

<p>
مثال 9.26:
</p>

<p>
فرض کنید که اطلاعات زیر در مورد یک سیستم LTI داده شده است:
</p>

<ul>
  <li>سیستم علی است.</li>
  <li>تابع سیستم معقول است و تنها دو قطب در \( s = 2 \) و \( s = 4 \) دارد.</li>
</ul>

</div>

<div dir="rtl" align="justify">

<p>
3. اگر \( x(t) = 1 \)، آنگاه \( y(t) = 0 \).
</p>

<p>
4. مقدار پاسخ ضربه‌ای در \( t = 0+ \) برابر با 4 است.
</p>

<p>
با استفاده از این اطلاعات، ما می‌خواهیم تابع سیستم سیستم را تعیین کنیم.
</p>

<p>
از دو حقیقت اول، می‌دانیم که سیستم ناپایدار است (زیرا علی است و یک قطب در \( s = 4 \) با بخش حقیقی مثبت دارد) و تابع سیستم از فرم زیر است:
</p>

<p align="center">
$$
H(s) = \frac{p(s)}{(s + 2)(s-4)} p(s) = \frac{s^2 - 2s - 8}{s^2}.
$$
</p>

<p>
چون پاسخ \( y(t) \) به ورودی \( x(t) = 1 \) باید برابر با \( H(0) \), می‌توان نتیجه گرفت که \( p(0) = 0 \)، یعنی \( p(s) \) باید ریشه‌ای در \( s = 0 \) داشته باشد و بنابراین به فرم زیر است:
</p>

<p align="center">
$$
p(s) = s q(s),
$$
</p>

<p>
در نهایت، از حقیقت چهارم و قضیه مقدار اولیه در بخش 9.5.10، مشاهده می‌کنیم که:
</p>

<p align="center">
$$
\lim_{s \to \infty} sH(s) = \lim_{s \to \infty} s^2 q(s) = 4.
$$
</p>

<p>
هنگامی که \( s \) به بی‌نهایت میل می‌کند، عبارت‌هایی با بالاترین درجه در صورت و مخرج \( sH(s) \) غالب می‌شوند و بنابراین تنها همین موارد در ارزیابی معادله (9.138) اهمیت دارند. علاوه بر این، اگر درجه صورت از درجه مخرج بیشتر باشد، حد آن انحراف پیدا می‌کند. به‌این‌ترتیب، ما می‌توانیم یک مقدار ثابت و غیر صفر برای حد بدست آوریم فقط اگر درجه صورت تابع \( sH(s) \) با درجه مخرج آن برابر باشد. از آنجا که درجه مخرج برابر با 2 است، نتیجه می‌گیریم که برای اینکه معادله (9.138) برقرار باشد، \( q(s) \) باید یک ثابت باشد—یعنی \( q(s) = K \). با استفاده از این ثابت می‌توانیم مقدار آن را محاسبه کنیم:
</p>

<p align="center">
$$
\lim_{s \to \infty} \frac{Ks^2}{s^2 - 25s - 8} = K.
$$
</p>

<p>
با معادل قرار دادن معادلات (9.138) و (9.139)، می‌بینیم که \( K = 4 \) و بنابراین،
</p>

<p align="center">
$$
H(s) = \frac{4s}{(s + 2)(s-4)}.
$$
</p>

<p>
مثال 9.27:
</p>

<p>
یک سیستم پایدار و علی با پاسخ ضربه‌ای \( h(t) \) و تابع سیستم \( H(s) \) را در نظر بگیرید. فرض کنید که \( H(s) \) معقول است، شامل یک قطب در \( s = 2 \) است و در مبدا صفر ندارد. موقعیت سایر قطب‌ها و صفرها مشخص نیست. برای هر یک از عبارات زیر، اجازه دهید تعیین کنیم که آیا می‌توانیم به‌طور قطعی بگوییم که درست است، نادرست است یا اطلاعات کافی برای تعیین صحت آن وجود ندارد:
</p>

<ul>
  <li>(الف) \( \mathcal{F}\{ h(t) e^{3t} \} \) همگرا است.</li>
  <li>(ب) \( \int_{-\infty}^{\infty} h(t) \, dt = 0 \).</li>
  <li>(ج) \( t h(t) \) پاسخ ضربه‌ای یک سیستم علی و پایدار است.</li>
</ul>

</div>

<div dir="rtl" align="justify">

<p>
(d) \( \frac{d h(t)}{dt} \) حداقل یک قطب در تبدیل لاپلاس خود دارد.
</p>

<p>
(e) \( h(t) \) دارای مدت زمان محدود است.
</p>

<p>
(f) \( H(s) = H(-s) \).
</p>

<p>
(g) \( \lim_{x \to \infty} H(s) = 2 \).
</p>

<p>
عبارت (الف) نادرست است، زیرا \( \{h(t)e^{3t}\} \) معادل با مقدار تبدیل لاپلاس \( h(t) \) در \( s = -3 \) است. اگر این همگرا باشد، به این معناست که \( s = -3 \) در ناحیه همگرایی (ROC) قرار دارد. یک سیستم علی و پایدار باید همیشه ناحیه همگرایی خود را به راست‌ترین قطب‌های خود داشته باشد. با این حال، \( s = 3 \) در راستای قطب در \( s = -2 \) قرار ندارد.
</p>

<p>
عبارت (ب) نادرست است، زیرا معادل با این است که بگوییم \( H(0) = 0 \). این با حقیقتی که می‌گوید \( H(s) \) در مبدا صفر ندارد مغایرت دارد.
</p>

<p>
عبارت (ج) درست است. طبق جدول 9.1، خصوصیت ذکر شده در بخش 9.5.8، تبدیل لاپلاس \( t h(t) \) همان ناحیه همگرایی را دارد که \( H(s) \). این ناحیه همگرایی شامل محور \( jw \) است و بنابراین، سیستم مربوطه پایدار است. همچنین، از آنجا که \( h(t) = 0 \) برای \( t < 0 \) است، این بدان معناست که \( t h(t) = 0 \) برای \( t < 0 \) است. بنابراین، \( t h(t) \) پاسخ ضربه‌ای یک سیستم علی است.
</p>

<p>
عبارت (د) درست است. طبق جدول 9.1، تبدیل لاپلاس \( \frac{d h(t)}{dt} \) برابر با \( s H(s) \) است. ضرب در \( s \) قطب در \( s = -2 \) را از بین نمی‌برد.
</p>

<p>
عبارت (ه) نادرست است. اگر \( h(t) \) دارای مدت زمان محدود باشد، آنگاه اگر تبدیل لاپلاس آن هر نقطه‌ای در ناحیه همگرایی خود داشته باشد، ناحیه همگرایی باید تمام صفحه \( s \) باشد. اما این با حقیقتی که می‌گوید \( H(s) \) یک قطب در \( s = -2 \) دارد مغایرت دارد.
</p>

<p>
عبارت (و) نادرست است. اگر این درست بود، پس از آنکه \( H(s) \) یک قطب در \( s = 2 \) دارد، باید همچنین یک قطب در \( s = -2 \) داشته باشد. این با حقیقتی که می‌گوید تمام قطب‌های یک سیستم علی و پایدار باید در نیم‌صفحه چپ صفحه \( s \) قرار داشته باشند مغایرت دارد.
</p>

<p>
صحت عبارت (ز) با اطلاعات داده‌شده قابل تعیین نیست. این عبارت نیاز دارد که درجه صورت و مخرج \( H(s) \) برابر باشد، و اطلاعات کافی برای تعیین این مورد در اختیار نداریم.
</p>

<p>
9.7.5 فیلترهای باتروورث
</p>

<p>
در مثال 6.3، به طور مختصر کلاس معروف سیستم‌های LTI به نام فیلترهای باتروورث را معرفی کردیم. فیلترهای این کلاس تعدادی ویژگی دارند که از جمله آنها می‌توان به ویژگی‌های بزرگ‌نمایی پاسخ فرکانسی هر یک از این فیلترها در محدوده عبوری اشاره کرد، که آنها را برای پیاده‌سازی عملی جذاب می‌کند. به‌عنوان یک نمونه بیشتر از کاربرد تبدیل‌های لاپلاس، در این بخش از تکنیک‌های تبدیل لاپلاس برای تعیین تابع سیستم یک فیلتر باتروورث از مشخصات بزرگ‌نمایی پاسخ فرکانسی آن استفاده می‌کنیم.
</p>

<p>
یک فیلتر پایین‌گذر باتروورث از مرتبه \( N \) دارای یک پاسخ فرکانسی است که مربع اندازه آن به‌صورت زیر داده شده است:
</p>

<p align="center">
$$
B(j\omega)^2 = \frac{1}{1 + \left(\frac{j\omega}{j\omega_c}\right)^2}
$$
</p>

<p>
که در آن \( N \) مرتبه فیلتر است. از معادله (9.140)، ما می‌خواهیم تابع سیستم \( B(s) \) را پیدا کنیم که موجب ایجاد \( |B(j\omega)|^2 \) می‌شود. ابتدا توجه می‌کنیم که بر اساس تعریف،
</p>

<p align="center">
$$
B(j\omega)^2 = B(j\omega) B^*(j\omega).
$$
</p>

</div>

<div dir="rtl" align="justify">

<p>
اگر پاسخ ضربه‌ای فیلتر باتروورث را محدود کنیم تا حقیقی باشد، آنگاه از خاصیت تقارن مزدوج در تبدیل فوریه،
</p>

<p align="center">
$$
B^*(j\omega) = B(-j\omega),
$$
</p>

<p>
بنابراین
</p>

<p align="center">
$$
B(j\omega)B(-j\omega) = \frac{1}{1 + \left(\frac{j\omega}{j\omega_c}\right)^2}
$$
</p>

<p>
اکنون توجه می‌کنیم که \( B(s) \) در \( s = j\omega \) برابر با \( B(j\omega) \) است و بنابراین، از معادله (9.143) به دست می‌آید:
</p>

<p align="center">
$$
B(s)B(-s) = \frac{1}{1 + \left(\frac{s}{j\omega_c}\right)^2}
$$
</p>

<p>
ریشه‌های چند جمله‌ای مخرج که مربوط به قطب‌های ترکیبی \( B(s)B(-s) \) هستند در
</p>

<p align="center">
$$
s = \left(-1\right)^{1/2N} j\omega_c.
$$
</p>

<p>
معادله (9.145) برای هر مقدار \( s = s_p \) که
</p>

<p align="center">
$$
s_p = \omega_c \exp \left( j\left( \pi \left( 2k + 1 \right) \right) \frac{1}{2N} + \frac{\pi}{2} \right),
$$
</p>

<p>
که در آن \( k \) یک عدد صحیح است.
</p>

<p>
در شکل 9.28 موقعیت قطب‌های \( B(s)B(-s) \) برای \( N = 1 \), \( N = 2 \), \( N = 3 \) و \( N = 6 \) نشان داده شده است. به طور کلی، ملاحظات زیر را می‌توان درباره این قطب‌ها بیان کرد:
</p>

<ol>
  <li> تعداد ۲N قطب به صورت یکنواخت در زاویه‌ای بر روی یک دایره با شعاع \( \omega_c \) در صفحه \( s \) قرار دارند.</li>
  <li> قطبی هرگز روی محور \( jw \) قرار ندارد و برای \( N \) فرد، در محور \( o \) قرار دارد، اما برای \( N \) زوج، قرار ندارد.</li>
  <li> فاصله زاویه‌ای بین قطب‌های \( B(s)B(-s) \) برابر با \( \frac{\pi}{N} \) رادیان است.</li>
</ol>

<p>
برای تعیین قطب‌های \( B(s) \) با توجه به قطب‌های \( B(s)B(-s) \)، مشاهده می‌کنیم که قطب‌های \( B(s)B(-s) \) به صورت جفت‌جفت ظاهر می‌شوند، بنابراین اگر قطبی در \( s = s_p \) وجود داشته باشد، قطب دیگری در \( s = -s_p \) نیز وجود دارد. بنابراین، برای ساخت \( B(s) \)، یکی از قطب‌ها از هر جفت را انتخاب می‌کنیم. اگر سیستم را محدود به سیستم‌های پایدار و علی کنیم، آنگاه قطب‌هایی که با \( B(s) \) مرتبط هستند، قطب‌های واقع در نیم‌دایره در نیمه‌صفحه چپ صفحه \( s \) خواهند بود. مکان‌های قطب‌ها فقط تا یک ضریب مقیاس مشخص می‌کنند. با این حال، از معادله (9.144) می‌بینیم که
</p>

<p align="center">
$$
B^2(s) \Big|_{s = 0} = 1,
$$
</p>

<p>
یا معادل آن، از معادله (9.140)، ضریب مقیاس به گونه‌ای انتخاب می‌شود که مربع اندازه پاسخ فرکانسی در \( \omega = 0 \) دارای بهره یک باشد. برای نمایش نحوه تعیین \( B(s) \)، بیایید به موارد \( N = 1 \), \( N = 2 \) و \( N = 3 \) توجه کنیم. در شکل 9.28 قطب‌های \( B(s)B(-s) \) که از معادله (9.148) به دست آمده است، نشان داده شده است.
</p>

</p>

$$
</p>

<p>

</p>

<p align="center">
$$
H(s) = H_1(s) + H_2(s).
$$
</p>

<p>
به طور مشابه، پاسخ ضربه‌ای برای اتصال سری در شکل 9.30(b) به صورت زیر است:
</p>

<p align="center">
$$
h(t) = h_1(t) * h_2(t),
$$
</p>

<p>
و تابع سیستم مرتبط به آن به صورت زیر خواهد بود:
</p>

<p align="center">
$$
H(s) = H_1(s) H_2(s).
$$
</p>

<p>
در شکل‌های 9.30، ترکیب‌های موازی و سری دو سیستم LTI نشان داده شده‌اند.
</p>

<p>

</p>

<p align="center">
$$

$$
</p>

<p>
در ادامه به تحلیل اتصال فیدبک دو سیستم LTI پرداخته‌ایم. از شکل 9.31، داریم:
</p>

<p align="center">

</p>

<p align="center">
$$

$$
</p>

<p>
از این رو، رابطه زیر را به دست می‌آوریم:
</p>

<p align="center">

Y(s) = H_1(s) \left[ X(s) - H_2(s) Y(s) \right],
$$
</p>

<p>
که از آن نتیجه می‌شود:
</p>

<h1>پاسخ‌های فیلتر باتروورث</h1>

<div dir="rtl" align="justify">

<p>
اگر پاسخ ضربه‌ای فیلتر باتروورث را محدود کنیم تا حقیقی باشد، آنگاه از خاصیت تقارن مزدوج در تبدیل فوریه،
</p>

<p class="math">
$$
B^*(j\omega) = B(-j\omega),
$$
</p>

<p>
بنابراین
</p>

<p class="math">
$$
B(j\omega)B(-j\omega) = \frac{1}{1 + \left(\frac{j\omega}{j\omega_c}\right)^2}
$$
</p>

<p>
اکنون توجه می‌کنیم که \( B(s) \) در \( s = j\omega \) برابر با \( B(j\omega) \) است و بنابراین، از معادله (9.143) به دست می‌آید:
</p>

<p class="math">
$$
B(s)B(-s) = \frac{1}{1 + \left(\frac{s}{j\omega_c}\right)^2}
$$
</p>

<p>
ریشه‌های چند جمله‌ای مخرج که مربوط به قطب‌های ترکیبی \( B(s)B(-s) \) هستند در
</p>

<p class="math">
$$
s = \left(-1\right)^{1/2N} j\omega_c.
$$
</p>

<p>
معادله (9.145) برای هر مقدار \( s = s_p \) که
</p>

<p class="math">
$$
s_p = \omega_c \exp \left( j\left( \pi \left( 2k + 1 \right) \right) \frac{1}{2N} + \frac{\pi}{2} \right),
$$
</p>

<p>
که در آن \( k \) یک عدد صحیح است.
</p>

<p>
در شکل 9.28 موقعیت قطب‌های \( B(s)B(-s) \) برای \( N = 1 \), \( N = 2 \), \( N = 3 \) و \( N = 6 \) نشان داده شده است. به طور کلی، ملاحظات زیر را می‌توان درباره این قطب‌ها بیان کرد:
</p>

<ol>
  <li> تعداد ۲N قطب به صورت یکنواخت در زاویه‌ای بر روی یک دایره با شعاع \( \omega_c \) در صفحه \( s \) قرار دارند.</li>
  <li> قطبی هرگز روی محور \( jw \) قرار ندارد و برای \( N \) فرد، در محور \( o \) قرار دارد، اما برای \( N \) زوج، قرار ندارد.</li>
  <li> فاصله زاویه‌ای بین قطب‌های \( B(s)B(-s) \) برابر با \( \frac{\pi}{N} \) رادیان است.</li>
</ol>

<p>
برای تعیین قطب‌های \( B(s) \) با توجه به قطب‌های \( B(s)B(-s) \)، مشاهده می‌کنیم که قطب‌های \( B(s)B(-s) \) به صورت جفت‌جفت ظاهر می‌شوند، بنابراین اگر قطبی در \( s = s_p \) وجود داشته باشد، قطب دیگری در \( s = -s_p \) نیز وجود دارد. بنابراین، برای ساخت \( B(s) \)، یکی از قطب‌ها از هر جفت را انتخاب می‌کنیم. اگر سیستم را محدود به سیستم‌های پایدار و علی کنیم، آنگاه قطب‌هایی که با \( B(s) \) مرتبط هستند، قطب‌های واقع در نیم‌دایره در نیمه‌صفحه چپ صفحه \( s \) خواهند بود. مکان‌های قطب‌ها فقط تا یک ضریب مقیاس مشخص می‌کنند. با این حال، از معادله (9.144) می‌بینیم که
</p>

<p class="math">
$$
B^2(s) \Big|_{s = 0} = 1,
$$
</p>

<p>
یا معادل آن، از معادله (9.140)، ضریب مقیاس به گونه‌ای انتخاب می‌شود که مربع اندازه پاسخ فرکانسی در \( \omega = 0 \) دارای بهره یک باشد. برای نمایش نحوه تعیین \( B(s) \)، بیایید به موارد \( N = 1 \), \( N = 2 \) و \( N = 3 \) توجه کنیم. در شکل 9.28 قطب‌های \( B(s)B(-s) \) که از معادله (9.148) به دست آمده است، نشان داده شده است.
</p>

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# تعریف متغیرها
s, t, A, alpha, beta, t0 = sp.symbols('s t A alpha beta t0')

# سوال 9.2 (الف) - تبدیل لاپلاس x(t) = e^(-alpha*(t-1)) * u(t-1)
x_t_1 = sp.exp(-alpha*(t-1)) * sp.Heaviside(t-1)
X_s_1 = sp.laplace_transform(x_t_1, t, s)
X_s_result_1 = X_s_1[0]

# سوال 9.3 - تبدیل لاپلاس x(t) = e^(alpha*t) * u(t) + e^(beta*t) * u(t)
x_t_2 = sp.exp(alpha*t) * sp.Heaviside(t) + sp.exp(beta*t) * sp.Heaviside(t)
X_s_2 = sp.laplace_transform(x_t_2, t, s)
X_s_result_2 = X_s_2[0]

# رسم نمودار برای سیگنال‌های مختلف
t_vals = np.linspace(0, 10, 400)
x_vals_1 = np.exp(-1*(t_vals-1)) * (t_vals >= 1)  # سیگنال x(t) = e^(-alpha(t-1)) * u(t-1)
x_vals_2 = np.exp(1*t_vals) * (t_vals >= 0) + np.exp(2*t_vals) * (t_vals >= 0)  # سیگنال x(t) = e^(alpha*t) * u(t) + e^(beta*t) * u(t)

# رسم سیگنال‌ها
plt.figure(figsize=(12, 6))

# رسم سیگنال 9.2 (الف)
plt.subplot(1, 2, 1)
plt.plot(t_vals, x_vals_1, label="x(t) = e^(-alpha(t-1)) * u(t-1)", color='b')
plt.title("سیگنال x(t) = e^(-alpha(t-1)) * u(t-1)")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)

# رسم سیگنال 9.3
plt.subplot(1, 2, 2)
plt.plot(t_vals, x_vals_2, label="x(t) = e^(alpha*t) * u(t) + e^(beta*t) * u(t)", color='r')
plt.title("سیگنال x(t) = e^(alpha*t) * u(t) + e^(beta*t) * u(t)")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
